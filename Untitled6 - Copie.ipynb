```python
import geopandas as gpd
import networkx as nx
import numpy as np
from shapely.geometry import Point, LineString, MultiLineString
import random
import matplotlib.pyplot as plt

# === 1.   ===
gdf_route = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_building = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

#    UTM   
gdf_route = gdf_route.to_crs(epsg=32631)
gdf_building = gdf_building.to_crs(epsg=32631)

# === 2.       ===
G = nx.Graph()

for _, row in gdf_route.iterrows():
    geom = row.geometry

    if isinstance(geom, LineString):
        lines = [geom]
    elif isinstance(geom, MultiLineString):
        lines = list(geom.geoms)
    else:
        continue  #      

    for line in lines:
        coords = list(line.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i + 1]
            edge_line = LineString([p1, p2])

            #      
            if not any(gdf_building.intersects(edge_line)):
                dist = np.linalg.norm(np.array(p1) - np.array(p2))
                G.add_edge(p1, p2, weight=dist)

# === 3.   PSO ===
class Particle:
    def __init__(self, graph, start, end):
        self.graph = graph
        self.start = start
        self.end = end
        self.path = self.random_path()
        self.best_path = self.path
        self.best_cost = self.path_cost(self.path)
        self.velocity = []

    def random_path(self):
        try:
            return nx.shortest_path(self.graph, self.start, self.end, weight=None)
        except:
            return []

    def path_cost(self, path):
        cost = 0
        for i in range(len(path) - 1):
            if self.graph.has_edge(path[i], path[i+1]):
                cost += self.graph[path[i]][path[i+1]]['weight']
            else:
                cost += 1e6  #     
        return cost

    def update_velocity(self, global_best):
        if not global_best:
            return
        new_path = []
        for a, b in zip(self.path, global_best):
            new_path.append(random.choice([a, b]))
        self.velocity = new_path

    def update_position(self):
        self.path = self.velocity
        cost = self.path_cost(self.path)
        if cost < self.best_cost:
            self.best_path = self.path
            self.best_cost = cost

# === 4.  PSO ===
def pso_path(graph, start, end, num_particles=30, iterations=50):
    swarm = [Particle(graph, start, end) for _ in range(num_particles)]
    global_best = min(swarm, key=lambda p: p.best_cost).best_path

    for _ in range(iterations):
        for particle in swarm:
            particle.update_velocity(global_best)
            particle.update_position()

        current_best = min(swarm, key=lambda p: p.best_cost)
        if current_best.path_cost(current_best.path) < Particle(graph, start, end).path_cost(global_best):
            global_best = current_best.best_path

    return global_best

# === 5.     ===
#         
point_a = list(G.nodes)[0]  #    
point_b = list(G.nodes)[-1]  #  

# === 6.   ===
shortest_path = pso_path(G, point_a, point_b)

# === 7.   ===
gdf_route.plot(color='gray', linewidth=0.5)
gdf_building.plot(color='red', alpha=0.4)

if shortest_path:
    path_line = LineString(shortest_path)
    gdf_path = gpd.GeoSeries([path_line], crs=gdf_route.crs)
    gdf_path.plot(color='blue', linewidth=3)

plt.title("Shortest Path using PSO in Bjaa")
plt.xlabel("Easting")
plt.ylabel("Northing")
plt.show()
```
