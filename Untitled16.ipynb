```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
import time
import math

# ===================================================================
# 1. CONFIGURATION AND HYPERPARAMETERS
# ===================================================================

# --- Environment Setup ---
START_POS = np.array([5, 5])
GOAL_POS = np.array([95, 95])
OBSTACLES = [
    # (center_x, center_y, radius)
    (30, 40, 10),
    (60, 60, 15),
    (75, 25, 8)
]
N_WAYPOINTS = 10 

# Search space boundaries for the waypoints
SEARCH_BOUNDS_X = (0, 100)
SEARCH_BOUNDS_Y = (0, 100)

# --- MGWO Hyperparameters ---
N_WOLVES = 50
N_ITERATIONS = 500
OBSTACLE_PENALTY = 1000 # Cost penalty for hitting an obstacle

# --- Levy Flight Parameters for the "M"odification ---
LEVY_BETA = 1.5


# ===================================================================
# 2. HELPER FUNCTIONS AND OBJECTIVE FUNCTION
# ===================================================================

def line_intersects_circle(p1, p2, circle_center, radius):
    """Checks if the line segment from p1 to p2 intersects a circle."""
    p1, p2, circle_center = np.array(p1), np.array(p2), np.array(circle_center)
    d, f = p2 - p1, p1 - circle_center
    a = np.dot(d, d)
    b = 2 * np.dot(f, d)
    c = np.dot(f, f) - radius**2
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return False
    else:
        t1 = (-b - np.sqrt(discriminant)) / (2*a)
        t2 = (-b + np.sqrt(discriminant)) / (2*a)
        return 0 <= t1 <= 1 or 0 <= t2 <= 1

def calculate_path_cost(waypoints_flat):
    """The objective function. Calculates the cost of a path."""
    waypoints = waypoints_flat.reshape(N_WAYPOINTS, 2)
    full_path = np.vstack([START_POS, waypoints, GOAL_POS])
    
    path_length = np.sum(np.sqrt(np.sum(np.diff(full_path, axis=0)**2, axis=1)))
    
    collision_cost = 0
    for i in range(len(full_path) - 1):
        p1, p2 = full_path[i], full_path[i+1]
        for ox, oy, r in OBSTACLES:
            if line_intersects_circle(p1, p2, (ox, oy), r):
                collision_cost += OBSTACLE_PENALTY
                
    return path_length + collision_cost

def get_levy_step():
    """Generates a step size based on the Levy distribution for exploration."""
    sigma_u = (math.gamma(1 + LEVY_BETA) * np.sin(np.pi * LEVY_BETA / 2) / 
               (math.gamma((1 + LEVY_BETA) / 2) * LEVY_BETA * 2**((LEVY_BETA - 1) / 2)))**(1 / LEVY_BETA)
    sigma_v = 1
    u = np.random.normal(0, sigma_u, 1)
    v = np.random.normal(0, sigma_v, 1)
    step = u / (np.abs(v)**(1 / LEVY_BETA))
    return 0.01 * step # Scale down the step

def plot_environment(path=None, title="MGWO Path Planning"):
    """Visualizes the environment, obstacles, and optionally a path."""
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')

    for obs in OBSTACLES:
        circle = patches.Circle((obs[0], obs[1]), obs[2], color='b', alpha=0.5)
        ax.add_patch(circle)

    if path is not None:
        path_points = np.vstack([START_POS, path, GOAL_POS])
        ax.plot(path_points[:, 0], path_points[:, 1], 'y-', label='Path')
        ax.plot(path_points[:, 0], path_points[:, 1], 'k.', markersize=5)

    ax.set_xlim(SEARCH_BOUNDS_X)
    ax.set_ylim(SEARCH_BOUNDS_Y)
    ax.set_title(title)
    ax.legend()
    ax.grid(True)
    return fig, ax

# ===================================================================
# 3. MODIFIED GREY WOLF OPTIMIZER ALGORITHM
# ===================================================================

def mgwo_path_planning():
    """The main MGWO algorithm logic."""
    n_dimensions = N_WAYPOINTS * 2
    min_bound = np.array([SEARCH_BOUNDS_X[0], SEARCH_BOUNDS_Y[0]] * N_WAYPOINTS)
    max_bound = np.array([SEARCH_BOUNDS_X[1], SEARCH_BOUNDS_Y[1]] * N_WAYPOINTS)
    
    # --- INITIALIZATION ---
    positions = np.random.rand(N_WOLVES, n_dimensions) * (max_bound - min_bound) + min_bound
    
    # Initialize Alpha, Beta, and Delta wolves
    alpha_pos, beta_pos, delta_pos = np.zeros(n_dimensions), np.zeros(n_dimensions), np.zeros(n_dimensions)
    alpha_score, beta_score, delta_score = float('inf'), float('inf'), float('inf')
    
    history = []

    print("\nStarting MGWO algorithm...")
    start_time = time.time()

    # --- MAIN LOOP ---
    for i in range(N_ITERATIONS):
        # 1. Calculate fitness for each wolf
        for j in range(N_WOLVES):
            fitness = calculate_path_cost(positions[j])
            if fitness < alpha_score:
                alpha_score, alpha_pos = fitness, positions[j].copy()
            elif alpha_score < fitness < beta_score:
                beta_score, beta_pos = fitness, positions[j].copy()
            elif beta_score < fitness < delta_score:
                delta_score, delta_pos = fitness, positions[j].copy()
        
        # 2. Update the 'a' parameter (linearly from 2 to 0)
        a = 2 - i * (2 / N_ITERATIONS)
        
        # 3. Update positions of Omega wolves
        for j in range(N_WOLVES):
            r1, r2 = np.random.rand(n_dimensions), np.random.rand(n_dimensions)
            A1, C1 = 2 * a * r1 - a, 2 * r2
            D_alpha = np.abs(C1 * alpha_pos - positions[j])
            X1 = alpha_pos - A1 * D_alpha
            
            r1, r2 = np.random.rand(n_dimensions), np.random.rand(n_dimensions)
            A2, C2 = 2 * a * r1 - a, 2 * r2
            D_beta = np.abs(C2 * beta_pos - positions[j])
            X2 = beta_pos - A2 * D_beta
            
            r1, r2 = np.random.rand(n_dimensions), np.random.rand(n_dimensions)
            A3, C3 = 2 * a * r1 - a, 2 * r2
            D_delta = np.abs(C3 * delta_pos - positions[j])
            X3 = delta_pos - A3 * D_delta
            
            # Update wolf position based on the average of the leaders
            positions[j] = (X1 + X2 + X3) / 3

        # 4. (The "M" Part) Apply Levy Flight to the Alpha wolf to escape local minima
        levy_step = get_levy_step() * (alpha_pos - beta_pos) # Use difference for direction
        alpha_pos += levy_step
        
        # 5. Boundary checking
        positions = np.clip(positions, min_bound, max_bound)
        alpha_pos = np.clip(alpha_pos, min_bound, max_bound)
        
        history.append(alpha_pos.copy())
        if (i + 1) % 50 == 0:
            print(f"Iteration {i+1}/{N_ITERATIONS}: Best Cost (Alpha) = {alpha_score:.2f}")

    end_time = time.time()
    print(f"\nFinished! Total time: {end_time - start_time:.2f} seconds.")
    print(f"Final Best Cost = {alpha_score:.2f}")
    return alpha_pos, history

# ===================================================================
# 4. MAIN EXECUTION BLOCK
# ===================================================================

if __name__ == "__main__":
    # --- Step 1: Show the initial environment ---
    fig, ax = plot_environment(title="Initial Environment Setup")
    plt.show()

    # --- Step 2: Run the MGWO algorithm ---
    best_path_flat, history = mgwo_path_planning()
    
    # --- Step 3: Show the final best path ---
    final_path_waypoints = best_path_flat.reshape(N_WAYPOINTS, 2)
    fig, ax = plot_environment(path=final_path_waypoints, title="Final Best Path Found by MGWO")
    plt.show()

    # --- Step 4: Create and save the convergence animation ---
    print("\nCreating and saving animation... (this may take a moment)")
    fig, ax = plot_environment(title="MGWO Path Evolution")
    
    initial_waypoints = history[0].reshape(N_WAYPOINTS, 2)
    initial_path_points = np.vstack([START_POS, initial_waypoints, GOAL_POS])
    line, = ax.plot(initial_path_points[:, 0], initial_path_points[:, 1], 'y-', label='Path')
    dots, = ax.plot(initial_waypoints[:, 0], initial_waypoints[:, 1], 'k.', markersize=5)
    title_text = ax.text(0.5, 1.01, '', ha='center', va='bottom', transform=ax.transAxes, fontsize=12)

    def update_animation(frame):
        waypoints = history[frame].reshape(N_WAYPOINTS, 2)
        path = np.vstack([START_POS, waypoints, GOAL_POS])
        line.set_data(path[:, 0], path[:, 1])
        dots.set_data(waypoints[:, 0], waypoints[:, 1])
        title_text.set_text(f"MGWO Path Evolution (Iteration {frame})")
        return line, dots, title_text

    ani = FuncAnimation(fig, update_animation, frames=N_ITERATIONS, blit=False, interval=50)

    try:
        ani.save('mgwo_path_animation.mp4', writer='ffmpeg', dpi=100)
        print("Animation successfully saved to 'mgwo_path_animation.mp4'")
    except Exception as e:
        print(f"Error saving animation: {e}")
        print("Please ensure you have ffmpeg installed and accessible in your system's PATH.")
    
    plt.close(fig)
```
