```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
import time

# ===================================================================
# 1. CONFIGURATION AND HYPERPARAMETERS
# ===================================================================

# --- Environment Setup ---
START_POS = np.array([5, 5])
GOAL_POS = np.array([95, 95])
OBSTACLES = [
    # (center_x, center_y, radius)
    (30, 40, 10),
    (60, 60, 15),
    (75, 25, 8)
]
N_WAYPOINTS = 10 

# Search space boundaries for the waypoints
SEARCH_BOUNDS_X = (0, 100)
SEARCH_BOUNDS_Y = (0, 100)

# --- BFO Hyperparameters ---
# The total number of evaluations is N_ED * N_REPRO * N_CHEMO * N_BACTERIA
N_BACTERIA = 50      # Number of bacteria in the colony (population size)
N_CHEMO = 20         # Number of chemotactic steps
N_REPRO = 10         # Number of reproduction steps
N_ED = 5             # Number of elimination-dispersal events
N_RUN = 4            # Max steps in a single run direction
STEP_SIZE = 0.1      # Size of a chemotactic step (C_i)
PED = 0.25           # Elimination-dispersal probability

OBSTACLE_PENALTY = 1000 # Cost penalty for hitting an obstacle


# ===================================================================
# 2. HELPER FUNCTIONS AND OBJECTIVE FUNCTION
# ===================================================================

def line_intersects_circle(p1, p2, circle_center, radius):
    p1, p2, circle_center = np.array(p1), np.array(p2), np.array(circle_center)
    d, f = p2 - p1, p1 - circle_center
    a = np.dot(d, d)
    b = 2 * np.dot(f, d)
    c = np.dot(f, f) - radius**2
    discriminant = b**2 - 4*a*c
    if discriminant < 0: return False
    t1 = (-b - np.sqrt(discriminant)) / (2*a); t2 = (-b + np.sqrt(discriminant)) / (2*a)
    return 0 <= t1 <= 1 or 0 <= t2 <= 1

def calculate_path_cost(waypoints_flat):
    waypoints = waypoints_flat.reshape(N_WAYPOINTS, 2)
    full_path = np.vstack([START_POS, waypoints, GOAL_POS])
    path_length = np.sum(np.sqrt(np.sum(np.diff(full_path, axis=0)**2, axis=1)))
    collision_cost = 0
    for i in range(len(full_path) - 1):
        for ox, oy, r in OBSTACLES:
            if line_intersects_circle(full_path[i], full_path[i+1], (ox, oy), r):
                collision_cost += OBSTACLE_PENALTY
    return path_length + collision_cost

def plot_environment(path=None, title="BFO Path Planning"):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')
    for ox, oy, r in OBSTACLES:
        circle = patches.Circle((ox, oy), r, color='b', alpha=0.3)
        ax.add_patch(circle)
    if path is not None:
        path_points = np.vstack([START_POS, path, GOAL_POS])
        ax.plot(path_points[:, 0], path_points[:, 1], 'y-', label='Path')
        ax.plot(path_points[:, 0], path_points[:, 1], 'k.', markersize=5)
    ax.set_xlim(SEARCH_BOUNDS_X); ax.set_ylim(SEARCH_BOUNDS_Y)
    ax.set_title(title); ax.legend(); ax.grid(True)
    return fig, ax

# ===================================================================
# 3. BACTERIAL FORAGING OPTIMIZATION ALGORITHM
# ===================================================================

def bfo_path_planning():
    """The main BFO algorithm logic."""
    n_dimensions = N_WAYPOINTS * 2
    min_bound = np.array([SEARCH_BOUNDS_X[0], SEARCH_BOUNDS_Y[0]] * N_WAYPOINTS)
    max_bound = np.array([SEARCH_BOUNDS_X[1], SEARCH_BOUNDS_Y[1]] * N_WAYPOINTS)
    
    # --- INITIALIZATION ---
    positions = np.random.rand(N_BACTERIA, n_dimensions) * (max_bound - min_bound) + min_bound
    
    best_pos_global = None
    best_cost_global = float('inf')
    history = []

    print("\nStarting BFO algorithm...")
    start_time = time.time()

    # --- ELIMINATION-DISPERSAL LOOP ---
    for l in range(N_ED):
        # --- REPRODUCTION LOOP ---
        for k in range(N_REPRO):
            health = np.zeros(N_BACTERIA)
            
            # --- CHEMOTAXIS LOOP ---
            for j in range(N_CHEMO):
                costs = np.array([calculate_path_cost(p) for p in positions])
                
                # Update health and global best
                health += costs
                best_idx_current_step = np.argmin(costs)
                if costs[best_idx_current_step] < best_cost_global:
                    best_cost_global = costs[best_idx_current_step]
                    best_pos_global = positions[best_idx_current_step].copy()

                # Tumble and run for each bacterium
                for i in range(N_BACTERIA):
                    # Tumble: generate a random direction
                    delta = np.random.uniform(-1, 1, n_dimensions)
                    direction = delta / np.linalg.norm(delta)
                    
                    # Run: move in that direction
                    new_pos = positions[i] + STEP_SIZE * direction
                    new_pos = np.clip(new_pos, min_bound, max_bound) # Boundary check
                    
                    # If life improves, keep trying to run in that direction
                    if calculate_path_cost(new_pos) < costs[i]:
                        positions[i] = new_pos

            # --- REPRODUCTION LOGIC ---
            sorted_indices = np.argsort(health)
            n_survivors = N_BACTERIA // 2
            
            # The unhealthiest half dies
            # The healthiest half reproduces
            survivor_positions = positions[sorted_indices[:n_survivors]]
            positions = np.vstack([survivor_positions, survivor_positions]) # Duplicate
            
            # Store the best path of this generation for animation
            history.append(best_pos_global.copy())
            print(f"E-D {l+1}/{N_ED}, Repro {k+1}/{N_REPRO}: Best Cost = {best_cost_global:.2f}")

        # --- ELIMINATION-DISPERSAL LOGIC ---
        for i in range(N_BACTERIA):
            if np.random.rand() < PED:
                positions[i] = np.random.rand(n_dimensions) * (max_bound - min_bound) + min_bound

    end_time = time.time()
    print(f"\nFinished! Total time: {end_time - start_time:.2f} seconds.")
    print(f"Final Best Cost = {best_cost_global:.2f}")
    return best_pos_global, history

# ===================================================================
# 4. MAIN EXECUTION BLOCK
# ===================================================================

if __name__ == "__main__":
    # --- Step 1: Run the BFO algorithm ---
    best_path_flat, history = bfo_path_planning()
    
    # --- Step 2: Show the final best path ---
    final_path_waypoints = best_path_flat.reshape(N_WAYPOINTS, 2)
    fig, ax = plot_environment(path=final_path_waypoints, title="Final Best Path Found by BFO")
    plt.show()

    # --- Step 3: Create and save the convergence animation ---
    print("\nCreating and saving animation... (this may take a moment)")
    fig, ax = plot_environment(title="BFO Path Evolution")
    
    initial_waypoints = history[0].reshape(N_WAYPOINTS, 2)
    initial_path_points = np.vstack([START_POS, initial_waypoints, GOAL_POS])
    line, = ax.plot(initial_path_points[:, 0], initial_path_points[:, 1], 'y-', label='Path')
    dots, = ax.plot(initial_waypoints[:, 0], initial_waypoints[:, 1], 'k.', markersize=5)
    title_text = ax.text(0.5, 1.01, '', ha='center', va='bottom', transform=ax.transAxes, fontsize=12)

    def update_animation(frame):
        waypoints = history[frame].reshape(N_WAYPOINTS, 2)
        path = np.vstack([START_POS, waypoints, GOAL_POS])
        line.set_data(path[:, 0], path[:, 1])
        dots.set_data(waypoints[:, 0], waypoints[:, 1])
        title_text.set_text(f"BFO Path Evolution (Generation {frame+1})")
        return line, dots, title_text

    ani = FuncAnimation(fig, update_animation, frames=len(history), blit=False, interval=100)

    try:
        ani.save('bfo_path_animation.mp4', writer='ffmpeg', dpi=100)
        print("Animation successfully saved to 'bfo_path_animation.mp4'")
    except Exception as e:
        print(f"Error saving animation: {e}")
        print("Please ensure you have ffmpeg installed and accessible in your system's PATH.")
        
    plt.close(fig)
```
