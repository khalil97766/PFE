```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
import time

# ===================================================================
# 1. CONFIGURATION AND HYPERPARAMETERS
# ===================================================================

# --- Environment Setup ---
START_POS = np.array([5.0, 5.0]) # Use float for precise movements
GOAL_POS = np.array([95.0, 95.0])
OBSTACLES = [
    # (center_x, center_y, radius)
    (30, 40, 10),
    (60, 60, 15),
    (75, 25, 8)
]
X_MIN, X_MAX = 0, 100
Y_MIN, Y_MAX = 0, 100

# --- APF Hyperparameters ---
# These require careful tuning!
ZETA = 1.0           # Attractive force gain
ETA = 100.0          # Repulsive force gain
RHO_0 = 20.0         # Obstacle influence radius (distance at which repulsive force starts)

STEP_SIZE = 1.0      # How far the robot moves in each step
MAX_ITER = 500       # Maximum number of steps to take
GOAL_THRESHOLD = 1.5 # How close to the goal to be considered "arrived"


# ===================================================================
# 2. FORCE CALCULATIONS AND HELPER FUNCTIONS
# ===================================================================

def calculate_attractive_force(current_pos, goal_pos):
    """Calculates the attractive force pulling the robot towards the goal."""
    return ZETA * (goal_pos - current_pos)

def calculate_repulsive_force(current_pos, obstacles):
    """Calculates the total repulsive force from all obstacles."""
    total_repulsive_force = np.zeros(2)
    for ox, oy, r in obstacles:
        obs_center = np.array([ox, oy])
        dist_to_center = np.linalg.norm(current_pos - obs_center)
        
        # Calculate distance to the obstacle's surface (rho)
        rho = dist_to_center - r
        
        # If outside the influence radius, this obstacle has no effect
        if rho > RHO_0:
            continue
        
        # If inside the obstacle (or very close), apply a huge force
        if rho <= 0.1:
            rho = 0.1
        
        # Calculate the repulsive force for this single obstacle
        force_magnitude = ETA * ((1.0 / rho) - (1.0 / RHO_0)) * (1.0 / rho**2)
        force_direction = (current_pos - obs_center) / dist_to_center # Unit vector away from center
        
        total_repulsive_force += force_magnitude * force_direction
        
    return total_repulsive_force

def plot_environment(path=None, current_pos=None, title="APF Path Planning"):
    """Visualizes the environment, path, and current robot position."""
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')

    for ox, oy, r in OBSTACLES:
        circle = patches.Circle((ox, oy), r, color='b', alpha=0.3)
        ax.add_patch(circle)
        # Plot the influence radius
        influence_circle = patches.Circle((ox, oy), r + RHO_0, color='b', fill=False, linestyle='--', alpha=0.2)
        ax.add_patch(influence_circle)

    if path is not None and len(path) > 0:
        path_arr = np.array(path)
        ax.plot(path_arr[:, 0], path_arr[:, 1], 'y-', label='Path')

    if current_pos is not None:
        ax.plot(current_pos[0], current_pos[1], 'ms', markersize=10, label='Robot')

    ax.set_xlim(X_MIN, X_MAX)
    ax.set_ylim(Y_MIN, Y_MAX)
    ax.set_title(title)
    ax.legend()
    ax.grid(True)
    return fig, ax

# ===================================================================
# 3. ARTIFICIAL POTENTIAL FIELD ALGORITHM
# ===================================================================

def apf_path_planning():
    """The main APF path generation logic."""
    current_pos = np.copy(START_POS)
    path = [np.copy(current_pos)]
    
    print("\nStarting APF algorithm...")
    start_time = time.time()
    
    for i in range(MAX_ITER):
        # Check if goal is reached
        if np.linalg.norm(current_pos - GOAL_POS) < GOAL_THRESHOLD:
            print(f"Goal reached in {i+1} steps!")
            break

        # Calculate forces
        f_attractive = calculate_attractive_force(current_pos, GOAL_POS)
        f_repulsive = calculate_repulsive_force(current_pos, OBSTACLES)
        f_total = f_attractive + f_repulsive
        
        # Normalize the total force to get the direction of movement
        if np.linalg.norm(f_total) > 0:
            direction = f_total / np.linalg.norm(f_total)
        else:
            # If forces cancel out (local minimum), we are stuck
            print(f"Local minimum reached at step {i+1}. Cannot move.")
            break

        # Move the robot and update path
        current_pos += STEP_SIZE * direction
        path.append(np.copy(current_pos))

    else: # This 'else' belongs to the 'for' loop, executes if loop finishes without 'break'
        print(f"Algorithm finished after {MAX_ITER} steps (max iterations reached).")

    end_time = time.time()
    print(f"Total time: {end_time - start_time:.2f} seconds.")
    return path

# ===================================================================
# 4. MAIN EXECUTION BLOCK
# ===================================================================

if __name__ == "__main__":
    # --- Step 1: Run the APF algorithm ---
    path_history = apf_path_planning()
    
    # --- Step 2: Show the final path ---
    final_pos = path_history[-1]
    fig, ax = plot_environment(path=path_history, current_pos=final_pos, title="Final Path Found by APF")
    plt.show()

    # --- Step 3: Create and save the convergence animation ---
    print("\nCreating and saving animation... (this may take a moment)")
    fig, ax = plot_environment(title="APF Path Evolution")
    
    line, = ax.plot([], [], 'y-', label='Path')
    robot_marker, = ax.plot([], [], 'ms', markersize=10, label='Robot')
    title_text = ax.text(0.5, 1.01, '', ha='center', va='bottom', transform=ax.transAxes, fontsize=12)

    def update_animation(frame):
        # Draw the path up to the current frame
        path_so_far = np.array(path_history[:frame+1])
        line.set_data(path_so_far[:, 0], path_so_far[:, 1])
        
        # Update the robot's current position
        current_pos = path_history[frame]
        robot_marker.set_data([current_pos[0]], [current_pos[1]])
        
        title_text.set_text(f"APF Path Evolution (Step {frame})")
        return line, robot_marker, title_text

    ani = FuncAnimation(fig, update_animation, frames=len(path_history), blit=False, interval=50)

    try:
        ani.save('apf_path_animation.mp4', writer='ffmpeg', dpi=100)
        print("Animation successfully saved to 'apf_path_animation.mp4'")
    except Exception as e:
        print(f"Error saving animation: {e}")
        print("Please ensure you have ffmpeg installed and accessible in your system's PATH.")
        
    plt.close(fig)
```
