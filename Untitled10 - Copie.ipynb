```python
import geopandas as gpd
import networkx as nx
from shapely.geometry import LineString, Point
import matplotlib.pyplot as plt
import heapq

# Load data
gdf_route = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_building = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# Create graph from route lines
G = nx.Graph()

for idx, row in gdf_route.iterrows():
    geom = row.geometry
    if isinstance(geom, LineString):
        coords = list(geom.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i + 1]
            dist = Point(p1).distance(Point(p2))
            G.add_edge(p1, p2, weight=dist)

# Remove edges that intersect buildings
for u, v in list(G.edges()):
    line = LineString([u, v])
    if gdf_building.intersects(line).any():
        G.remove_edge(u, v)

# Define start and goal (first and last nodes in the graph)
start = list(G.nodes)[0]
goal = list(G.nodes)[-1]

# Define heuristic (Euclidean distance)
def heuristic(node, goal):
    return Point(node).distance(Point(goal))

# Greedy Best-First Search algorithm
def greedy_bfs(graph, start, goal):
    visited = set()
    came_from = {}
    queue = [(heuristic(start, goal), start)]
    heapq.heapify(queue)

    while queue:
        _, current = heapq.heappop(queue)
        if current == goal:
            # Reconstruct path
            path = [current]
            while current in came_from:
                current = came_from[current]
                path.append(current)
            return list(reversed(path))

        if current in visited:
            continue
        visited.add(current)

        for neighbor in graph.neighbors(current):
            if neighbor not in visited:
                came_from[neighbor] = current
                heapq.heappush(queue, (heuristic(neighbor, goal), neighbor))

    return None  # No path found

# Run Greedy BFS
path = greedy_bfs(G, start, goal)

# Plotting
fig, ax = plt.subplots()
gdf_route.plot(ax=ax, color='gray', linewidth=0.5)
gdf_building.plot(ax=ax, color='red', alpha=0.5)

if path:
    gpd.GeoSeries([LineString(path)]).plot(ax=ax, color='red', linewidth=2)
    plt.scatter(*start, color='green', label='Start')
    plt.scatter(*goal, color='purple', label='Goal')
    plt.legend()
else:
    print("No path found.")

plt.show()
```
