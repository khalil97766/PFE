```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from collections import deque
import random
import time # To measure performance

# --- 1. Environment and Simulation ---

# Constants
GOAL_POS = (25.0, 25.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 500
GOAL_RADIUS = 1.0
SEQUENCE_LENGTH = 5

# Obstacles represented as rectangles [x_min, y_min, x_max, y_max]
OBSTACLES = np.array([ # Use NumPy array for faster access
    [10, 0, 12, 15],
    [10, 20, 12, 30],
    [18, 10, 20, 25]
])

class Robot:
    def __init__(self, x, y, theta):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega):
        # Clip omega to its max/min values
        omega = np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)
        self.theta += omega * DT
        # Normalize angle to be within [-pi, pi]
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += ROBOT_SPEED * math.cos(self.theta) * DT
        self.y += ROBOT_SPEED * math.sin(self.theta) * DT
        return (self.x, self.y), self.theta


### --- OPTIMIZATION 1: Vectorized Lidar Simulation --- ###
# This version is significantly faster than the original because it avoids Python loops.
def get_lidar_readings_vectorized(robot_pos, robot_theta, num_sensors=5, max_dist=10.0, ray_step=0.1):
    """Simulates 5 Lidar sensors using NumPy for speed."""
    readings = np.full(num_sensors, max_dist)
    angles = np.linspace(-np.pi/2, np.pi/2, num_sensors) + robot_theta
    
    # Create all ray points at once
    dist_steps = np.arange(ray_step, max_dist, ray_step)
    
    # Calculate x, y coordinates for all points on all rays
    # Shape: (num_steps, num_sensors)
    ray_points_x = robot_pos[0] + dist_steps[:, np.newaxis] * np.cos(angles)
    ray_points_y = robot_pos[1] + dist_steps[:, np.newaxis] * np.sin(angles)

    # Check for collisions with all obstacles
    for obs in OBSTACLES:
        # Create a boolean mask for points inside the current obstacle
        in_obstacle = (
            (ray_points_x >= obs[0]) & (ray_points_x <= obs[2]) &
            (ray_points_y >= obs[1]) & (ray_points_y <= obs[3])
        )
        
        # Find the first collision distance for each ray
        # `argmax` finds the index of the first `True` value
        collision_indices = np.argmax(in_obstacle, axis=0)
        
        for i in range(num_sensors):
            if in_obstacle[collision_indices[i], i]: # If a collision was found for this ray
                dist = dist_steps[collision_indices[i]]
                # Update the reading only if this collision is closer
                if dist < readings[i]:
                    readings[i] = dist
                    
    return readings


def get_state(robot_pos, robot_theta):
    """Generates the 7-dimensional state vector for the neural network."""
    # Use the new, faster Lidar function
    lidar = get_lidar_readings_vectorized(robot_pos, robot_theta)
    
    dx = GOAL_POS[0] - robot_pos[0]
    dy = GOAL_POS[1] - robot_pos[1]
    
    dist_to_goal = math.sqrt(dx**2 + dy**2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = math.atan2(math.sin(angle_to_goal - robot_theta), math.cos(angle_to_goal - robot_theta))
    
    state = np.concatenate((lidar / 10.0, [dist_to_goal / 30.0, angle_diff / np.pi]))
    return state
    
# --- 2. The Expert and The Learner Models ---

def apf_expert_policy(state):
    """A simple APF-based expert. It's good but can get stuck."""
    lidar_readings = state[:5] * 10.0
    angle_to_goal = state[6] * np.pi
    
    repulsive_omega = 0.0
    weights = [0.1, 0.3, 1.0, 0.3, 0.1] 
    turn_directions = [np.pi/2, np.pi/4, 0, -np.pi/4, -np.pi/2]

    for i in range(5):
        if lidar_readings[i] < 3.0:
            repulsive_omega += weights[i] * (1.0 / max(lidar_readings[i], 0.1)) * turn_directions[i]

    attractive_omega = 2.0 * angle_to_goal
    total_omega = attractive_omega - repulsive_omega
    return np.clip(total_omega, -MAX_TURN_RATE, MAX_TURN_RATE)

def create_lstm_model(sequence_length, n_inputs, n_outputs):
    model = Sequential([
        Input(shape=(sequence_length, n_inputs)),
        LSTM(64, return_sequences=True),
        LSTM(64),
        Dense(32, activation='relu'),
        Dense(n_outputs, activation='tanh')
    ])
    # Use a faster, modern optimizer like AdamW if available, otherwise Adam is fine.
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    model.summary()
    return model

### --- OPTIMIZATION 2 & 3: Faster Prediction & @tf.function --- ###
# This function is compiled by TensorFlow for high performance.
# It calls the model directly, which is much faster than model.predict().
@tf.function
def predict_action(model, nn_input):
    """Performs a fast, compiled prediction."""
    # `training=False` is important for speed as it disables things like dropout.
    return model(nn_input, training=False)


def run_episode(start_pos, start_theta, policy_model, use_exploration=False, epsilon=0.0):
    """Runs a single episode, now using optimized components."""
    robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
    path_taken = [start_pos]
    state_history = deque(maxlen=SEQUENCE_LENGTH)
    
    episode_sequences = []
    episode_actions = []

    for _ in range(MAX_EPISODE_STEPS):
        pos = (robot.x, robot.y)
        
        current_state = get_state(pos, robot.theta)
        state_history.append(current_state)
        
        if len(state_history) < SEQUENCE_LENGTH:
            omega = 0.1 * (random.random() - 0.5)
        else:
            sequence = np.array(state_history)
            
            if policy_model == "expert":
                omega = apf_expert_policy(current_state)
            else: # It's a Keras model
                nn_input = np.reshape(sequence, (1, SEQUENCE_LENGTH, 7))
                nn_input_tensor = tf.convert_to_tensor(nn_input, dtype=tf.float32)

                # USE THE FAST PREDICTION FUNCTION
                normalized_omega_tensor = predict_action(policy_model, nn_input_tensor)
                normalized_omega = normalized_omega_tensor.numpy()[0][0]

                if use_exploration and random.random() < epsilon:
                    normalized_omega += random.uniform(-0.5, 0.5)
                
                omega = normalized_omega * MAX_TURN_RATE

            episode_sequences.append(sequence)
            episode_actions.append(omega / MAX_TURN_RATE)

        new_pos, _ = robot.move(omega)
        path_taken.append(new_pos)

        if math.dist(new_pos, GOAL_POS) < GOAL_RADIUS:
            return path_taken, True, episode_sequences, episode_actions

    return path_taken, False, episode_sequences, episode_actions


# --- 3. Main Execution Block ---

if __name__ == "__main__":
    
    ### --- OPTIMIZATION 4: Reduce episodes for faster testing --- ###
    # You can reduce these numbers for quicker runs, but the final model might be less optimal.
    NUM_EXPERT_EPISODES = 25 # Was 50
    NUM_RL_EPOCHS = 5        # Was 10
    EPISODES_PER_EPOCH = 15  # Was 20
    
    # =========================================================================
    # PART 1: Generate Training Data with the "Expert" (APF)
    # =========================================================================
    print("\n--- Part 1: Generating Data with APF Expert ---")
    start_time = time.time()
    
    all_expert_sequences = []
    all_expert_actions = []

    for i in range(NUM_EXPERT_EPISODES):
        print(f"Running expert episode {i+1}/{NUM_EXPERT_EPISODES}")
        start_pos = (random.uniform(0, 5), random.uniform(0, 5))
        start_theta = random.uniform(-np.pi, np.pi)
        _, success, sequences, actions = run_episode(start_pos, start_theta, "expert")
        if success:
            all_expert_sequences.extend(sequences)
            all_expert_actions.extend(actions)

    X_train_expert = np.array(all_expert_sequences)
    y_train_expert = np.array(all_expert_actions)
    print(f"Generated {X_train_expert.shape[0]} samples in {time.time() - start_time:.2f} seconds.")

    # =========================================================================
    # PART 2: Pre-train the LSTM model (Creating LSTM_FT)
    # =========================================================================
    if X_train_expert.shape[0] > 0:
        print("\n--- Part 2: Pre-training the LSTM_FT model ---")
        lstm_ft_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)
        lstm_ft_model.fit(X_train_expert, y_train_expert, epochs=15, batch_size=64, validation_split=0.2, verbose=2)
    else:
        print("\nNo expert data generated, skipping training. Creating a new model.")
        lstm_ft_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)

    # =========================================================================
    # PART 3: Fine-tune with Reinforcement Learning (Creating LSTM_FTR)
    # =========================================================================
    print("\n--- Part 3: Fine-tuning to create LSTM_FTR model ---")
    start_time = time.time()
    
    lstm_ftr_model = tf.keras.models.clone_model(lstm_ft_model)
    lstm_ftr_model.set_weights(lstm_ft_model.get_weights())
    optimizer = tf.keras.optimizers.AdamW(learning_rate=5e-4) # Use a slightly lower learning rate for fine-tuning
    lstm_ftr_model.compile(optimizer=optimizer, loss='mean_squared_error')

    epsilon = 0.5

    for epoch in range(NUM_RL_EPOCHS):
        print(f"\nRL Epoch {epoch+1}/{NUM_RL_EPOCHS}, Epsilon: {epsilon:.2f}")
        best_episodes_data = []
        
        for i in range(EPISODES_PER_EPOCH):
            start_pos = (random.uniform(0, 5), random.uniform(0, 5))
            start_theta = random.uniform(-np.pi, np.pi)
            path, success, sequences, actions = run_episode(start_pos, start_theta, lstm_ftr_model, use_exploration=True, epsilon=epsilon)
            
            if success:
                reward = 100 / len(path)
                best_episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})

        if not best_episodes_data:
            print("No successful paths in this RL epoch, skipping training.")
            continue
            
        best_episodes_data.sort(key=lambda x: x['reward'], reverse=True)
        top_episodes = best_episodes_data[:len(best_episodes_data)//2]
        
        X_fine_tune = []
        y_fine_tune = []
        for episode_data in top_episodes:
            X_fine_tune.extend(episode_data['sequences'])
            y_fine_tune.extend(episode_data['actions'])
            
        if X_fine_tune:
            print(f"Fine-tuning on {len(X_fine_tune)} samples from the best paths...")
            lstm_ftr_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        
        epsilon = max(0.1, epsilon * 0.9)
    
    print(f"RL Fine-tuning finished in {time.time() - start_time:.2f} seconds.")

    # =========================================================================
    # PART 4: Final Evaluation and Comparison
    # =========================================================================
    print("\n--- Part 4: Evaluating all three models ---")
    plt.figure(figsize=(12, 12))
    
    for obs in OBSTACLES:
        plt.gca().add_patch(plt.Rectangle((obs[0], obs[1]), obs[2]-obs[0], obs[3]-obs[1], color='gray'))
    plt.plot(GOAL_POS[0], GOAL_POS[1], 'r*', markersize=20, label='Goal')
    
    eval_start_pos = (2, 2)
    eval_start_theta = np.pi / 4

    # 1. Expert Path
    path_expert, _, _, _ = run_episode(eval_start_pos, eval_start_theta, "expert")
    path_expert_np = np.array(path_expert)
    plt.plot(path_expert_np[:, 0], path_expert_np[:, 1], 'b--', label=f'Expert (APF) Path - {len(path_expert)} steps')
    
    # 2. LSTM_FT Path
    path_ft, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_ft_model)
    path_ft_np = np.array(path_ft)
    plt.plot(path_ft_np[:, 0], path_ft_np[:, 1], 'g-.', label=f'LSTM_FT Path - {len(path_ft)} steps')

    # 3. LSTM_FTR Path
    path_ftr, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_ftr_model)
    path_ftr_np = np.array(path_ftr)
    plt.plot(path_ftr_np[:, 0], path_ftr_np[:, 1], 'm-', linewidth=2.5, label=f'LSTM_FTR Path (Final) - {len(path_ftr)} steps')
    
    plt.title("Comparison of Expert, LSTM_FT, and LSTM_FTR Models")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from collections import deque
import random
import time

# --- 1. Environment and Simulation ---

# Constants
GOAL_POS = (25.0, 25.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 500
GOAL_RADIUS = 1.0
SEQUENCE_LENGTH = 5

# --- IMPROVEMENT: SAFETY AND ROBOT SIZE ---
ROBOT_RADIUS = 0.5  # The robot has a physical radius of 0.5 units
PROXIMITY_PENALTY_WEIGHT = 0.5  # How much to penalize getting close
MIN_SAFE_DISTANCE = 1.5  # Any lidar reading below this will be penalized

# Obstacles represented as rectangles [x_min, y_min, x_max, y_max]
OBSTACLES = np.array([
    [10, 0, 12, 15],
    [10, 20, 12, 30],
    [18, 10, 20, 25]
])

# Pre-calculate "inflated" obstacles for efficient collision checking
INFLATED_OBSTACLES = OBSTACLES + np.array([-ROBOT_RADIUS, -ROBOT_RADIUS, ROBOT_RADIUS, ROBOT_RADIUS])


class Robot:
    def __init__(self, x, y, theta):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega):
        omega = np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)
        self.theta += omega * DT
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += ROBOT_SPEED * math.cos(self.theta) * DT
        self.y += ROBOT_SPEED * math.sin(self.theta) * DT
        return (self.x, self.y), self.theta


def get_lidar_readings_vectorized(robot_pos, robot_theta, num_sensors=5, max_dist=10.0, ray_step=0.1):
    """Simulates 5 Lidar sensors using NumPy for speed."""
    readings = np.full(num_sensors, max_dist)
    angles = np.linspace(-np.pi / 2, np.pi / 2, num_sensors) + robot_theta
    dist_steps = np.arange(ray_step, max_dist, ray_step)
    ray_points_x = robot_pos[0] + dist_steps[:, np.newaxis] * np.cos(angles)
    ray_points_y = robot_pos[1] + dist_steps[:, np.newaxis] * np.sin(angles)

    for obs in OBSTACLES:
        in_obstacle = (
            (ray_points_x >= obs[0]) & (ray_points_x <= obs[2]) &
            (ray_points_y >= obs[1]) & (ray_points_y <= obs[3])
        )
        collision_indices = np.argmax(in_obstacle, axis=0)
        for i in range(num_sensors):
            if in_obstacle[collision_indices[i], i]:
                dist = dist_steps[collision_indices[i]]
                if dist < readings[i]:
                    readings[i] = dist
    return readings


def get_state(robot_pos, robot_theta):
    """Generates the 7-dimensional state vector."""
    lidar = get_lidar_readings_vectorized(robot_pos, robot_theta)
    dx = GOAL_POS[0] - robot_pos[0]
    dy = GOAL_POS[1] - robot_pos[1]
    dist_to_goal = math.sqrt(dx ** 2 + dy ** 2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = math.atan2(math.sin(angle_to_goal - robot_theta), math.cos(angle_to_goal - robot_theta))
    state = np.concatenate((lidar / 10.0, [dist_to_goal / 30.0, angle_diff / np.pi]))
    return state


# --- 2. The Expert and The Learner Models ---

def apf_expert_policy(state):
    """A more cautious APF-based expert that respects a safety margin."""
    lidar_readings = state[:5] * 10.0
    angle_to_goal = state[6] * np.pi
    repulsive_omega = 0.0
    weights = [0.2, 0.5, 1.0, 0.5, 0.2]
    turn_directions = [np.pi / 2, np.pi / 4, 0, -np.pi / 4, -np.pi / 2]

    # --- IMPROVEMENT: EXPERT IS MORE CAUTIOUS ---
    # Activate repulsive force from further away and with more strength
    for i in range(5):
        if lidar_readings[i] < 4.0:
            repulsive_omega += weights[i] * (1.5 / max(lidar_readings[i], 0.1)) * turn_directions[i]

    attractive_omega = 2.0 * angle_to_goal
    total_omega = attractive_omega - repulsive_omega
    return np.clip(total_omega, -MAX_TURN_RATE, MAX_TURN_RATE)


def create_lstm_model(sequence_length, n_inputs, n_outputs):
    model = Sequential([
        Input(shape=(sequence_length, n_inputs)),
        LSTM(64, return_sequences=True),
        LSTM(64),
        Dense(32, activation='relu'),
        Dense(n_outputs, activation='tanh')
    ])
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    model.summary()
    return model


@tf.function
def predict_action(model, nn_input):
    """Performs a fast, compiled prediction."""
    return model(nn_input, training=False)


def run_episode(start_pos, start_theta, policy_model, use_exploration=False, epsilon=0.0):
    """Runs a single episode with collision checks and proximity penalties."""
    robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
    path_taken = [start_pos]
    state_history = deque(maxlen=SEQUENCE_LENGTH)
    episode_sequences = []
    episode_actions = []
    total_proximity_penalty = 0

    for _ in range(MAX_EPISODE_STEPS):
        pos = (robot.x, robot.y)

        # --- IMPROVEMENT: COLLISION CHECK ---
        for obs in INFLATED_OBSTACLES:
            if obs[0] <= pos[0] <= obs[2] and obs[1] <= pos[1] <= obs[3]:
                return path_taken, False, [], [], 0 # Collision, failed episode

        current_state = get_state(pos, robot.theta)
        state_history.append(current_state)

        # --- IMPROVEMENT: CALCULATE PROXIMITY PENALTY ---
        lidar_readings = current_state[:5] * 10.0
        if np.any(lidar_readings < MIN_SAFE_DISTANCE):
            penalty_per_step = (1 / np.min(lidar_readings))
            total_proximity_penalty += penalty_per_step

        if len(state_history) < SEQUENCE_LENGTH:
            omega = 0.1 * (random.random() - 0.5)
        else:
            sequence = np.array(state_history)
            if policy_model == "expert":
                omega = apf_expert_policy(current_state)
            else:
                nn_input = np.reshape(sequence, (1, SEQUENCE_LENGTH, 7))
                nn_input_tensor = tf.convert_to_tensor(nn_input, dtype=tf.float32)
                normalized_omega_tensor = predict_action(policy_model, nn_input_tensor)
                normalized_omega = normalized_omega_tensor.numpy()[0][0]
                if use_exploration and random.random() < epsilon:
                    normalized_omega += random.uniform(-0.5, 0.5)
                omega = normalized_omega * MAX_TURN_RATE

            episode_sequences.append(sequence)
            episode_actions.append(omega / MAX_TURN_RATE)

        new_pos, _ = robot.move(omega)
        path_taken.append(new_pos)

        if math.dist(new_pos, GOAL_POS) < GOAL_RADIUS:
            path_length_reward = 100 / len(path_taken)
            final_reward = path_length_reward - (total_proximity_penalty * PROXIMITY_PENALTY_WEIGHT)
            return path_taken, True, episode_sequences, episode_actions, final_reward

    return path_taken, False, [], [], 0 # Timeout, failed episode


# --- 3. Main Execution Block ---

if __name__ == "__main__":
    
    # Reduced numbers for quicker testing runs
    NUM_EXPERT_EPISODES = 25
    NUM_RL_EPOCHS = 10
    EPISODES_PER_EPOCH = 20

    # =========================================================================
    # PART 1: Generate Training Data with the (now safer) "Expert"
    # =========================================================================
    print("\n--- Part 1: Generating Data with Cautious APF Expert ---")
    start_time = time.time()
    all_expert_sequences = []
    all_expert_actions = []

    for i in range(NUM_EXPERT_EPISODES):
        print(f"Running expert episode {i + 1}/{NUM_EXPERT_EPISODES}")
        start_pos = (random.uniform(0, 5), random.uniform(0, 5))
        start_theta = random.uniform(-np.pi, np.pi)
        # The `run_episode` for the expert doesn't need the reward value
        path, success, sequences, actions, _ = run_episode(start_pos, start_theta, "expert")
        if success:
            all_expert_sequences.extend(sequences)
            all_expert_actions.extend(actions)

    X_train_expert = np.array(all_expert_sequences)
    y_train_expert = np.array(all_expert_actions)
    print(f"Generated {X_train_expert.shape[0]} samples in {time.time() - start_time:.2f} seconds.")

    # =========================================================================
    # PART 2: Pre-train the LSTM model (Creating LSTM_FT)
    # =========================================================================
    if X_train_expert.shape[0] > 0:
        print("\n--- Part 2: Pre-training the LSTM_FT model ---")
        lstm_ft_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)
        lstm_ft_model.fit(X_train_expert, y_train_expert, epochs=15, batch_size=64, validation_split=0.2, verbose=2)
    else:
        print("\nNo expert data generated, creating a new model from scratch.")
        lstm_ft_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)

    # =========================================================================
    # PART 3: Fine-tune with Safety-Aware Reinforcement Learning (Creating LSTM_FTR)
    # =========================================================================
    print("\n--- Part 3: Fine-tuning to create LSTM_FTR model ---")
    start_time = time.time()
    
    lstm_ftr_model = tf.keras.models.clone_model(lstm_ft_model)
    lstm_ftr_model.set_weights(lstm_ft_model.get_weights())
    optimizer = tf.keras.optimizers.AdamW(learning_rate=5e-4)
    lstm_ftr_model.compile(optimizer=optimizer, loss='mean_squared_error')
    epsilon = 0.5

    for epoch in range(NUM_RL_EPOCHS):
        print(f"\nRL Epoch {epoch + 1}/{NUM_RL_EPOCHS}, Epsilon: {epsilon:.2f}")
        episodes_data = []
        for i in range(EPISODES_PER_EPOCH):
            start_pos = (random.uniform(0, 5), random.uniform(0, 5))
            start_theta = random.uniform(-np.pi, np.pi)
            path, success, sequences, actions, reward = run_episode(
                start_pos, start_theta, lstm_ftr_model, use_exploration=True, epsilon=epsilon
            )
            if success:
                episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})

        if not episodes_data:
            print("No successful paths in this RL epoch, skipping training.")
            epsilon = max(0.1, epsilon * 0.9) # Still decay epsilon
            continue

        episodes_data.sort(key=lambda x: x['reward'], reverse=True)
        top_episodes = episodes_data[:len(episodes_data) // 2]
        
        X_fine_tune = []
        y_fine_tune = []
        for episode_data in top_episodes:
            X_fine_tune.extend(episode_data['sequences'])
            y_fine_tune.extend(episode_data['actions'])

        if X_fine_tune:
            print(f"Fine-tuning on {len(X_fine_tune)} samples from the best paths...")
            lstm_ftr_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        
        epsilon = max(0.1, epsilon * 0.9)

    print(f"RL Fine-tuning finished in {time.time() - start_time:.2f} seconds.")

    # =========================================================================
    # PART 4: Final Evaluation and Comparison
    # =========================================================================
    print("\n--- Part 4: Evaluating all three models ---")
    plt.figure(figsize=(12, 12))
    ax = plt.gca()
    
    # Plot normal obstacles for visualization
    for obs in OBSTACLES:
        ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2] - obs[0], obs[3] - obs[1], color='gray'))
    # Optional: Plot inflated obstacles to see the "no-go" zones
    for obs in INFLATED_OBSTACLES:
        ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2] - obs[0], obs[3] - obs[1], color='red', alpha=0.2, hatch='/'))
        
    plt.plot(GOAL_POS[0], GOAL_POS[1], 'r*', markersize=20, label='Goal')
    
    eval_start_pos = (2, 2)
    eval_start_theta = np.pi / 4

    # 1. Expert Path
    path_expert, _, _, _, _ = run_episode(eval_start_pos, eval_start_theta, "expert")
    path_expert_np = np.array(path_expert)
    plt.plot(path_expert_np[:, 0], path_expert_np[:, 1], 'b--', label=f'Expert (APF) Path')
    
    # 2. LSTM_FT Path
    path_ft, _, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_ft_model)
    path_ft_np = np.array(path_ft)
    plt.plot(path_ft_np[:, 0], path_ft_np[:, 1], 'g-.', label=f'LSTM_FT Path')

    # 3. LSTM_FTR Path
    path_ftr, _, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_ftr_model)
    path_ftr_np = np.array(path_ftr)
    plt.plot(path_ftr_np[:, 0], path_ftr_np[:, 1], 'm-', linewidth=2.5, label=f'LSTM_FTR Path (Final)')
    
    plt.title("Comparison of Expert, LSTM_FT, and LSTM_FTR Models")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from collections import deque
import random
import time

# --- 1. Environment and Simulation ---

# Constants
GOAL_POS = (25.0, 25.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 500
GOAL_RADIUS = 1.0
SEQUENCE_LENGTH = 5

# Safety and Robot Size Constants
ROBOT_RADIUS = 0.5
PROXIMITY_PENALTY_WEIGHT = 0.5
MIN_SAFE_DISTANCE = 1.5

# Obstacles
OBSTACLES = np.array([
    [10, 0, 12, 15],
    [10, 20, 12, 30],
    [18, 10, 20, 25]
])
INFLATED_OBSTACLES = OBSTACLES + np.array([-ROBOT_RADIUS, -ROBOT_RADIUS, ROBOT_RADIUS, ROBOT_RADIUS])


class Robot:
    def __init__(self, x, y, theta):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega):
        omega = np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)
        self.theta += omega * DT
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += ROBOT_SPEED * math.cos(self.theta) * DT
        self.y += ROBOT_SPEED * math.sin(self.theta) * DT
        return (self.x, self.y), self.theta

# --- Helper Functions for Simulation ---

def generate_challenging_start_pos():
    """Generates a start position from one of several challenging regions."""
    region = random.choice(['easy', 'corridor1', 'corridor2', 'corridor_approach'])
    
    if region == 'easy':
        x = random.uniform(0, 5)
        y = random.uniform(0, 5)
    elif region == 'corridor1':
        x = random.uniform(13, 17)
        y = random.uniform(10, 15)
    elif region == 'corridor2':
        x = random.uniform(13, 17)
        y = random.uniform(20, 25)
    elif region == 'corridor_approach':
        x = random.uniform(13, 17)
        y = random.uniform(5, 8)
        
    theta = random.uniform(-np.pi, np.pi)
    return (x, y), theta

def get_lidar_readings_vectorized(robot_pos, robot_theta, num_sensors=5, max_dist=10.0, ray_step=0.1):
    """Simulates 5 Lidar sensors using NumPy for speed."""
    readings = np.full(num_sensors, max_dist)
    angles = np.linspace(-np.pi / 2, np.pi / 2, num_sensors) + robot_theta
    dist_steps = np.arange(ray_step, max_dist, ray_step)
    ray_points_x = robot_pos[0] + dist_steps[:, np.newaxis] * np.cos(angles)
    ray_points_y = robot_pos[1] + dist_steps[:, np.newaxis] * np.sin(angles)

    for obs in OBSTACLES:
        in_obstacle = (
            (ray_points_x >= obs[0]) & (ray_points_x <= obs[2]) &
            (ray_points_y >= obs[1]) & (ray_points_y <= obs[3])
        )
        collision_indices = np.argmax(in_obstacle, axis=0)
        for i in range(num_sensors):
            if in_obstacle[collision_indices[i], i]:
                dist = dist_steps[collision_indices[i]]
                if dist < readings[i]:
                    readings[i] = dist
    return readings

def get_state(robot_pos, robot_theta):
    """Generates the 7-dimensional state vector."""
    lidar = get_lidar_readings_vectorized(robot_pos, robot_theta)
    dx = GOAL_POS[0] - robot_pos[0]
    dy = GOAL_POS[1] - robot_pos[1]
    dist_to_goal = math.sqrt(dx ** 2 + dy ** 2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = math.atan2(math.sin(angle_to_goal - robot_theta), math.cos(angle_to_goal - robot_theta))
    state = np.concatenate((lidar / 10.0, [dist_to_goal / 30.0, angle_diff / np.pi]))
    return state


# --- 2. The Expert and The Learner Models ---

def apf_expert_policy(state):
    """A more cautious APF-based expert that respects a safety margin."""
    lidar_readings = state[:5] * 10.0
    angle_to_goal = state[6] * np.pi
    repulsive_omega = 0.0
    weights = [0.2, 0.5, 1.0, 0.5, 0.2]
    turn_directions = [np.pi / 2, np.pi / 4, 0, -np.pi / 4, -np.pi / 2]

    for i in range(5):
        if lidar_readings[i] < 4.0:
            repulsive_omega += weights[i] * (1.5 / max(lidar_readings[i], 0.1)) * turn_directions[i]

    attractive_omega = 2.0 * angle_to_goal
    total_omega = attractive_omega - repulsive_omega
    return np.clip(total_omega, -MAX_TURN_RATE, MAX_TURN_RATE)

def create_lstm_model(sequence_length, n_inputs, n_outputs):
    model = Sequential([
        Input(shape=(sequence_length, n_inputs)),
        LSTM(64, return_sequences=True),
        LSTM(64),
        Dense(32, activation='relu'),
        Dense(n_outputs, activation='tanh')
    ])
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    model.summary()
    return model

@tf.function
def predict_action(model, nn_input):
    """Performs a fast, compiled prediction."""
    return model(nn_input, training=False)

def run_episode(start_pos, start_theta, policy_model, use_exploration=False, epsilon=0.0):
    """
    Runs a single episode with all features, including the critical
    REWARD SHAPING to guide the agent out of local minima.
    """
    robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
    path_taken = [start_pos]
    state_history = deque(maxlen=SEQUENCE_LENGTH)
    episode_sequences, episode_actions = [], []
    total_proximity_penalty = 0.0

    ### --- THE KEY IMPROVEMENT: REWARD SHAPING --- ###
    # We need the distance at the previous step to know if we're making progress.
    last_dist_to_goal = math.dist(start_pos, GOAL_POS)
    # This will accumulate the rewards/penalties from getting closer/further.
    progress_reward = 0.0
    PROGRESS_REWARD_WEIGHT = 2.0 # Controls the importance of making progress

    for _ in range(MAX_EPISODE_STEPS):
        pos = (robot.x, robot.y)

        # Collision Check
        for obs in INFLATED_OBSTACLES:
            if obs[0] <= pos[0] <= obs[2] and obs[1] <= pos[1] <= obs[3]:
                return path_taken, False, [], [], 0.0

        current_state = get_state(pos, robot.theta)
        state_history.append(current_state)

        # Proximity Penalty
        lidar_readings = current_state[:5] * 10.0
        if np.any(lidar_readings < MIN_SAFE_DISTANCE):
            penalty_per_step = (1 / np.min(lidar_readings))
            total_proximity_penalty += penalty_per_step

        if len(state_history) < SEQUENCE_LENGTH:
            omega = 0.1 * (random.random() - 0.5)
        else:
            sequence = np.array(state_history)
            if policy_model == "expert":
                omega = apf_expert_policy(current_state)
                omega += np.random.normal(0, 0.2) # Give expert slightly more noise
            else:
                nn_input = np.reshape(sequence, (1, SEQUENCE_LENGTH, 7))
                nn_input_tensor = tf.convert_to_tensor(nn_input, dtype=tf.float32)
                normalized_omega_tensor = predict_action(policy_model, nn_input_tensor)
                normalized_omega = normalized_omega_tensor.numpy()[0][0]
                if use_exploration and random.random() < epsilon:
                    normalized_omega += random.uniform(-0.5, 0.5)
                omega = normalized_omega * MAX_TURN_RATE

            episode_sequences.append(sequence)
            episode_actions.append(omega / MAX_TURN_RATE)

        new_pos, _ = robot.move(omega)
        path_taken.append(new_pos)

        ### --- REWARD SHAPING CALCULATION --- ###
        current_dist_to_goal = math.dist(new_pos, GOAL_POS)
        # Did we make progress?
        distance_gained = last_dist_to_goal - current_dist_to_goal
        progress_reward += distance_gained
        # Update for the next step
        last_dist_to_goal = current_dist_to_goal


        # Check for success
        if current_dist_to_goal < GOAL_RADIUS:
            # Final large reward for reaching the goal
            success_reward = 100.0
            
            # Combine all reward components
            final_reward = (
                success_reward
                + (progress_reward * PROGRESS_REWARD_WEIGHT)
                - (total_proximity_penalty * PROXIMITY_PENALTY_WEIGHT)
            )
            # We no longer need the path length reward as progress is already rewarded
            
            return path_taken, True, episode_sequences, episode_actions, final_reward

    # Return 0 reward for timeout or collision
    return path_taken, False, [], [], 0.0


# --- 3. Main Execution Block ---

if __name__ == "__main__":
    
    NUM_EXPERT_EPISODES = 50  # Increased for more diverse data
    NUM_RL_EPOCHS = 15
    EPISODES_PER_EPOCH = 25

    # =========================================================================
    # PART 1: Generate Training Data from Challenging Starts
    # =========================================================================
    print("\n--- Part 1: Generating Data with Cautious APF Expert ---")
    start_time = time.time()
    all_expert_sequences, all_expert_actions = [], []

    for i in range(NUM_EXPERT_EPISODES):
        print(f"Running expert episode {i + 1}/{NUM_EXPERT_EPISODES}")
        start_pos, start_theta = generate_challenging_start_pos()
        path, success, sequences, actions, _ = run_episode(start_pos, start_theta, "expert")
        if success:
            all_expert_sequences.extend(sequences)
            all_expert_actions.extend(actions)

    X_train_expert = np.array(all_expert_sequences)
    y_train_expert = np.array(all_expert_actions)
    print(f"Generated {X_train_expert.shape[0]} samples in {time.time() - start_time:.2f} seconds.")

    # =========================================================================
    # PART 2: Pre-train the LSTM model
    # =========================================================================
    if X_train_expert.shape[0] > 0:
        print("\n--- Part 2: Pre-training the LSTM_FT model ---")
        lstm_ft_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)
        lstm_ft_model.fit(X_train_expert, y_train_expert, epochs=20, batch_size=64, validation_split=0.2, verbose=2)
    else:
        print("\nNo expert data generated, creating a new model from scratch.")
        lstm_ft_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)

    # =========================================================================
    # PART 3: Fine-tune with Safety-Aware and Robust RL
    # =========================================================================
    print("\n--- Part 3: Fine-tuning to create LSTM_FTR model ---")
    start_time = time.time()
    
    lstm_ftr_model = tf.keras.models.clone_model(lstm_ft_model)
    lstm_ftr_model.set_weights(lstm_ft_model.get_weights())
    optimizer = tf.keras.optimizers.AdamW(learning_rate=5e-4)
    lstm_ftr_model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    epsilon = 0.7  # Start with higher exploration
    
    for epoch in range(NUM_RL_EPOCHS):
        print(f"\nRL Epoch {epoch + 1}/{NUM_RL_EPOCHS}, Epsilon: {epsilon:.2f}")
        episodes_data = []
        for i in range(EPISODES_PER_EPOCH):
            start_pos, start_theta = generate_challenging_start_pos()
            path, success, sequences, actions, reward = run_episode(
                start_pos, start_theta, lstm_ftr_model, use_exploration=True, epsilon=epsilon
            )
            if success:
                episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})

        if not episodes_data:
            print("No successful paths in this RL epoch, skipping training.")
            epsilon = max(0.1, epsilon * 0.95)
            continue

        episodes_data.sort(key=lambda x: x['reward'], reverse=True)
        top_episodes = episodes_data[:len(episodes_data) // 2]
        
        X_fine_tune, y_fine_tune = [], []
        for episode_data in top_episodes:
            X_fine_tune.extend(episode_data['sequences'])
            y_fine_tune.extend(episode_data['actions'])

        if X_fine_tune:
            print(f"Fine-tuning on {len(X_fine_tune)} samples from the best paths...")
            lstm_ftr_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        
        epsilon = max(0.1, epsilon * 0.95) # Slower decay

    print(f"RL Fine-tuning finished in {time.time() - start_time:.2f} seconds.")

    # =========================================================================
    # PART 4: Final Evaluation from a Difficult Starting Point
    # =========================================================================
    print("\n--- Part 4: Evaluating all three models from a difficult start ---")
    plt.figure(figsize=(12, 12))
    ax = plt.gca()
    
    for obs in OBSTACLES:
        ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2] - obs[0], obs[3] - obs[1], color='gray'))
    for obs in INFLATED_OBSTACLES:
        ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2] - obs[0], obs[3] - obs[1], color='red', alpha=0.2, hatch='/'))
        
    plt.plot(GOAL_POS[0], GOAL_POS[1], 'r*', markersize=20, label='Goal')
    
    # Start in the difficult canyon
    eval_start_pos = (15, 5) 
    eval_start_theta = np.pi / 2
    
    print("Evaluating Expert (APF)...")
    path_expert, _, _, _, _ = run_episode(eval_start_pos, eval_start_theta, "expert")
    path_expert_np = np.array(path_expert)
    plt.plot(path_expert_np[:, 0], path_expert_np[:, 1], 'b--', label=f'Expert (APF) Path')
    
    print("Evaluating Pre-Trained Model (LSTM_FT)...")
    path_ft, _, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_ft_model)
    path_ft_np = np.array(path_ft)
    plt.plot(path_ft_np[:, 0], path_ft_np[:, 1], 'g-.', label=f'LSTM_FT Path')

    print("Evaluating Fine-Tuned Model (LSTM_FTR)...")
    path_ftr, _, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_ftr_model)
    path_ftr_np = np.array(path_ftr)
    plt.plot(path_ftr_np[:, 0], path_ftr_np[:, 1], 'm-', linewidth=2.5, label=f'LSTM_FTR Path (Final)')
    
    plt.title("Comparison of Models from a Difficult Starting Position")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from collections import deque
import random
import time

# --- 1. Environment and Simulation (No changes from last version) ---

# Constants
GOAL_POS = (25.0, 25.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 500
GOAL_RADIUS = 1.0
SEQUENCE_LENGTH = 5

# Safety and Robot Size Constants
ROBOT_RADIUS = 0.5
PROXIMITY_PENALTY_WEIGHT = 0.5
MIN_SAFE_DISTANCE = 1.5
PROGRESS_REWARD_WEIGHT = 2.0

# Obstacles
OBSTACLES = np.array([
    [10, 0, 12, 15],
    [10, 20, 12, 30],
    [18, 10, 20, 25]
])
INFLATED_OBSTACLES = OBSTACLES + np.array([-ROBOT_RADIUS, -ROBOT_RADIUS, ROBOT_RADIUS, ROBOT_RADIUS])


class Robot:
    def __init__(self, x, y, theta):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega):
        omega = np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)
        self.theta += omega * DT
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += ROBOT_SPEED * math.cos(self.theta) * DT
        self.y += ROBOT_SPEED * math.sin(self.theta) * DT
        return (self.x, self.y), self.theta

# --- Helper Functions for Simulation ---
# (No changes to these functions from last version)

def get_lidar_readings_vectorized(robot_pos, robot_theta, num_sensors=5, max_dist=10.0, ray_step=0.1):
    readings = np.full(num_sensors, max_dist)
    angles = np.linspace(-np.pi / 2, np.pi / 2, num_sensors) + robot_theta
    dist_steps = np.arange(ray_step, max_dist, ray_step)
    ray_points_x = robot_pos[0] + dist_steps[:, np.newaxis] * np.cos(angles)
    ray_points_y = robot_pos[1] + dist_steps[:, np.newaxis] * np.sin(angles)
    for obs in OBSTACLES:
        in_obstacle = ((ray_points_x >= obs[0]) & (ray_points_x <= obs[2]) & (ray_points_y >= obs[1]) & (ray_points_y <= obs[3]))
        collision_indices = np.argmax(in_obstacle, axis=0)
        for i in range(num_sensors):
            if in_obstacle[collision_indices[i], i]:
                dist = dist_steps[collision_indices[i]]
                if dist < readings[i]: readings[i] = dist
    return readings

def get_state(robot_pos, robot_theta):
    lidar = get_lidar_readings_vectorized(robot_pos, robot_theta)
    dx = GOAL_POS[0] - robot_pos[0]
    dy = GOAL_POS[1] - robot_pos[1]
    dist_to_goal = math.sqrt(dx ** 2 + dy ** 2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = math.atan2(math.sin(angle_to_goal - robot_theta), math.cos(angle_to_goal - robot_theta))
    state = np.concatenate((lidar / 10.0, [dist_to_goal / 30.0, angle_diff / np.pi]))
    return state

# --- 2. Models and Policies ---
# (No changes to these functions from last version)

def apf_expert_policy(state):
    lidar_readings = state[:5] * 10.0
    angle_to_goal = state[6] * np.pi
    repulsive_omega = 0.0
    weights = [0.2, 0.5, 1.0, 0.5, 0.2]
    turn_directions = [np.pi / 2, np.pi / 4, 0, -np.pi / 4, -np.pi / 2]
    for i in range(5):
        if lidar_readings[i] < 4.0:
            repulsive_omega += weights[i] * (1.5 / max(lidar_readings[i], 0.1)) * turn_directions[i]
    attractive_omega = 2.0 * angle_to_goal
    total_omega = attractive_omega - repulsive_omega
    return np.clip(total_omega, -MAX_TURN_RATE, MAX_TURN_RATE)

def create_lstm_model(sequence_length, n_inputs, n_outputs):
    model = Sequential([Input(shape=(sequence_length, n_inputs)), LSTM(64, return_sequences=True), LSTM(64), Dense(32, activation='relu'), Dense(n_outputs, activation='tanh')])
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

@tf.function
def predict_action(model, nn_input):
    return model(nn_input, training=False)

def run_episode(start_pos, start_theta, policy_model, use_exploration=False, epsilon=0.0):
    robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
    path_taken = [start_pos]
    state_history = deque(maxlen=SEQUENCE_LENGTH)
    episode_sequences, episode_actions = [], []
    total_proximity_penalty, progress_reward = 0.0, 0.0
    last_dist_to_goal = math.dist(start_pos, GOAL_POS)

    for _ in range(MAX_EPISODE_STEPS):
        pos = (robot.x, robot.y)
        for obs in INFLATED_OBSTACLES:
            if obs[0] <= pos[0] <= obs[2] and obs[1] <= pos[1] <= obs[3]: return path_taken, False, [], [], 0.0
        current_state = get_state(pos, robot.theta)
        state_history.append(current_state)
        lidar_readings = current_state[:5] * 10.0
        if np.any(lidar_readings < MIN_SAFE_DISTANCE): total_proximity_penalty += (1 / np.min(lidar_readings))
        if len(state_history) < SEQUENCE_LENGTH: omega = 0.1 * (random.random() - 0.5)
        else:
            sequence = np.array(state_history)
            if policy_model == "expert":
                omega = apf_expert_policy(current_state)
                omega += np.random.normal(0, 0.2)
            else:
                nn_input = np.reshape(sequence, (1, SEQUENCE_LENGTH, 7))
                nn_input_tensor = tf.convert_to_tensor(nn_input, dtype=tf.float32)
                normalized_omega = predict_action(policy_model, nn_input_tensor).numpy()[0][0]
                if use_exploration and random.random() < epsilon: normalized_omega += random.uniform(-0.5, 0.5)
                omega = normalized_omega * MAX_TURN_RATE
            episode_sequences.append(sequence)
            episode_actions.append(omega / MAX_TURN_RATE)
        new_pos, _ = robot.move(omega)
        path_taken.append(new_pos)
        current_dist_to_goal = math.dist(new_pos, GOAL_POS)
        distance_gained = last_dist_to_goal - current_dist_to_goal
        progress_reward += distance_gained
        last_dist_to_goal = current_dist_to_goal
        if current_dist_to_goal < GOAL_RADIUS:
            final_reward = 100.0 + (progress_reward * PROGRESS_REWARD_WEIGHT) - (total_proximity_penalty * PROXIMITY_PENALTY_WEIGHT)
            return path_taken, True, episode_sequences, episode_actions, final_reward
    return path_taken, False, [], [], 0.0

# --- 3. Main Execution Block with CURRICULUM LEARNING ---

if __name__ == "__main__":
    
    # --- CURRICULUM LEVEL 1: Train on Easy Tasks ---
    print("\n" + "="*50)
    print("=== CURRICULUM LEVEL 1: IMITATION LEARNING ON EASY TASKS ===")
    print("="*50)
    
    all_expert_sequences, all_expert_actions = [], []
    for i in range(50): # 50 episodes from easy starts
        print(f"Running easy expert episode {i+1}/50")
        start_pos = (random.uniform(0, 5), random.uniform(0, 5))
        start_theta = random.uniform(-np.pi, np.pi)
        path, success, sequences, actions, _ = run_episode(start_pos, start_theta, "expert")
        if success:
            all_expert_sequences.extend(sequences)
            all_expert_actions.extend(actions)

    lstm_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)
    if all_expert_sequences:
        X_train_expert = np.array(all_expert_sequences)
        y_train_expert = np.array(all_expert_actions)
        print(f"\nTraining on {X_train_expert.shape[0]} samples from easy expert paths...")
        lstm_model.fit(X_train_expert, y_train_expert, epochs=20, batch_size=64, validation_split=0.2, verbose=1)
    else:
        print("\nNo successful expert paths, starting with a random model.")
    
    # --- CURRICULUM LEVEL 2: RL from Canyon Exit ---
    print("\n" + "="*50)
    print("=== CURRICULUM LEVEL 2: RL - ESCAPING THE CANYON ===")
    print("="*50)
    
    # Re-compile with a smaller learning rate for fine-tuning
    lstm_model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=5e-4), loss='mean_squared_error')
    epsilon = 0.6
    for epoch in range(10): # 10 RL epochs for this task
        print(f"\nRL Canyon Exit - Epoch {epoch+1}/10, Epsilon: {epsilon:.2f}")
        episodes_data = []
        for i in range(20): # 20 episodes per epoch
            # START AT THE EXIT OF THE CANYON
            start_pos = (random.uniform(13, 17), random.uniform(16, 19))
            start_theta = random.uniform(-np.pi, np.pi)
            _, success, sequences, actions, reward = run_episode(start_pos, start_theta, lstm_model, use_exploration=True, epsilon=epsilon)
            if success: episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})

        if episodes_data:
            episodes_data.sort(key=lambda x: x['reward'], reverse=True)
            top_episodes = episodes_data[:len(episodes_data) // 2]
            X_fine_tune, y_fine_tune = [], []
            for ep_data in top_episodes:
                X_fine_tune.extend(ep_data['sequences'])
                y_fine_tune.extend(ep_data['actions'])
            if X_fine_tune:
                print(f"Fine-tuning on {len(X_fine_tune)} samples...")
                lstm_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        epsilon = max(0.1, epsilon * 0.9)
        
    # --- CURRICULUM LEVEL 3: RL from Canyon Entrance (The Final Task) ---
    print("\n" + "="*50)
    print("=== CURRICULUM LEVEL 3: RL - NAVIGATING THE FULL CANYON ===")
    print("="*50)
    
    epsilon = 0.5 # Reset epsilon for the new, harder task
    for epoch in range(15): # 15 RL epochs for the final, hardest task
        print(f"\nRL Full Canyon - Epoch {epoch+1}/15, Epsilon: {epsilon:.2f}")
        episodes_data = []
        for i in range(30): # More episodes to increase chance of success
            # START AT THE ENTRANCE OF THE CANYON
            start_pos = (random.uniform(13, 17), random.uniform(5, 8))
            start_theta = random.uniform(np.pi/4, 3*np.pi/4) # Point generally upwards
            _, success, sequences, actions, reward = run_episode(start_pos, start_theta, lstm_model, use_exploration=True, epsilon=epsilon)
            if success: episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})
        
        if episodes_data:
            episodes_data.sort(key=lambda x: x['reward'], reverse=True)
            top_episodes = episodes_data[:len(episodes_data) // 2]
            X_fine_tune, y_fine_tune = [], []
            for ep_data in top_episodes:
                X_fine_tune.extend(ep_data['sequences'])
                y_fine_tune.extend(ep_data['actions'])
            if X_fine_tune:
                print(f"Fine-tuning on {len(X_fine_tune)} samples...")
                lstm_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        epsilon = max(0.1, epsilon * 0.9)

    # --- FINAL EVALUATION ---
    print("\n--- Part 4: Final Evaluation ---")
    plt.figure(figsize=(12, 12))
    ax = plt.gca()
    for obs in OBSTACLES: ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2]-obs[0], obs[3]-obs[1], color='gray'))
    for obs in INFLATED_OBSTACLES: ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2]-obs[0], obs[3]-obs[1], color='red', alpha=0.2, hatch='/'))
    plt.plot(GOAL_POS[0], GOAL_POS[1], 'r*', markersize=20, label='Goal')
    
    eval_start_pos = (15, 5) 
    eval_start_theta = np.pi / 2
    
    print("Evaluating Final Model...")
    path_ftr, success, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_model)
    if success: print("SUCCESS: Model reached the goal!")
    else: print("FAILURE: Model did not reach the goal.")
    
    path_ftr_np = np.array(path_ftr)
    plt.plot(path_ftr_np[:, 0], path_ftr_np[:, 1], 'm-', linewidth=2.5, label=f'Final Curriculum-Trained Model')
    
    plt.title("Evaluation of Final Model from Difficult Start")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from collections import deque
import random
import time

# --- 1. Environment and Simulation ---

# Constants
GOAL_POS = (25.0, 25.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 500
GOAL_RADIUS = 1.0
SEQUENCE_LENGTH = 5

# Safety and Robot Size Constants - (MODIFIED)
ROBOT_RADIUS = 0.5
PROXIMITY_PENALTY_WEIGHT = 1.5 # زيادة من 0.5 لزيادة عقوبة الاقتراب
MIN_SAFE_DISTANCE = 1.5
PROGRESS_REWARD_WEIGHT = 2.0

# Obstacles
OBSTACLES = np.array([
    [10, 0, 12, 15],
    [10, 20, 12, 30],
    [18, 10, 20, 25]
])
INFLATED_OBSTACLES = OBSTACLES + np.array([-ROBOT_RADIUS, -ROBOT_RADIUS, ROBOT_RADIUS, ROBOT_RADIUS])


class Robot:
    def __init__(self, x, y, theta):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega):
        omega = np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)
        self.theta += omega * DT
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += ROBOT_SPEED * math.cos(self.theta) * DT
        self.y += ROBOT_SPEED * math.sin(self.theta) * DT
        return (self.x, self.y), self.theta

# --- Helper Functions for Simulation ---

def get_lidar_readings_vectorized(robot_pos, robot_theta, num_sensors=5, max_dist=10.0, ray_step=0.1):
    readings = np.full(num_sensors, max_dist)
    angles = np.linspace(-np.pi / 2, np.pi / 2, num_sensors) + robot_theta
    dist_steps = np.arange(ray_step, max_dist, ray_step)
    ray_points_x = robot_pos[0] + dist_steps[:, np.newaxis] * np.cos(angles)
    ray_points_y = robot_pos[1] + dist_steps[:, np.newaxis] * np.sin(angles)
    for obs in OBSTACLES:
        in_obstacle = ((ray_points_x >= obs[0]) & (ray_points_x <= obs[2]) & (ray_points_y >= obs[1]) & (ray_points_y <= obs[3]))
        collision_indices = np.argmax(in_obstacle, axis=0)
        for i in range(num_sensors):
            if in_obstacle[collision_indices[i], i]:
                dist = dist_steps[collision_indices[i]]
                if dist < readings[i]: readings[i] = dist
    return readings

def get_state(robot_pos, robot_theta, add_noise=False):
    """
    Generates the state vector. 
    If add_noise is True, it adds a small amount of random noise 
    to simulate sensor imprecision and improve model robustness.
    """
    lidar = get_lidar_readings_vectorized(robot_pos, robot_theta)
    
    dx = GOAL_POS[0] - robot_pos[0]
    dy = GOAL_POS[1] - robot_pos[1]
    
    dist_to_goal = math.sqrt(dx**2 + dy**2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = math.atan2(math.sin(angle_to_goal - robot_theta), math.cos(angle_to_goal - robot_theta))
    
    if add_noise:
        lidar_noise = np.random.normal(0, 0.02, size=lidar.shape) * lidar
        lidar += lidar_noise
        dist_noise = np.random.normal(0, 0.02) * dist_to_goal
        angle_noise = np.random.normal(0, 0.02) * angle_diff
        dist_to_goal += dist_noise
        angle_diff += angle_noise

    state = np.concatenate((lidar / 10.0, [dist_to_goal / 30.0, angle_diff / np.pi]))
    return state


# --- 2. Models and Policies ---

def apf_expert_policy(state):
    lidar_readings = state[:5] * 10.0
    angle_to_goal = state[6] * np.pi
    repulsive_omega = 0.0
    weights = [0.2, 0.5, 1.0, 0.5, 0.2]
    turn_directions = [np.pi / 2, np.pi / 4, 0, -np.pi / 4, -np.pi / 2]
    for i in range(5):
        if lidar_readings[i] < 4.0:
            repulsive_omega += weights[i] * (1.5 / max(lidar_readings[i], 0.1)) * turn_directions[i]
    attractive_omega = 2.0 * angle_to_goal
    total_omega = attractive_omega - repulsive_omega
    return np.clip(total_omega, -MAX_TURN_RATE, MAX_TURN_RATE)

# (MODIFIED) - زيادة سعة النموذج
def create_lstm_model(sequence_length, n_inputs, n_outputs):
    model = Sequential([
        Input(shape=(sequence_length, n_inputs)),
        LSTM(128, return_sequences=True), # زيادة السعة من 64 إلى 128
        LSTM(128),                       # زيادة السعة من 64 إلى 128
        Dense(64, activation='relu'),    # زيادة السعة من 32 إلى 64
        Dense(n_outputs, activation='tanh')
    ])
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-3)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

@tf.function
def predict_action(model, nn_input):
    return model(nn_input, training=False)

# (MODIFIED) - تعديل دالة المكافأة
def run_episode(start_pos, start_theta, policy_model, use_exploration=False, epsilon=0.0, use_state_noise=False):
    robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
    path_taken = [start_pos]
    state_history = deque(maxlen=SEQUENCE_LENGTH)
    episode_sequences, episode_actions = [], []
    total_proximity_penalty, progress_reward = 0.0, 0.0
    last_dist_to_goal = math.dist(start_pos, GOAL_POS)
    total_steps = 0 # إضافة عداد الخطوات

    for _ in range(MAX_EPISODE_STEPS):
        total_steps += 1 # زيادة العداد في كل خطوة
        pos = (robot.x, robot.y)
        for obs in INFLATED_OBSTACLES:
            if obs[0] <= pos[0] <= obs[2] and obs[1] <= pos[1] <= obs[3]: return path_taken, False, [], [], 0.0
        
        current_state = get_state(pos, robot.theta, add_noise=use_state_noise)
        state_history.append(current_state)
        
        lidar_readings = current_state[:5] * 10.0
        if np.any(lidar_readings < MIN_SAFE_DISTANCE): total_proximity_penalty += (1 / np.min(lidar_readings))
        
        if len(state_history) < SEQUENCE_LENGTH: omega = 0.1 * (random.random() - 0.5)
        else:
            sequence = np.array(state_history)
            if policy_model == "expert":
                omega = apf_expert_policy(current_state)
                omega += np.random.normal(0, 0.2)
            else:
                nn_input = np.reshape(sequence, (1, SEQUENCE_LENGTH, 7))
                nn_input_tensor = tf.convert_to_tensor(nn_input, dtype=tf.float32)
                normalized_omega = predict_action(policy_model, nn_input_tensor).numpy()[0][0]
                if use_exploration and random.random() < epsilon: normalized_omega += random.uniform(-0.5, 0.5)
                omega = normalized_omega * MAX_TURN_RATE
            episode_sequences.append(sequence)
            episode_actions.append(omega / MAX_TURN_RATE)
            
        new_pos, _ = robot.move(omega)
        path_taken.append(new_pos)
        
        current_dist_to_goal = math.dist(new_pos, GOAL_POS)
        distance_gained = last_dist_to_goal - current_dist_to_goal
        progress_reward += distance_gained
        last_dist_to_goal = current_dist_to_goal
        
        if current_dist_to_goal < GOAL_RADIUS:
            # إضافة عقوبة الوقت إلى المكافأة النهائية
            final_reward = 100.0 + (progress_reward * PROGRESS_REWARD_WEIGHT) - (total_proximity_penalty * PROXIMITY_PENALTY_WEIGHT) - (total_steps * 0.1)
            return path_taken, True, episode_sequences, episode_actions, final_reward
            
    return path_taken, False, [], [], 0.0

# --- 3. Main Execution Block with CURRICULUM LEARNING ---

if __name__ == "__main__":
    
    # --- CURRICULUM LEVEL 1: Train on Easy Tasks ---
    print("\n" + "="*50)
    print("=== CURRICULUM LEVEL 1: IMITATION LEARNING ON EASY TASKS ===")
    print("="*50)
    
    all_expert_sequences, all_expert_actions = [], []
    for i in range(50):
        print(f"Running easy expert episode {i+1}/50")
        start_pos = (random.uniform(0, 5), random.uniform(0, 5))
        start_theta = random.uniform(-np.pi, np.pi)
        path, success, sequences, actions, _ = run_episode(start_pos, start_theta, "expert", use_state_noise=False)
        if success:
            all_expert_sequences.extend(sequences)
            all_expert_actions.extend(actions)

    lstm_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)
    if all_expert_sequences:
        X_train_expert = np.array(all_expert_sequences)
        y_train_expert = np.array(all_expert_actions)
        print(f"\nTraining on {X_train_expert.shape[0]} samples from easy expert paths...")
        lstm_model.fit(X_train_expert, y_train_expert, epochs=20, batch_size=64, validation_split=0.2, verbose=1)
    else:
        print("\nNo successful expert paths, starting with a random model.")
    
    # --- CURRICULUM LEVEL 2: RL from Canyon Exit - (MODIFIED) - زيادة التدريب
    print("\n" + "="*50)
    print("=== CURRICULUM LEVEL 2: RL - ESCAPING THE CANYON ===")
    print("="*50)
    
    lstm_model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=5e-4), loss='mean_squared_error')
    epsilon = 0.6
    for epoch in range(25): # زيادة الحلقات من 10 إلى 25
        print(f"\nRL Canyon Exit - Epoch {epoch+1}/25, Epsilon: {epsilon:.2f}")
        episodes_data = []
        for i in range(40): # زيادة التجارب من 20 إلى 40
            start_pos = (random.uniform(13, 17), random.uniform(16, 19))
            start_theta = random.uniform(-np.pi, np.pi)
            _, success, sequences, actions, reward = run_episode(start_pos, start_theta, lstm_model, use_exploration=True, epsilon=epsilon, use_state_noise=True)
            if success: episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})

        if episodes_data:
            episodes_data.sort(key=lambda x: x['reward'], reverse=True)
            top_episodes = episodes_data[:len(episodes_data) // 2]
            X_fine_tune, y_fine_tune = [], []
            for ep_data in top_episodes:
                X_fine_tune.extend(ep_data['sequences'])
                y_fine_tune.extend(ep_data['actions'])
            if X_fine_tune:
                print(f"Fine-tuning on {len(X_fine_tune)} samples...")
                lstm_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        epsilon = max(0.1, epsilon * 0.95) # إبطاء معدل انخفاض إبسيلون
        
    # --- CURRICULUM LEVEL 3: RL from Canyon Entrance - (MODIFIED) - زيادة التدريب
    print("\n" + "="*50)
    print("=== CURRICULUM LEVEL 3: RL - NAVIGATING THE FULL CANYON ===")
    print("="*50)
    
    epsilon = 0.5
    for epoch in range(50): # زيادة الحلقات من 20 إلى 50
        print(f"\nRL Full Canyon - Epoch {epoch+1}/50, Epsilon: {epsilon:.2f}")
        episodes_data = []
        for i in range(60): # زيادة التجارب من 40 إلى 60
            start_pos = (random.uniform(13, 17), random.uniform(5, 8))
            start_theta = random.uniform(np.pi/4, 3*np.pi/4)
            _, success, sequences, actions, reward = run_episode(start_pos, start_theta, lstm_model, use_exploration=True, epsilon=epsilon, use_state_noise=True)
            if success: episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})
        
        if episodes_data:
            episodes_data.sort(key=lambda x: x['reward'], reverse=True)
            top_episodes = episodes_data[:len(episodes_data) // 2]
            X_fine_tune, y_fine_tune = [], []
            for ep_data in top_episodes:
                X_fine_tune.extend(ep_data['sequences'])
                y_fine_tune.extend(ep_data['actions'])
            if X_fine_tune:
                print(f"Fine-tuning on {len(X_fine_tune)} samples...")
                lstm_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=32, verbose=0)
        epsilon = max(0.1, epsilon * 0.96) # إبطاء معدل انخفاض إبسيلون

    # --- FINAL EVALUATION ---
    print("\n--- Part 4: Final Evaluation ---")
    plt.figure(figsize=(12, 12))
    ax = plt.gca()
    for obs in OBSTACLES: ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2]-obs[0], obs[3]-obs[1], color='gray'))
    for obs in INFLATED_OBSTACLES: ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2]-obs[0], obs[3]-obs[1], color='red', alpha=0.2, hatch='/'))
    plt.plot(GOAL_POS[0], GOAL_POS[1], 'r*', markersize=20, label='Goal')
    
    eval_start_pos = (15, 5) 
    eval_start_theta = np.pi / 2
    
    print("Evaluating Final Model...")
    path_ftr, success, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_model, use_state_noise=False) # No noise for final test
    if success: print("SUCCESS: Model reached the goal!")
    else: print("FAILURE: Model did not reach the goal.")
    
    path_ftr_np = np.array(path_ftr)
    plt.plot(path_ftr_np[:, 0], path_ftr_np[:, 1], 'm-', linewidth=2.5, label=f'Final Curriculum-Trained Model')
    
    plt.title("Evaluation of Final Model from Difficult Start")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show() 
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from collections import deque
import random
import time

# --- 1. Environment and Simulation ---

# (تعديل) تعريف نقاط التفتيش والهدف النهائي
CHECKPOINTS = [
    (15.0, 10.0), # مدخل الممر
    (15.0, 18.0), # منتصف الممر
    (25.0, 25.0)  # الهدف النهائي
]
FINAL_GOAL_POS = (25.0, 25.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 700 # زيادة طفيفة للسماح بالوصول
GOAL_RADIUS = 1.5 # زيادة نصف القطر قليلاً لتسهيل الوصول لنقاط التفتيش
SEQUENCE_LENGTH = 5

# Safety and Robot Size Constants
ROBOT_RADIUS = 0.5
PROXIMITY_PENALTY_WEIGHT = 1.5
MIN_SAFE_DISTANCE = 1.5
PROGRESS_REWARD_WEIGHT = 2.0

# Obstacles
OBSTACLES = np.array([
    [10, 0, 12, 15],
    [10, 20, 12, 30],
    [18, 10, 20, 25]
])
INFLATED_OBSTACLES = OBSTACLES + np.array([-ROBOT_RADIUS, -ROBOT_RADIUS, ROBOT_RADIUS, ROBOT_RADIUS])


class Robot:
    def __init__(self, x, y, theta):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega):
        omega = np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)
        self.theta += omega * DT
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += ROBOT_SPEED * math.cos(self.theta) * DT
        self.y += ROBOT_SPEED * math.sin(self.theta) * DT
        return (self.x, self.y), self.theta

# --- Helper Functions for Simulation ---

def get_lidar_readings_vectorized(robot_pos, robot_theta, num_sensors=5, max_dist=10.0, ray_step=0.1):
    readings = np.full(num_sensors, max_dist)
    angles = np.linspace(-np.pi / 2, np.pi / 2, num_sensors) + robot_theta
    dist_steps = np.arange(ray_step, max_dist, ray_step)
    ray_points_x = robot_pos[0] + dist_steps[:, np.newaxis] * np.cos(angles)
    ray_points_y = robot_pos[1] + dist_steps[:, np.newaxis] * np.sin(angles)
    for obs in OBSTACLES:
        in_obstacle = ((ray_points_x >= obs[0]) & (ray_points_x <= obs[2]) & (ray_points_y >= obs[1]) & (ray_points_y <= obs[3]))
        collision_indices = np.argmax(in_obstacle, axis=0)
        for i in range(num_sensors):
            if in_obstacle[collision_indices[i], i]:
                dist = dist_steps[collision_indices[i]]
                if dist < readings[i]: readings[i] = dist
    return readings

# (تعديل) دالة الحالة لتأخذ الهدف الحالي كمعامل
def get_state(robot_pos, robot_theta, current_goal, add_noise=False):
    lidar = get_lidar_readings_vectorized(robot_pos, robot_theta)
    
    dx = current_goal[0] - robot_pos[0]
    dy = current_goal[1] - robot_pos[1]
    
    dist_to_goal = math.sqrt(dx**2 + dy**2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = math.atan2(math.sin(angle_to_goal - robot_theta), math.cos(angle_to_goal - robot_theta))
    
    if add_noise:
        lidar += np.random.normal(0, 0.02, size=lidar.shape) * lidar
        dist_to_goal += np.random.normal(0, 0.02) * dist_to_goal
        angle_diff += np.random.normal(0, 0.02) * angle_diff

    state = np.concatenate((lidar / 10.0, [dist_to_goal / 30.0, angle_diff / np.pi]))
    return state


# --- 2. Models and Policies ---

def create_lstm_model(sequence_length, n_inputs, n_outputs):
    model = Sequential([
        Input(shape=(sequence_length, n_inputs)),
        LSTM(128, return_sequences=True),
        LSTM(128),
        Dense(64, activation='relu'),
        Dense(n_outputs, activation='tanh')
    ])
    # (تعديل) تخفيض معدل التعلم للمرحلة النهائية لتثبيت التدريب
    optimizer = tf.keras.optimizers.AdamW(learning_rate=1e-4)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    return model

@tf.function
def predict_action(model, nn_input):
    return model(nn_input, training=False)

# (تعديل كبير) دالة تشغيل الحلقة لتستخدم نظام نقاط التفتيش
def run_episode(start_pos, start_theta, policy_model, use_exploration=False, epsilon=0.0, use_state_noise=False):
    robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
    path_taken = [start_pos]
    state_history = deque(maxlen=SEQUENCE_LENGTH)
    episode_sequences, episode_actions = [], []
    
    checkpoint_index = 0
    current_goal = CHECKPOINTS[checkpoint_index]
    
    total_reward = 0
    
    for step in range(MAX_EPISODE_STEPS):
        pos = (robot.x, robot.y)
        if any(obs[0] <= pos[0] <= obs[2] and obs[1] <= pos[1] <= obs[3] for obs in INFLATED_OBSTACLES):
            return path_taken, False, [], [], total_reward - 50 # عقوبة كبيرة على الاصطدام

        current_state = get_state(pos, robot.theta, current_goal, add_noise=use_state_noise)
        state_history.append(current_state)
        
        if len(state_history) < SEQUENCE_LENGTH:
            omega = 0.1 * (random.random() - 0.5)
        else:
            sequence = np.array(state_history)
            nn_input = np.reshape(sequence, (1, SEQUENCE_LENGTH, 7))
            nn_input_tensor = tf.convert_to_tensor(nn_input, dtype=tf.float32)
            normalized_omega = predict_action(policy_model, nn_input_tensor).numpy()[0][0]
            if use_exploration and random.random() < epsilon:
                normalized_omega += random.uniform(-0.5, 0.5)
            omega = normalized_omega * MAX_TURN_RATE
            
            episode_sequences.append(sequence)
            episode_actions.append(omega / MAX_TURN_RATE)
            
        new_pos, _ = robot.move(omega)
        path_taken.append(new_pos)
        
        dist_to_checkpoint = math.dist(new_pos, current_goal)
        
        if dist_to_checkpoint < GOAL_RADIUS:
            total_reward += 50 # مكافأة الوصول لنقطة تفتيش
            checkpoint_index += 1
            if checkpoint_index >= len(CHECKPOINTS):
                print("SUCCESS: Reached the final goal!")
                total_reward += 100 # مكافأة إضافية للهدف النهائي
                return path_taken, True, episode_sequences, episode_actions, total_reward
            else:
                print(f"Reached checkpoint {checkpoint_index}, new goal: {CHECKPOINTS[checkpoint_index]}")
                current_goal = CHECKPOINTS[checkpoint_index]

    return path_taken, False, [], [], total_reward # الفشل بسبب انتهاء الوقت

# --- 3. Main Execution Block ---

if __name__ == "__main__":
    
    # لم نعد بحاجة للتعلم المنهجي المعقد، سنبدأ مباشرة بالتدريب على المهمة الكاملة
    # باستخدام نظام نقاط التفتيش الذي يسهل التعلم بشكل طبيعي
    
    print("\n" + "="*50)
    print("=== TRAINING WITH CHECKPOINT-BASED REWARD SYSTEM ===")
    print("="*50)
    
    lstm_model = create_lstm_model(SEQUENCE_LENGTH, 7, 1)
    
    epsilon = 0.7 # بدء باستكشاف أعلى قليلاً
    for epoch in range(100): # زيادة كبيرة في عدد الحلقات الإجمالي
        print(f"\nTraining Epoch {epoch+1}/100, Epsilon: {epsilon:.2f}")
        episodes_data = []
        
        # تشغيل 50 حلقة في كل دورة تدريبية
        for i in range(50):
            # البدء دائماً من نقطة البداية الصعبة
            start_pos = (random.uniform(13, 17), random.uniform(5, 8))
            start_theta = random.uniform(np.pi/4, 3*np.pi/4)
            
            path, success, sequences, actions, reward = run_episode(
                start_pos, start_theta, lstm_model, 
                use_exploration=True, epsilon=epsilon, use_state_noise=True
            )
            
            # سنقوم بالتدريب على الحلقات التي حققت بعض التقدم على الأقل
            if reward > 0 or success:
                print(f"  Episode {i+1}: {'Success' if success else 'Progress'}, Reward: {reward:.2f}")
                episodes_data.append({'reward': reward, 'sequences': sequences, 'actions': actions})

        if episodes_data:
            episodes_data.sort(key=lambda x: x['reward'], reverse=True)
            # اختيار أفضل 40% من الحلقات للتدريب عليها
            top_episodes = episodes_data[:int(len(episodes_data) * 0.4)]
            
            X_fine_tune, y_fine_tune = [], []
            for ep_data in top_episodes:
                X_fine_tune.extend(ep_data['sequences'])
                y_fine_tune.extend(ep_data['actions'])
                
            if X_fine_tune:
                print(f"Fine-tuning on {len(X_fine_tune)} samples from best episodes...")
                lstm_model.fit(np.array(X_fine_tune), np.array(y_fine_tune), epochs=5, batch_size=64, verbose=0)
                
        epsilon = max(0.1, epsilon * 0.97) # معدل انخفاض بطيء جداً للاستكشاف

    # --- FINAL EVALUATION ---
    print("\n--- Part 4: Final Evaluation ---")
    plt.figure(figsize=(12, 12))
    ax = plt.gca()
    for obs in OBSTACLES: ax.add_patch(plt.Rectangle((obs[0], obs[1]), obs[2]-obs[0], obs[3]-obs[1], color='gray'))
    # رسم نقاط التفتيش
    checkpoints_np = np.array(CHECKPOINTS)
    plt.plot(checkpoints_np[:, 0], checkpoints_np[:, 1], 'bo', markersize=10, label='Checkpoints')
    plt.plot(FINAL_GOAL_POS[0], FINAL_GOAL_POS[1], 'r*', markersize=20, label='Goal')
    
    eval_start_pos = (15, 5) 
    eval_start_theta = np.pi / 2
    
    print("Evaluating Final Model...")
    path_ftr, success, _, _, _ = run_episode(eval_start_pos, eval_start_theta, lstm_model, use_state_noise=False)
    if success: print("SUCCESS: Model reached the goal!")
    else: print("FAILURE: Model did not reach the goal.")
    
    path_ftr_np = np.array(path_ftr)
    plt.plot(path_ftr_np[:, 0], path_ftr_np[:, 1], 'm-', linewidth=2.5, label=f'Final Model Path')
    
    plt.title("Evaluation of Final Model with Checkpoints")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show() 

```
