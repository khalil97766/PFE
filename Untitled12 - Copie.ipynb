```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from shapely.geometry import LineString, Polygon, Point
# No osmnx needed if we are simulating camera-detected obstacles
# import osmnx as ox
# import geopandas as gpd # Still useful if camera output is geo-like

# --- Configuration ---

# 1. PSO Parameters (can be similar to before)
NUM_PARTICLES = 50
MAX_ITERATIONS = 150
NUM_WAYPOINTS = 7
DIMENSIONS = NUM_WAYPOINTS * 2
W = 0.5; C1 = 1.5; C2 = 1.5

# 2. Fitness function weights
W_LENGTH = 1.0
W_OBSTACLE = 200.0
W_BOUNDS = 30.0

# 3. Environment and Robot Setup
# These are now in a generic Cartesian coordinate system (e.g., meters)
# that your robot and camera operate in.
MAP_BOUNDS = {'x_min': 0, 'x_max': 100, 'y_min': 0, 'y_max': 80} # Example map size
START_POS = np.array([5, 40])
GOAL_POS = np.array([95, 40])

# --- Placeholder for Camera Data Processing ---
def get_obstacles_from_camera_data(robot_pose=None, camera_fov=None, max_detection_range=None):
    """
    SIMULATES detecting obstacles from camera data and returning them as Shapely Polygons.
    In a real system, this function would:
    1. Capture an image/depth data.
    2. Run object detection/segmentation.
    3. Project detected obstacles from image coordinates to world/map coordinates.
    4. Represent them as polygons.

    For this example, we'll just define some static obstacles as if they were detected.
    """
    print("Simulating obstacle detection from camera data...")
    simulated_obstacles_coords = [
        # Obstacle 1 (e.g., a box detected)
        [(20, 30), (20, 50), (30, 50), (30, 30)],
        # Obstacle 2 (e.g., another object)
        [(50, 10), (50, 35), (60, 35), (60, 10)],
        # Obstacle 3 (e.g., a wall segment or larger object)
        [(70, 55), (70, 70), (85, 70), (85, 55)],
        # Obstacle 4 (closer to start)
        [(10,5),(10,20),(15,20),(15,5)]
    ]
    
    # Ensure obstacles are within map bounds (or handle how out-of-bounds detections are treated)
    valid_simulated_obstacles = []
    for obs_coords in simulated_obstacles_coords:
        # Basic check if any point of obstacle is roughly within map
        # In reality, projection might place parts outside, or detection could be noisy
        temp_poly = Polygon(obs_coords)
        if temp_poly.bounds[0] < MAP_BOUNDS['x_max'] and temp_poly.bounds[2] > MAP_BOUNDS['x_min'] and \
           temp_poly.bounds[1] < MAP_BOUNDS['y_max'] and temp_poly.bounds[3] > MAP_BOUNDS['y_min']:
            valid_simulated_obstacles.append(Polygon(obs_coords))
        else:
            print(f"Simulated obstacle {obs_coords} largely outside map bounds, skipping.")


    print(f"Simulated detection of {len(valid_simulated_obstacles)} obstacles.")
    return valid_simulated_obstacles

# --- PSO Helper Functions (largely same as before) ---

def check_segment_collision(p1_coords, p2_coords, obstacles_list):
    segment = LineString([p1_coords, p2_coords])
    for obs_poly in obstacles_list:
        if segment.intersects(obs_poly):
            return True
    return False

def fitness_function(particle_waypoints_flat, start_pos_map, goal_pos_map, obstacles_list_map, map_bnds_map):
    waypoints = particle_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    full_path_nodes = np.vstack([start_pos_map, waypoints, goal_pos_map])
    cost_length = 0
    cost_obstacle = 0
    cost_bounds = 0
    for i in range(len(full_path_nodes) - 1):
        p1 = full_path_nodes[i]
        p2 = full_path_nodes[i+1]
        cost_length += np.linalg.norm(p2 - p1)
        if check_segment_collision(p1, p2, obstacles_list_map):
            cost_obstacle += 1
    for wp_idx in range(NUM_WAYPOINTS):
        wp_x = particle_waypoints_flat[wp_idx*2]
        wp_y = particle_waypoints_flat[wp_idx*2+1]
        if not (map_bnds_map['x_min'] <= wp_x <= map_bnds_map['x_max'] and \
                map_bnds_map['y_min'] <= wp_y <= map_bnds_map['y_max']):
            cost_bounds += 1
    total_fitness = (W_LENGTH * cost_length +
                     W_OBSTACLE * cost_obstacle +
                     W_BOUNDS * cost_bounds)
    return total_fitness

def plot_environment_camera_sim(iteration, particles_pos_flat, gbest_pos_flat,
                                start_map, goal_map, obstacles_map_list,
                                map_bnds_map, ax_plot, title_suffix=""):
    ax_plot.clear()
    # Plot obstacles detected by "camera"
    for obs_poly in obstacles_map_list:
        if obs_poly.exterior and len(obs_poly.exterior.coords) > 0:
            ax_plot.add_patch(patches.Polygon(obs_poly.exterior.coords, closed=True, color='maroon', alpha=0.7, zorder=2))

    ax_plot.plot(start_map[0], start_map[1], 'go', markersize=10, label='Start', zorder=5)
    ax_plot.plot(goal_map[0], goal_map[1], 'ro', markersize=10, label='Goal', zorder=5)

    if gbest_pos_flat is not None and len(gbest_pos_flat) > 0:
        gbest_waypoints = gbest_pos_flat.reshape(NUM_WAYPOINTS, 2)
        gbest_path_nodes = np.vstack([start_map, gbest_waypoints, goal_map])
        ax_plot.plot(gbest_path_nodes[:, 0], gbest_path_nodes[:, 1], 'b-', linewidth=2.5, label='Best Path', zorder=4)

    ax_plot.set_xlim(map_bnds_map['x_min'], map_bnds_map['x_max'])
    ax_plot.set_ylim(map_bnds_map['y_min'], map_bnds_map['y_max'])
    ax_plot.set_title(f"PSO with Camera-Simulated Obstacles - Iter: {iteration} {title_suffix}")
    ax_plot.set_xlabel("X coordinate (Map)")
    ax_plot.set_ylabel("Y coordinate (Map)")
    if not ax_plot.get_legend():
        ax_plot.legend(loc='best')
    ax_plot.set_aspect('equal', adjustable='box')
    plt.pause(0.01)


# --- Main Execution ---
if __name__ == "__main__":
    # --- 1. Get Obstacles (Simulated from Camera Data) ---
    # In a real robot, you might pass the robot's current pose if the detection
    # needs to know where the camera is in the world.
    DETECTED_OBSTACLES = get_obstacles_from_camera_data(robot_pose=START_POS)

    if not DETECTED_OBSTACLES:
        print("Warning: No obstacles were detected/simulated. Path will be straight.")

    # --- 2. PSO Initialization ---
    particles_pos = np.zeros((NUM_PARTICLES, DIMENSIONS))
    for i in range(NUM_PARTICLES):
        for d_wp in range(NUM_WAYPOINTS):
            particles_pos[i, d_wp*2] = random.uniform(MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            particles_pos[i, d_wp*2+1] = random.uniform(MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])

    particles_vel = np.random.randn(NUM_PARTICLES, DIMENSIONS) * 0.1
    pbest_pos = np.copy(particles_pos)
    pbest_fitness = np.array([fitness_function(p, START_POS, GOAL_POS, DETECTED_OBSTACLES, MAP_BOUNDS) for p in pbest_pos])
    gbest_idx = np.argmin(pbest_fitness)
    gbest_pos = np.copy(pbest_pos[gbest_idx])
    gbest_fitness = pbest_fitness[gbest_idx]

    fig_pso, ax_pso = plt.subplots(figsize=(10, 8))
    plt.ion()

    print(f"\nStarting PSO with camera-simulated obstacles. Initial best fitness: {gbest_fitness:.2f}")
    plot_environment_camera_sim(0, particles_pos, gbest_pos,
                                START_POS, GOAL_POS, DETECTED_OBSTACLES,
                                MAP_BOUNDS, ax_pso)

    # --- 3. PSO Main Loop ---
    for iteration in range(MAX_ITERATIONS):
        for i in range(NUM_PARTICLES):
            r1 = np.random.rand(DIMENSIONS); r2 = np.random.rand(DIMENSIONS)
            particles_vel[i] = (W * particles_vel[i] +
                                C1 * r1 * (pbest_pos[i] - particles_pos[i]) +
                                C2 * r2 * (gbest_pos - particles_pos[i]))
            
            vel_limit_x = (MAP_BOUNDS['x_max'] - MAP_BOUNDS['x_min']) * 0.05
            vel_limit_y = (MAP_BOUNDS['y_max'] - MAP_BOUNDS['y_min']) * 0.05
            particles_vel[i, ::2] = np.clip(particles_vel[i, ::2], -vel_limit_x, vel_limit_x)
            particles_vel[i, 1::2] = np.clip(particles_vel[i, 1::2], -vel_limit_y, vel_limit_y)

            particles_pos[i] += particles_vel[i]
            current_waypoints = particles_pos[i].reshape(NUM_WAYPOINTS, 2)
            current_waypoints[:, 0] = np.clip(current_waypoints[:, 0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            current_waypoints[:, 1] = np.clip(current_waypoints[:, 1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
            particles_pos[i] = current_waypoints.flatten()

            current_fitness = fitness_function(particles_pos[i], START_POS, GOAL_POS, DETECTED_OBSTACLES, MAP_BOUNDS)
            if current_fitness < pbest_fitness[i]:
                pbest_fitness[i] = current_fitness
                pbest_pos[i] = np.copy(particles_pos[i])
                if current_fitness < gbest_fitness:
                    gbest_fitness = current_fitness
                    gbest_pos = np.copy(particles_pos[i])

        if (iteration + 1) % 10 == 0 or iteration == MAX_ITERATIONS - 1:
            plot_environment_camera_sim(iteration + 1, particles_pos, gbest_pos,
                                        START_POS, GOAL_POS, DETECTED_OBSTACLES,
                                        MAP_BOUNDS, ax_pso)
            print(f"Iter {iteration+1}/{MAX_ITERATIONS}: Best Fitness = {gbest_fitness:.2f}")

    # --- 4. Output Final Result ---
    final_waypoints_flat = gbest_pos
    final_waypoints_map = final_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    final_path_nodes_map = np.vstack([START_POS, final_waypoints_map, GOAL_POS])
    plt.figure(figsize=(4, 3))
    print("\n--- PSO Finished (Camera-Simulated Obstacles) ---")
    print(f"Best Fitness Found: {gbest_fitness:.2f}")
    print("Optimal Path (Map Coordinates):")
    for node_idx, node in enumerate(final_path_nodes_map):
        print(f"  Node {node_idx}: ({node[0]:.2f}, {node[1]:.2f})")

    plot_environment_camera_sim(MAX_ITERATIONS, particles_pos, gbest_pos,
                                START_POS, GOAL_POS, DETECTED_OBSTACLES,
                                MAP_BOUNDS, ax_pso, title_suffix=" - Final")
    plt.ioff()
    plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from shapely.geometry import LineString, Polygon, Point

# --- Configuration ---
NUM_PARTICLES = 80  # Increased
MAX_ITERATIONS = 300 # Increased
NUM_WAYPOINTS = 15   # <<-- Significantly Increased
DIMENSIONS = NUM_WAYPOINTS * 2

# Inertia weight decay parameters
W_MAX = 0.9
W_MIN = 0.4
C1 = 1.8 # Slightly higher emphasis on personal best
C2 = 1.8 # Slightly higher emphasis on global best

# Fitness function weights
W_LENGTH = 1.0
W_OBSTACLE_MULTIPLIER = 1.0 # For normal obstacles
NORMAL_OBSTACLE_PENALTY_CONSTANT = 20000.0
CRITICAL_OBSTACLE_PENALTY = float('inf') # The strongest penalty
W_BOUNDS = 70.0 # Increased penalty for going out of bounds

# Environment and Robot Setup
MAP_BOUNDS = {'x_min': 0, 'x_max': 100, 'y_min': 0, 'y_max': 55} # Adjusted y_max to fit image
START_POS = np.array([5, 40])
GOAL_POS = np.array([95, 40])
ROBOT_RADIUS = 1.0 # Adjust as needed. Set to 0 to disable inflation.

# --- Obstacle Simulation, Inflation, and Critical Marking ---
def get_and_process_obstacles(robot_radius):
    print("Processing simulated obstacles...")
    # Structure: list of dicts, each with 'coords' and 'is_critical'
    # Coordinates from the image (approximate)
    simulated_obstacles_data = [
        {'coords': [(15, 29), (15, 51), (30, 51), (30, 29)], 'is_critical': False, 'color': 'rosybrown'}, # Normal - Brownish one
        {'coords': [(48, 9), (48, 37), (60, 37), (60, 9)], 'is_critical': True, 'color': 'red'}     # Critical - Red one
        # Add more obstacles if needed:
        # {'coords': [(70, 5), (70, 25), (80, 25), (80, 5)], 'is_critical': False, 'color':'gray'}
    ]
    
    all_inflated_obstacles = []
    critical_inflated_obstacles = []
    normal_inflated_obstacles = []
    obstacle_colors = [] # To store colors for plotting

    for obs_data in simulated_obstacles_data:
        coords = obs_data['coords']
        is_critical = obs_data['is_critical']
        color = obs_data['color']
        
        original_poly = Polygon(coords)
        if not original_poly.is_valid:
            # print(f"DEBUG: Original simulated obstacle is invalid: {original_poly.wkt[:50]}")
            continue
        
        # Optional: Check if original polygon is within map bounds before inflation (can save computation)
        # if not (original_poly.bounds[0] < MAP_BOUNDS['x_max'] and original_poly.bounds[2] > MAP_BOUNDS['x_min'] and \
        #         original_poly.bounds[1] < MAP_BOUNDS['y_max'] and original_poly.bounds[3] > MAP_BOUNDS['y_min']):
        #     # print(f"DEBUG: Simulated obstacle {coords} largely outside map bounds, skipping.")
        #     continue

        inflated_poly = original_poly
        if robot_radius > 0:
            inflated_poly = original_poly.buffer(robot_radius, resolution=8)
            if not inflated_poly.is_valid:
                # print(f"DEBUG: Inflated obstacle (from {coords}) became invalid: {inflated_poly.wkt[:50]}")
                continue
        
        all_inflated_obstacles.append(inflated_poly)
        obstacle_colors.append(color) # Store color for the inflated obstacle

        if is_critical:
            critical_inflated_obstacles.append(inflated_poly)
            # print(f"DEBUG: Added CRITICAL inflated obstacle, bounds: {inflated_poly.bounds}")
        else:
            normal_inflated_obstacles.append(inflated_poly)
            # print(f"DEBUG: Added NORMAL inflated obstacle, bounds: {inflated_poly.bounds}")

    print(f"Generated {len(all_inflated_obstacles)} total inflated obstacles.")
    print(f"  {len(critical_inflated_obstacles)} critical inflated obstacles.")
    print(f"  {len(normal_inflated_obstacles)} normal inflated obstacles.")
    return all_inflated_obstacles, critical_inflated_obstacles, normal_inflated_obstacles, obstacle_colors

# --- PSO Helper Functions ---
def check_segment_collision_with_list(p1_coords, p2_coords, obstacles_to_check):
    if np.array_equal(p1_coords, p2_coords): return False
    try: segment = LineString([p1_coords, p2_coords])
    except Exception as e:
        # print(f"DEBUG: Error creating LineString from {p1_coords} to {p2_coords}: {e}")
        return True # Assume collision if line cannot be formed
    for obs_poly in obstacles_to_check:
        if segment.intersects(obs_poly):
            return True # Collision detected with one of the obstacles in the list
    return False # No collision with any obstacle in the list

def fitness_function(particle_waypoints_flat, start_pos_map, goal_pos_map, 
                     normal_obs_list, critical_obs_list, map_bnds_map,
                     particle_idx_debug=-1, iteration_debug=-1): # Debug params
    waypoints = particle_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    full_path_nodes = np.vstack([start_pos_map, waypoints, goal_pos_map])
    
    path_length = 0.0
    num_normal_collisions = 0
    num_critical_collisions = 0
    num_out_of_bounds = 0

    for i in range(len(full_path_nodes) - 1):
        p1 = full_path_nodes[i]
        p2 = full_path_nodes[i+1]
        path_length += np.linalg.norm(p2 - p1)

        # Check CRITICAL collisions first for this segment
        if check_segment_collision_with_list(p1, p2, critical_obs_list):
            num_critical_collisions += 1
            # if particle_idx_debug == 0 and iteration_debug % 20 == 0: # Selective debug
            #     print(f"DEBUG fitness (P{particle_idx_debug},Iter{iteration_debug}): Seg {i} HIT CRITICAL. Path Pts: {particle_waypoints_flat[:4]}")
            # No need to check normal if critical is hit for this segment,
            # as the path will be penalized with infinity.
            # We count all critical collisions to understand the extent if debugging.
            
        # Check NORMAL collisions for this segment (only if not already critically hit, for precise counting, though 'inf' dominates)
        # For simplicity, we check critical first, then normal. If any critical, whole path is 'inf'.
        elif check_segment_collision_with_list(p1, p2, normal_obs_list):
            num_normal_collisions += 1
            # if particle_idx_debug == 0 and iteration_debug % 20 == 0:
            #     print(f"DEBUG fitness (P{particle_idx_debug},Iter{iteration_debug}): Seg {i} HIT NORMAL. Path Pts: {particle_waypoints_flat[:4]}")
            
    for wp_idx in range(NUM_WAYPOINTS):
        wp_x = particle_waypoints_flat[wp_idx*2]
        wp_y = particle_waypoints_flat[wp_idx*2+1]
        if not (map_bnds_map['x_min'] <= wp_x <= map_bnds_map['x_max'] and \
                map_bnds_map['y_min'] <= wp_y <= map_bnds_map['y_max']):
            num_out_of_bounds += 1
    
    if num_critical_collisions > 0:
        # if particle_idx_debug == 0 and iteration_debug % 20 == 0:
        #     print(f"DEBUG fitness (P{particle_idx_debug},Iter{iteration_debug}): RETURNING CRITICAL_PENALTY ({num_critical_collisions} critical hits).")
        return CRITICAL_OBSTACLE_PENALTY 
    
    fitness_val = (W_LENGTH * path_length +
                   W_OBSTACLE_MULTIPLIER * num_normal_collisions * NORMAL_OBSTACLE_PENALTY_CONSTANT +
                   W_BOUNDS * num_out_of_bounds)
    
    # if particle_idx_debug == 0 and iteration_debug % 20 == 0:
    #     print(f"DEBUG fitness (P{particle_idx_debug},Iter{iteration_debug}): Total={fitness_val:.2f} (L:{path_length:.2f}, NormC:{num_normal_collisions}, CritC:{num_critical_collisions}, OB:{num_out_of_bounds})")
    return fitness_val


def plot_environment(iteration, gbest_pos_flat,
                     start_map, goal_map, 
                     all_obstacles_infl, critical_obstacles_infl, obstacle_cols, # Pass colors
                     map_bnds_map, ax_plot, title_suffix=""):
    ax_plot.clear()
    # Plot all inflated obstacles with their original specified colors
    for i, obs_poly in enumerate(all_obstacles_infl):
        color_to_use = obstacle_cols[i] # Get the stored color
        # Make critical obstacles more prominent if needed, though color already defines it
        alpha_val = 0.7
        # for crit_obs in critical_obstacles_infl:
        #     if obs_poly.equals(crit_obs):
        #         color_to_use = 'red' # Override to ensure critical are red
        #         alpha_val = 0.9
        #         break
        if obs_poly.exterior and len(obs_poly.exterior.coords) > 0 :
            ax_plot.add_patch(patches.Polygon(obs_poly.exterior.coords, closed=True, color=color_to_use, alpha=alpha_val, zorder=2))
    
    ax_plot.plot(start_map[0], start_map[1], 'go', markersize=10, label='Start', zorder=5) # Green start
    ax_plot.plot(goal_map[0], goal_map[1], 'mo', markersize=10, label='Goal', zorder=5)   # Magenta goal

    if gbest_pos_flat is not None and len(gbest_pos_flat) > 0 and np.isfinite(gbest_pos_flat).all():
        gbest_waypoints = gbest_pos_flat.reshape(NUM_WAYPOINTS, 2)
        # Ensure the path starts from START_POS for plotting
        gbest_path_nodes_plot = np.vstack([start_map, gbest_waypoints, goal_map])
        ax_plot.plot(gbest_path_nodes_plot[:, 0], gbest_path_nodes_plot[:, 1], 'b-', linewidth=2.5, label='Best Path', zorder=4)

    ax_plot.set_xlim(map_bnds_map['x_min'], map_bnds_map['x_max'])
    ax_plot.set_ylim(map_bnds_map['y_min'], map_bnds_map['y_max'])
    ax_plot.set_title(f"PSO Path Planning - Iter: {iteration} {title_suffix}")
    ax_plot.set_xlabel("X (Map)")
    ax_plot.set_ylabel("Y (Map)")
    
    handles, labels = ax_plot.get_legend_handles_labels()
    if handles: # Only show legend if there are items to show
         ax_plot.legend(loc='best')
    else: # If no handles (e.g. gbest is inf and not plotted), create a dummy legend or skip
        pass

    ax_plot.set_aspect('equal', adjustable='box')
    plt.pause(0.01)

# --- Main Execution ---
if __name__ == "__main__":
    ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, NORMAL_INFLATED_OBSTACLES, OBSTACLE_COLORS = \
        get_and_process_obstacles(ROBOT_RADIUS)

    if not ALL_INFLATED_OBSTACLES: print("Warning: No obstacles available.")
    if not CRITICAL_INFLATED_OBSTACLES: print("Info: No obstacles marked as critical.")


    particles_pos = np.zeros((NUM_PARTICLES, DIMENSIONS))
    for i in range(NUM_PARTICLES):
        for d_wp in range(NUM_WAYPOINTS):
            particles_pos[i, d_wp*2] = random.uniform(MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            particles_pos[i, d_wp*2+1] = random.uniform(MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])

    particles_vel = np.random.randn(NUM_PARTICLES, DIMENSIONS) * 0.1 # Initial small random velocities
    pbest_pos = np.copy(particles_pos)
    pbest_fitness = np.full(NUM_PARTICLES, np.inf)
    gbest_pos = np.full(DIMENSIONS, np.nan) # Initialize gbest_pos with NaN to indicate no valid path yet
    gbest_fitness = np.inf

    for i in range(NUM_PARTICLES):
        fitness_val = fitness_function(particles_pos[i], START_POS, GOAL_POS, 
                                       NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, 
                                       MAP_BOUNDS)
        pbest_fitness[i] = fitness_val
        # pbest_pos[i] remains particles_pos[i] (already copied)
        if fitness_val < gbest_fitness:
            gbest_fitness = fitness_val
            gbest_pos = np.copy(particles_pos[i])

    fig_pso, ax_pso = plt.subplots(figsize=(10, 7)) # Adjusted figure size
    plt.ion()

    print(f"\nStarting PSO. Initial best fitness: {gbest_fitness:.2f}")
    plot_environment(0, gbest_pos, # Pass gbest_pos directly
                     START_POS, GOAL_POS, 
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, title_suffix="Initial State")

    for iteration in range(MAX_ITERATIONS):
        current_W = W_MAX - (W_MAX - W_MIN) * (iteration / MAX_ITERATIONS) # Inertia weight decay

        num_finite_particles_this_iter = 0
        for i in range(NUM_PARTICLES):
            r1 = np.random.rand(DIMENSIONS); r2 = np.random.rand(DIMENSIONS)
            
            # Ensure pbest_pos[i] and gbest_pos are finite for velocity calculation
            # If gbest is still inf (no valid path found yet), particles might not move towards it effectively.
            # A common strategy is to let particles move more randomly or only based on pbest if gbest is invalid.
            gbest_term = np.zeros_like(particles_pos[i])
            if np.isfinite(gbest_fitness): # Only use gbest if it's a valid path
                 gbest_term = C2 * r2 * (gbest_pos - particles_pos[i])

            particles_vel[i] = (current_W * particles_vel[i] +
                                C1 * r1 * (pbest_pos[i] - particles_pos[i]) +
                                gbest_term)
            
            vel_limit_x = (MAP_BOUNDS['x_max'] - MAP_BOUNDS['x_min']) * 0.035 # Adjusted velocity limit
            vel_limit_y = (MAP_BOUNDS['y_max'] - MAP_BOUNDS['y_min']) * 0.035
            particles_vel[i, ::2] = np.clip(particles_vel[i, ::2], -vel_limit_x, vel_limit_x)
            particles_vel[i, 1::2] = np.clip(particles_vel[i, 1::2], -vel_limit_y, vel_limit_y)
            
            particles_pos[i] += particles_vel[i]
            
            current_waypoints = particles_pos[i].reshape(NUM_WAYPOINTS, 2)
            current_waypoints[:, 0] = np.clip(current_waypoints[:, 0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            current_waypoints[:, 1] = np.clip(current_waypoints[:, 1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
            particles_pos[i] = current_waypoints.flatten()

            current_fitness = fitness_function(particles_pos[i], START_POS, GOAL_POS, 
                                               NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                               MAP_BOUNDS, particle_idx_debug=i, iteration_debug=iteration) # Pass debug params
            
            if np.isfinite(current_fitness):
                num_finite_particles_this_iter +=1

            if current_fitness < pbest_fitness[i]:
                pbest_fitness[i] = current_fitness
                pbest_pos[i] = np.copy(particles_pos[i])
                if current_fitness < gbest_fitness:
                    gbest_fitness = current_fitness
                    gbest_pos = np.copy(particles_pos[i])
                    # print(f"Iter {iteration}, P{i}: New gbest! Fitness: {gbest_fitness:.2f}")


        if (iteration + 1) % 10 == 0 or iteration == MAX_ITERATIONS - 1:
            plot_environment(iteration + 1, gbest_pos,
                             START_POS, GOAL_POS, 
                             ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                             MAP_BOUNDS, ax_pso)
            print(f"Iter {iteration+1}/{MAX_ITERATIONS}: W={current_W:.2f}, BestFit={gbest_fitness:.2f}, FiniteParticles={num_finite_particles_this_iter}/{NUM_PARTICLES}")
            
            if np.isfinite(gbest_fitness) and gbest_fitness < NORMAL_OBSTACLE_PENALTY_CONSTANT / 100 : # If best fitness is significantly less than one normal obstacle penalty
                # And check if gbest itself has no critical collisions
                gbest_path_nodes_check = np.vstack([START_POS, gbest_pos.reshape(NUM_WAYPOINTS, 2), GOAL_POS])
                crit_coll_gbest = 0
                for k_seg in range(len(gbest_path_nodes_check) - 1):
                    if check_segment_collision_with_list(gbest_path_nodes_check[k_seg], gbest_path_nodes_check[k_seg+1], CRITICAL_INFLATED_OBSTACLES):
                        crit_coll_gbest +=1
                        break
                if crit_coll_gbest == 0:
                    print(f"--- Potentially converged early at iter {iteration+1} ---")
                    # break # Uncomment for early stopping

    # Final Path Construction and Output
    if not np.isfinite(gbest_fitness):
        print("\n--- PSO Finished: FAILED TO FIND A VALID PATH (gbest is inf) ---")
        final_path_nodes_map = np.array([START_POS, GOAL_POS]) # Show straight line if no path
    else:
        final_waypoints_map = gbest_pos.reshape(NUM_WAYPOINTS, 2)
        final_path_nodes_map = np.vstack([START_POS, final_waypoints_map, GOAL_POS])
        print("\n--- PSO Finished ---")
        print(f"Best Fitness Found: {gbest_fitness:.2f}")
    
    final_path_critical_collisions = 0
    final_path_normal_collisions = 0
    for i in range(len(final_path_nodes_map) - 1):
        p1_node = final_path_nodes_map[i]
        p2_node = final_path_nodes_map[i+1]
        if check_segment_collision_with_list(p1_node, p2_node, CRITICAL_INFLATED_OBSTACLES):
            final_path_critical_collisions += 1
        elif check_segment_collision_with_list(p1_node, p2_node, NORMAL_INFLATED_OBSTACLES):
            final_path_normal_collisions += 1

    if final_path_critical_collisions == 0: print("Final best path is CRITICALLY collision-free.")
    else: print(f"WARNING: Final best path has {final_path_critical_collisions} CRITICAL colliding segments.")
    if final_path_normal_collisions == 0: print("Final best path is also normally collision-free.")
    else: print(f"Final best path has {final_path_normal_collisions} normal colliding segments.")

    print("Optimal Path (Map Coordinates):")
    for node_idx, node in enumerate(final_path_nodes_map): print(f"  Node {node_idx}: ({node[0]:.2f}, {node[1]:.2f})")

    plot_environment(MAX_ITERATIONS, gbest_pos,
                     START_POS, GOAL_POS, 
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, title_suffix=" - Final")
    plt.ioff()
    plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from shapely.geometry import LineString, Polygon, Point

# --- Configuration ---
NUM_PARTICLES = 150  # Increased
MAX_ITERATIONS = 500 # Increased
NUM_WAYPOINTS = 15   # Kept high for flexibility
DIMENSIONS = NUM_WAYPOINTS * 2

# Inertia weight decay parameters
W_MAX = 0.9
W_MIN = 0.4
C1 = 1.8
C2 = 1.8

# Fitness function weights
W_LENGTH = 1.0
W_OBSTACLE_MULTIPLIER = 1.0
NORMAL_OBSTACLE_PENALTY_CONSTANT = 20000.0
CRITICAL_OBSTACLE_PENALTY = float('inf')
W_BOUNDS = 70.0

# Environment and Robot Setup
MAP_BOUNDS = {'x_min': 0, 'x_max': 100, 'y_min': 0, 'y_max': 55}
START_POS = np.array([5, 40])
GOAL_POS = np.array([95, 40])
ROBOT_RADIUS = 0.1 # <<-- SET TO A VERY SMALL VALUE OR 0 FOR THIS DEBUGGING RUN

# --- Obstacle Simulation, Inflation, and Critical Marking ---
def get_and_process_obstacles(robot_radius):
    print("Processing simulated obstacles...")
    simulated_obstacles_data = [
        {'coords': [(15, 29), (15, 51), (30, 51), (30, 29)], 'is_critical': False, 'color': 'rosybrown'},
        {'coords': [(48, 9), (48, 37), (60, 37), (60, 9)], 'is_critical': True, 'color': 'red'}
    ]
    
    all_inflated_obstacles = []
    critical_inflated_obstacles = []
    normal_inflated_obstacles = []
    obstacle_colors = []

    for obs_data in simulated_obstacles_data:
        coords = obs_data['coords']
        is_critical = obs_data['is_critical']
        color = obs_data['color']
        
        original_poly = Polygon(coords)
        if not original_poly.is_valid:
            print(f"DEBUG: Original simulated obstacle is invalid: {original_poly.wkt[:50]}")
            continue
        
        inflated_poly = original_poly
        if robot_radius > 0:
            inflated_poly = original_poly.buffer(robot_radius, resolution=8)
            if not inflated_poly.is_valid:
                print(f"DEBUG: Inflated obstacle (from {coords}) became invalid: {inflated_poly.wkt[:50]}")
                continue
        
        all_inflated_obstacles.append(inflated_poly)
        obstacle_colors.append(color)

        if is_critical:
            critical_inflated_obstacles.append(inflated_poly)
            # print(f"DEBUG: Added CRITICAL inflated obstacle, bounds: {inflated_poly.bounds}")
        else:
            normal_inflated_obstacles.append(inflated_poly)
            # print(f"DEBUG: Added NORMAL inflated obstacle, bounds: {inflated_poly.bounds}")

    print(f"Generated {len(all_inflated_obstacles)} total inflated obstacles.")
    print(f"  {len(critical_inflated_obstacles)} critical inflated obstacles.")
    print(f"  {len(normal_inflated_obstacles)} normal inflated obstacles.")
    return all_inflated_obstacles, critical_inflated_obstacles, normal_inflated_obstacles, obstacle_colors

# --- PSO Helper Functions ---
def check_segment_collision_with_list(p1_coords, p2_coords, obstacles_to_check):
    if np.array_equal(p1_coords, p2_coords): return False
    try: segment = LineString([p1_coords, p2_coords])
    except Exception as e: return True 
    for obs_poly in obstacles_to_check:
        if segment.intersects(obs_poly): return True
    return False

def fitness_function(particle_waypoints_flat, start_pos_map, goal_pos_map, 
                     normal_obs_list, critical_obs_list, map_bnds_map,
                     particle_idx_debug=-1, iteration_debug=-1):
    waypoints = particle_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    full_path_nodes = np.vstack([start_pos_map, waypoints, goal_pos_map])
    
    path_length = 0.0
    num_normal_collisions = 0
    num_critical_collisions = 0
    num_out_of_bounds = 0

    # UNCOMMENT THE DEBUG BLOCK BELOW IF EVEN GUIDED PARTICLE IS INF
    # detailed_debug_active = (particle_idx_debug == 0) # and iteration_debug < 2) # Only for first particle for first few iterations

    for i in range(len(full_path_nodes) - 1):
        p1 = full_path_nodes[i]
        p2 = full_path_nodes[i+1]
        path_length += np.linalg.norm(p2 - p1)

        segment_hit_critical_this_pass = False
        if check_segment_collision_with_list(p1, p2, critical_obs_list):
            num_critical_collisions += 1
            segment_hit_critical_this_pass = True
            # if detailed_debug_active:
            #     print(f"  FIT_DBG (P{particle_idx_debug},Iter{iteration_debug},Seg{i}): {p1.round(1)}->{p2.round(1)} HIT CRITICAL.")
        
        # Only check normal if not critically hit (to avoid double counting penalty logic, though inf dominates)
        if not segment_hit_critical_this_pass: 
            if check_segment_collision_with_list(p1, p2, normal_obs_list):
                num_normal_collisions += 1
                # if detailed_debug_active:
                #     print(f"  FIT_DBG (P{particle_idx_debug},Iter{iteration_debug},Seg{i}): {p1.round(1)}->{p2.round(1)} HIT NORMAL.")
            
    for wp_idx in range(NUM_WAYPOINTS):
        wp_x = particle_waypoints_flat[wp_idx*2]
        wp_y = particle_waypoints_flat[wp_idx*2+1]
        if not (map_bnds_map['x_min'] <= wp_x <= map_bnds_map['x_max'] and \
                map_bnds_map['y_min'] <= wp_y <= map_bnds_map['y_max']):
            num_out_of_bounds += 1
    
    if num_critical_collisions > 0:
        # if detailed_debug_active:
        #     print(f"  FIT_DBG (P{particle_idx_debug},Iter{iteration_debug}): RETURNING INF ({num_critical_collisions} critical hits).")
        return CRITICAL_OBSTACLE_PENALTY 
    
    fitness_val = (W_LENGTH * path_length +
                   W_OBSTACLE_MULTIPLIER * num_normal_collisions * NORMAL_OBSTACLE_PENALTY_CONSTANT +
                   W_BOUNDS * num_out_of_bounds)
    
    # if detailed_debug_active:
    #     print(f"  FIT_DBG (P{particle_idx_debug},Iter{iteration_debug}): Total={fitness_val:.2f} (L:{path_length:.2f}, NormC:{num_normal_collisions}, CritC:{num_critical_collisions}, OB:{num_out_of_bounds})")
    return fitness_val

def plot_environment(iteration, gbest_pos_flat,
                     start_map, goal_map, 
                     all_obstacles_infl, critical_obstacles_infl, obstacle_cols,
                     map_bnds_map, ax_plot, title_suffix=""):
    ax_plot.clear()
    for i, obs_poly in enumerate(all_obstacles_infl):
        color_to_use = obstacle_cols[i]
        alpha_val = 0.7
        is_this_critical = False
        for crit_obs_check in critical_obstacles_infl: # Check if current obs is in critical list
            if obs_poly.equals(crit_obs_check):
                is_this_critical = True
                break
        if is_this_critical:
            alpha_val = 0.9 # Make critical more solid
            # color_to_use = 'red' # Force critical to be red if color mapping is off

        if obs_poly.exterior and len(obs_poly.exterior.coords) > 0 :
            ax_plot.add_patch(patches.Polygon(obs_poly.exterior.coords, closed=True, color=color_to_use, alpha=alpha_val, zorder=2))
    
    ax_plot.plot(start_map[0], start_map[1], 'go', markersize=10, label='Start', zorder=5)
    ax_plot.plot(goal_map[0], goal_map[1], 'mo', markersize=10, label='Goal', zorder=5)

    if gbest_pos_flat is not None and len(gbest_pos_flat) > 0 and np.isfinite(gbest_pos_flat).all() and np.isfinite(gbest_fitness): # Added gbest_fitness check
        gbest_waypoints = gbest_pos_flat.reshape(NUM_WAYPOINTS, 2)
        gbest_path_nodes_plot = np.vstack([start_map, gbest_waypoints, goal_map])
        ax_plot.plot(gbest_path_nodes_plot[:, 0], gbest_path_nodes_plot[:, 1], 'b-', linewidth=2.5, label='Best Path', zorder=4)

    ax_plot.set_xlim(map_bnds_map['x_min'], map_bnds_map['x_max'])
    ax_plot.set_ylim(map_bnds_map['y_min'], map_bnds_map['y_max'])
    ax_plot.set_title(f"PSO Path Planning - Iter: {iteration} {title_suffix}")
    ax_plot.set_xlabel("X (Map)")
    ax_plot.set_ylabel("Y (Map)")
    handles, labels = ax_plot.get_legend_handles_labels()
    if handles: ax_plot.legend(loc='best')
    ax_plot.set_aspect('equal', adjustable='box')
    plt.pause(0.01)

# --- Main Execution ---
if __name__ == "__main__":
    ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, NORMAL_INFLATED_OBSTACLES, OBSTACLE_COLORS = \
        get_and_process_obstacles(ROBOT_RADIUS)

    if not ALL_INFLATED_OBSTACLES: print("Warning: No obstacles available.")
    if not CRITICAL_INFLATED_OBSTACLES: print("Warning: No obstacles marked as critical by get_and_process_obstacles.")


    particles_pos = np.zeros((NUM_PARTICLES, DIMENSIONS))
    for i in range(NUM_PARTICLES):
        for d_wp in range(NUM_WAYPOINTS):
            particles_pos[i, d_wp*2] = random.uniform(MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            particles_pos[i, d_wp*2+1] = random.uniform(MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])

    # <<<< TEMPORARY GUIDED INITIALIZATION FOR DEBUGGING >>>>
    if NUM_PARTICLES > 0 and len(CRITICAL_INFLATED_OBSTACLES) > 0:
        print("DEBUG: Guiding first particle's initial waypoints...")
        guided_wps_array = np.zeros((NUM_WAYPOINTS, 2))
        
        # Target: Go above the first critical obstacle. Critical obstacle y-range approx [9,37] + radius.
        # If ROBOT_RADIUS=0.1, inflated y_max_crit approx 37.1. Let's aim for y=45.
        # Normal obstacle y-range approx [29,51] + radius.
        
        # Point before critical, high Y
        pt1_x = (CRITICAL_INFLATED_OBSTACLES[0].bounds[0] + ALL_INFLATED_OBSTACLES[0].bounds[2]) /2 # X between start and critical
        pt1_y = 45.0 
        if NORMAL_INFLATED_OBSTACLES: # If normal obstacle exists
             pt1_x = (NORMAL_INFLATED_OBSTACLES[0].bounds[2] + CRITICAL_INFLATED_OBSTACLES[0].bounds[0])/2 # X between normal and critical

        # Point after critical, high Y
        pt2_x = (CRITICAL_INFLATED_OBSTACLES[0].bounds[2] + GOAL_POS[0]) / 2 # X between critical and goal
        pt2_y = 45.0

        # Distribute waypoints along a path: START -> pt1 -> pt2 -> GOAL
        # This crude path has 3 segments. We have NUM_WAYPOINTS.
        # Assign roughly NUM_WAYPOINTS/3 to each segment.

        num_wp_seg1 = NUM_WAYPOINTS // 3
        num_wp_seg2 = NUM_WAYPOINTS // 3
        num_wp_seg3 = NUM_WAYPOINTS - num_wp_seg1 - num_wp_seg2

        current_wp_idx = 0
        # Segment 1: START_POS to pt1
        for k in range(num_wp_seg1):
            t = (k + 1) / (num_wp_seg1)
            guided_wps_array[current_wp_idx, 0] = START_POS[0] + t * (pt1_x - START_POS[0])
            guided_wps_array[current_wp_idx, 1] = START_POS[1] + t * (pt1_y - START_POS[1])
            current_wp_idx += 1
        
        # Segment 2: pt1 to pt2
        for k in range(num_wp_seg2):
            t = (k + 1) / (num_wp_seg2)
            guided_wps_array[current_wp_idx, 0] = pt1_x + t * (pt2_x - pt1_x)
            guided_wps_array[current_wp_idx, 1] = pt1_y + t * (pt2_y - pt1_y) # Stays high
            current_wp_idx += 1

        # Segment 3: pt2 to GOAL_POS
        for k in range(num_wp_seg3):
            t = (k + 1) / (num_wp_seg3)
            guided_wps_array[current_wp_idx, 0] = pt2_x + t * (GOAL_POS[0] - pt2_x)
            guided_wps_array[current_wp_idx, 1] = pt2_y + t * (GOAL_POS[1] - pt2_y) # Come down to goal Y
            current_wp_idx += 1
            
        particles_pos[0, :] = guided_wps_array.flatten()
        # print(f"DEBUG: Guided particle 0 waypoints (flattened): {particles_pos[0, :]}")

        # <<-- Detailed check of the guided particle's path BEFORE PSO starts -->>
        # print("\n--- DEBUGGING GUIDED PARTICLE 0 INITIAL PATH ---")
        # guided_particle_nodes = np.vstack([START_POS, particles_pos[0].reshape(NUM_WAYPOINTS, 2), GOAL_POS])
        # guided_path_hits_critical = False
        # for k_seg in range(len(guided_particle_nodes) - 1):
        #     p1_test = guided_particle_nodes[k_seg]
        #     p2_test = guided_particle_nodes[k_seg+1]
        #     # print(f"  GUIDED Seg {k_seg}: {p1_test.round(1)} -> {p2_test.round(1)}")
        #     if check_segment_collision_with_list(p1_test, p2_test, CRITICAL_INFLATED_OBSTACLES):
        #         # print(f"    WARN: GUIDED segment {k_seg} HITS CRITICAL!")
        #         guided_path_hits_critical = True
        #         # for c_idx, crit_obs_test in enumerate(CRITICAL_INFLATED_OBSTACLES):
        #         #     if check_segment_collision_with_list(p1_test,p2_test,[crit_obs_test]):
        #         #         print(f"        Collided with CRIT_OBS {c_idx}, bounds {crit_obs_test.bounds}")
        # if not guided_path_hits_critical:
        #     print("  GUIDED particle 0 path appears CRITICALLY CLEAR initially.")
        # else:
        #     print("  ERROR: GUIDED particle 0 path HITS CRITICAL initially. Review guided logic or obstacle definitions.")
        # print("--- END DEBUGGING GUIDED PARTICLE ---\n")
    # <<<< END TEMPORARY GUIDED INITIALIZATION >>>>

    particles_vel = np.random.randn(NUM_PARTICLES, DIMENSIONS) * 0.05 # Smaller initial velocities
    pbest_pos = np.copy(particles_pos)
    pbest_fitness = np.full(NUM_PARTICLES, np.inf)
    gbest_pos = np.full(DIMENSIONS, np.nan)
    gbest_fitness = np.inf

    # Initial fitness evaluation for all particles (including guided one)
    for i in range(NUM_PARTICLES):
        fitness_val = fitness_function(particles_pos[i], START_POS, GOAL_POS, 
                                       NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, 
                                       MAP_BOUNDS, particle_idx_debug=i, iteration_debug=-1) # -1 for initial eval
        pbest_fitness[i] = fitness_val
        if fitness_val < gbest_fitness:
            gbest_fitness = fitness_val
            gbest_pos = np.copy(particles_pos[i])
            # if i == 0 and np.isfinite(gbest_fitness):
            #     print(f"DEBUG: Guided particle 0 gave initial gbest_fitness: {gbest_fitness:.2f}")
            # elif i==0 and not np.isfinite(gbest_fitness):
            #     print(f"DEBUG: Guided particle 0 STILL has INF fitness. Problem is severe.")


    fig_pso, ax_pso = plt.subplots(figsize=(10, 7))
    plt.ion()

    print(f"\nStarting PSO. Initial best fitness: {gbest_fitness:.2f}") # This will show 'inf' if guided particle also failed
    plot_environment(0, gbest_pos,
                     START_POS, GOAL_POS, 
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, title_suffix="Initial State")
    
    if not np.isfinite(gbest_fitness):
        print("CRITICAL DEBUG: No particle (even guided) found a non-infinite path initially. Stopping.")
        # plt.ioff(); plt.show() # Show plot and exit if you want to stop here
        # exit()


    for iteration in range(MAX_ITERATIONS):
        current_W = W_MAX - (W_MAX - W_MIN) * (iteration / MAX_ITERATIONS)
        num_finite_particles_this_iter = 0

        for i in range(NUM_PARTICLES):
            r1 = np.random.rand(DIMENSIONS); r2 = np.random.rand(DIMENSIONS)
            gbest_term = np.zeros_like(particles_pos[i])
            if np.isfinite(gbest_fitness):
                 gbest_term = C2 * r2 * (gbest_pos - particles_pos[i])

            particles_vel[i] = (current_W * particles_vel[i] +
                                C1 * r1 * (pbest_pos[i] - particles_pos[i]) +
                                gbest_term)
            
            vel_limit_x = (MAP_BOUNDS['x_max'] - MAP_BOUNDS['x_min']) * 0.035
            vel_limit_y = (MAP_BOUNDS['y_max'] - MAP_BOUNDS['y_min']) * 0.035
            particles_vel[i, ::2] = np.clip(particles_vel[i, ::2], -vel_limit_x, vel_limit_x)
            particles_vel[i, 1::2] = np.clip(particles_vel[i, 1::2], -vel_limit_y, vel_limit_y)
            particles_pos[i] += particles_vel[i]
            
            current_waypoints = particles_pos[i].reshape(NUM_WAYPOINTS, 2)
            current_waypoints[:, 0] = np.clip(current_waypoints[:, 0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            current_waypoints[:, 1] = np.clip(current_waypoints[:, 1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
            particles_pos[i] = current_waypoints.flatten()

            current_fitness = fitness_function(particles_pos[i], START_POS, GOAL_POS, 
                                               NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                               MAP_BOUNDS, particle_idx_debug=i, iteration_debug=iteration)
            
            if np.isfinite(current_fitness): num_finite_particles_this_iter +=1
            if current_fitness < pbest_fitness[i]:
                pbest_fitness[i] = current_fitness
                pbest_pos[i] = np.copy(particles_pos[i])
                if current_fitness < gbest_fitness:
                    gbest_fitness = current_fitness
                    gbest_pos = np.copy(particles_pos[i])

        if (iteration + 1) % 10 == 0 or iteration == MAX_ITERATIONS - 1 or (iteration < 5 and np.isfinite(gbest_fitness)): # Plot more frequently at start if a path is found
            plot_environment(iteration + 1, gbest_pos,
                             START_POS, GOAL_POS, 
                             ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                             MAP_BOUNDS, ax_pso)
            print(f"Iter {iteration+1}/{MAX_ITERATIONS}: W={current_W:.2f}, BestFit={gbest_fitness:.2f}, FiniteInIter={num_finite_particles_this_iter}/{NUM_PARTICLES}")
            
            if np.isfinite(gbest_fitness) and gbest_fitness < (NORMAL_OBSTACLE_PENALTY_CONSTANT / 1000) and num_finite_particles_this_iter > 0 :
                gbest_path_nodes_check = np.vstack([START_POS, gbest_pos.reshape(NUM_WAYPOINTS, 2), GOAL_POS])
                crit_coll_gbest = 0
                for k_seg in range(len(gbest_path_nodes_check) - 1):
                    if check_segment_collision_with_list(gbest_path_nodes_check[k_seg], gbest_path_nodes_check[k_seg+1], CRITICAL_INFLATED_OBSTACLES):
                        crit_coll_gbest +=1; break
                if crit_coll_gbest == 0:
                    # print(f"--- Potentially converged early at iter {iteration+1} ---")
                    pass # break # Uncomment for early stopping

    # Final Path Construction and Output
    if not np.isfinite(gbest_fitness) or not np.isfinite(gbest_pos).all(): # Check gbest_pos too
        print("\n--- PSO Finished: FAILED TO FIND A VALID PATH (gbest is inf or NaN) ---")
        final_path_nodes_map = np.array([START_POS, GOAL_POS])
    else:
        final_waypoints_map = gbest_pos.reshape(NUM_WAYPOINTS, 2)
        final_path_nodes_map = np.vstack([START_POS, final_waypoints_map, GOAL_POS])
        print("\n--- PSO Finished ---")
        print(f"Best Fitness Found: {gbest_fitness:.2f}")
    
    final_path_critical_collisions = 0; final_path_normal_collisions = 0
    for i in range(len(final_path_nodes_map) - 1):
        p1_node = final_path_nodes_map[i]; p2_node = final_path_nodes_map[i+1]
        if check_segment_collision_with_list(p1_node, p2_node, CRITICAL_INFLATED_OBSTACLES): final_path_critical_collisions += 1
        elif check_segment_collision_with_list(p1_node, p2_node, NORMAL_INFLATED_OBSTACLES): final_path_normal_collisions += 1

    if final_path_critical_collisions == 0: print("Final best path is CRITICALLY collision-free.")
    else: print(f"WARNING: Final best path has {final_path_critical_collisions} CRITICAL colliding segments.")
    if final_path_normal_collisions == 0: print("Final best path is also normally collision-free.")
    else: print(f"Final best path has {final_path_normal_collisions} normal colliding segments.")

    print("Optimal Path (Map Coordinates):")
    for node_idx, node in enumerate(final_path_nodes_map): print(f"  Node {node_idx}: ({node[0]:.2f}, {node[1]:.2f})")

    plot_environment(MAX_ITERATIONS, gbest_pos,
                     START_POS, GOAL_POS, 
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, title_suffix=" - Final")
    plt.ioff()
    plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from shapely.geometry import LineString, Polygon, Point

# --- Configuration ---
NUM_PARTICLES = 150
MAX_ITERATIONS = 500 # Increase if paths are complex or not converging
NUM_WAYPOINTS = 15   # More waypoints for complex paths
DIMENSIONS = NUM_WAYPOINTS * 2

# Inertia weight decay parameters
W_MAX = 0.9
W_MIN = 0.4
C1 = 1.8 
C2 = 1.8 

# Fitness function weights
W_LENGTH = 1.0
W_OBSTACLE_MULTIPLIER = 1.0 # For normal obstacles
NORMAL_OBSTACLE_PENALTY_CONSTANT = 20000.0
CRITICAL_OBSTACLE_PENALTY = float('inf') # Effectively infinite penalty
W_BOUNDS = 70.0 

# Environment and Robot Setup
MAP_BOUNDS = {'x_min': 0, 'x_max': 100, 'y_min': 0, 'y_max': 55} 
START_POS = np.array([5, 40])
GOAL_POS = np.array([95, 40])
ROBOT_RADIUS = 0.1 # <<-- ADJUST THIS! Set to 0 for point robot, or actual radius.

# --- Obstacle Simulation, Inflation, and Critical Marking ---
def get_and_process_obstacles(robot_radius):
    print("Processing simulated obstacles...")
    # Define obstacles: coords are [(x1,y1), (x2,y2), ...], is_critical, color for plotting
    simulated_obstacles_data = [
        {'coords': [(15, 29), (15, 51), (30, 51), (30, 29)], 'is_critical': False, 'color': 'rosybrown'},
        {'coords': [(48, 9), (48, 37), (60, 37), (60, 9)], 'is_critical': True, 'color': 'red'}
        # Add more obstacles here if needed, e.g.:
        # {'coords': [(70, 45), (70, 50), (90, 50), (90, 45)], 'is_critical': False, 'color':'lightslategray'}
    ]
    
    all_inflated_obstacles = []
    critical_inflated_obstacles = []
    normal_inflated_obstacles = []
    obstacle_colors = [] # To store colors corresponding to all_inflated_obstacles

    for obs_data in simulated_obstacles_data:
        coords = obs_data['coords']
        is_critical = obs_data['is_critical']
        color = obs_data['color']
        
        original_poly = Polygon(coords)
        if not original_poly.is_valid:
            # print(f"DEBUG: Original simulated obstacle with coords {coords} is invalid.")
            continue # Skip invalid original polygons
        
        # Optional: Check if original polygon is roughly within map bounds to avoid unnecessary processing
        # This check is basic and might not be perfect for polygons near the edge.
        # if not (original_poly.bounds[2] > MAP_BOUNDS['x_min'] and original_poly.bounds[0] < MAP_BOUNDS['x_max'] and \
        #         original_poly.bounds[3] > MAP_BOUNDS['y_min'] and original_poly.bounds[1] < MAP_BOUNDS['y_max']):
        #     # print(f"DEBUG: Simulated obstacle {coords} is largely outside map bounds, skipping.")
        #     continue

        inflated_poly = original_poly
        if robot_radius > 0:
            # Ensure the original polygon is valid before buffering
            if original_poly.is_valid:
                 inflated_poly = original_poly.buffer(robot_radius, resolution=8)
                 if not inflated_poly.is_valid:
                    # print(f"DEBUG: Inflated obstacle (from {coords}) became invalid.")
                    continue # Skip if inflation results in an invalid polygon
            else: # Should have been caught by the earlier check
                # print(f"DEBUG: Cannot inflate invalid original polygon: {coords}")
                continue
        
        all_inflated_obstacles.append(inflated_poly)
        obstacle_colors.append(color) # Store color for the corresponding inflated_poly

        if is_critical:
            critical_inflated_obstacles.append(inflated_poly)
        else:
            normal_inflated_obstacles.append(inflated_poly)

    print(f"Generated {len(all_inflated_obstacles)} total inflated obstacles.")
    print(f"  {len(critical_inflated_obstacles)} critical inflated obstacles.")
    print(f"  {len(normal_inflated_obstacles)} normal inflated obstacles.")
    return all_inflated_obstacles, critical_inflated_obstacles, normal_inflated_obstacles, obstacle_colors

# --- PSO Helper Functions ---
def check_segment_collision_with_list(p1_coords, p2_coords, obstacles_to_check):
    """Checks if a line segment p1-p2 collides with any obstacle in the given list."""
    if np.array_equal(p1_coords, p2_coords): return False # Zero-length segment
    try:
        segment = LineString([p1_coords, p2_coords])
        if not segment.is_valid: return True # Cannot form valid segment
    except Exception:
        return True # Error forming segment, assume collision for safety
    for obs_poly in obstacles_to_check:
        if segment.intersects(obs_poly):
            return True 
    return False

def fitness_function(particle_waypoints_flat, start_pos_map, goal_pos_map, 
                     normal_obs_list, critical_obs_list, map_bnds_map,
                     particle_idx_debug=-1, iteration_debug=-1): # Debug params
    waypoints = particle_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    full_path_nodes = np.vstack([start_pos_map, waypoints, goal_pos_map])
    
    path_length = 0.0
    num_normal_collisions = 0
    num_critical_collisions = 0
    num_out_of_bounds = 0

    # Enable detailed debugging for specific particles/iterations if needed
    # detailed_debug_active = (particle_idx_debug == 0 and (iteration_debug == -1 or iteration_debug % 50 == 0) )
    # if detailed_debug_active: print(f"\nFIT_DBG P{particle_idx_debug} Iter{iteration_debug}: Path {particle_waypoints_flat[:4].round(1)}...")

    for i in range(len(full_path_nodes) - 1):
        p1 = full_path_nodes[i]
        p2 = full_path_nodes[i+1]
        path_length += np.linalg.norm(p2 - p1)

        segment_hit_critical_this_pass = False
        if check_segment_collision_with_list(p1, p2, critical_obs_list):
            num_critical_collisions += 1
            segment_hit_critical_this_pass = True
            # if detailed_debug_active: print(f"  FIT_DBG P{particle_idx_debug} Seg{i}: {p1.round(1)}->{p2.round(1)} HIT CRITICAL.")
        
        if not segment_hit_critical_this_pass: 
            if check_segment_collision_with_list(p1, p2, normal_obs_list):
                num_normal_collisions += 1
                # if detailed_debug_active: print(f"  FIT_DBG P{particle_idx_debug} Seg{i}: {p1.round(1)}->{p2.round(1)} HIT NORMAL.")
            
    for wp_idx in range(NUM_WAYPOINTS):
        wp_x, wp_y = particle_waypoints_flat[wp_idx*2], particle_waypoints_flat[wp_idx*2+1]
        if not (map_bnds_map['x_min'] <= wp_x <= map_bnds_map['x_max'] and \
                map_bnds_map['y_min'] <= wp_y <= map_bnds_map['y_max']):
            num_out_of_bounds += 1
    
    if num_critical_collisions > 0:
        # if detailed_debug_active: print(f"  FIT_DBG P{particle_idx_debug}: RETURNING INF ({num_critical_collisions} critical hits).")
        return CRITICAL_OBSTACLE_PENALTY 
    
    fitness_val = (W_LENGTH * path_length +
                   W_OBSTACLE_MULTIPLIER * num_normal_collisions * NORMAL_OBSTACLE_PENALTY_CONSTANT +
                   W_BOUNDS * num_out_of_bounds)
    
    # if detailed_debug_active: print(f"  FIT_DBG P{particle_idx_debug}: Total={fitness_val:.2f} (L:{path_length:.2f}, NormC:{num_normal_collisions}, CritC:{num_critical_collisions}, OB:{num_out_of_bounds})")
    return fitness_val

def plot_environment(iteration, gbest_pos_flat,
                     start_map, goal_map, 
                     all_obstacles_infl, critical_obstacles_infl, obstacle_cols, # Pass colors
                     map_bnds_map, ax_plot, current_gbest_fitness, title_suffix=""):
    ax_plot.clear()
    for i, obs_poly in enumerate(all_obstacles_infl): # Iterate through all obstacles to use stored colors
        color_to_use = obstacle_cols[i] 
        alpha_val = 0.6 # Default alpha
        is_this_critical = any(obs_poly.equals(crit_obs) for crit_obs in critical_obstacles_infl)
        if is_this_critical:
            alpha_val = 0.85 # Make critical obstacles more solid
            # color_to_use = 'red' # Optionally override color for critical ones if needed

        if obs_poly.exterior and len(obs_poly.exterior.coords) > 0 :
            ax_plot.add_patch(patches.Polygon(obs_poly.exterior.coords, closed=True, color=color_to_use, alpha=alpha_val, zorder=2))
    
    ax_plot.plot(start_map[0], start_map[1], 'go', markersize=10, label='Start', zorder=5)
    ax_plot.plot(goal_map[0], goal_map[1], 'mo', markersize=10, label='Goal', zorder=5)

    if gbest_pos_flat is not None and len(gbest_pos_flat) > 0 and \
       np.isfinite(gbest_pos_flat).all() and np.isfinite(current_gbest_fitness):
        gbest_waypoints = gbest_pos_flat.reshape(NUM_WAYPOINTS, 2)
        gbest_path_nodes_plot = np.vstack([start_map, gbest_waypoints, goal_map])
        ax_plot.plot(gbest_path_nodes_plot[:, 0], gbest_path_nodes_plot[:, 1], 'b-', linewidth=2.5, label='Best Path', zorder=4)
    elif iteration >= 0 and not np.isfinite(current_gbest_fitness) : # iteration >= 0 to avoid plotting this at init if gbest is inf
        ax_plot.text(np.mean(ax_plot.get_xlim()), np.mean(ax_plot.get_ylim()),
                      "No valid path found yet", ha='center', va='center', fontsize=12, color='gray', zorder=3)

    ax_plot.set_xlim(map_bnds_map['x_min'], map_bnds_map['x_max'])
    ax_plot.set_ylim(map_bnds_map['y_min'], map_bnds_map['y_max'])
    ax_plot.set_title(f"PSO Path Planning - Iter: {iteration} {title_suffix}")
    ax_plot.set_xlabel("X (Map)"); ax_plot.set_ylabel("Y (Map)")
    handles, labels = ax_plot.get_legend_handles_labels()
    if handles: ax_plot.legend(loc='best')
    ax_plot.set_aspect('equal', adjustable='box'); plt.pause(0.01)

# --- Main Execution ---
if __name__ == "__main__":
    ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, NORMAL_INFLATED_OBSTACLES, OBSTACLE_COLORS = \
        get_and_process_obstacles(ROBOT_RADIUS)

    if not ALL_INFLATED_OBSTACLES: print("Warning: No obstacles available for path planning.")
    if not CRITICAL_INFLATED_OBSTACLES: print("Info: No obstacles were marked as critical.")

    # Initialize particles' positions
    particles_pos = np.zeros((NUM_PARTICLES, DIMENSIONS))
    for i in range(NUM_PARTICLES):
        for d_wp in range(NUM_WAYPOINTS):
            particles_pos[i, d_wp*2] = random.uniform(MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            particles_pos[i, d_wp*2+1] = random.uniform(MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
    
    # << Optional: Guided Initialization for particle 0 - REMOVE OR COMMENT OUT IF NOT DEBUGGING "inf" issue >>
    if NUM_PARTICLES > 0 and len(CRITICAL_INFLATED_OBSTACLES) > 0:
        # print("DEBUG: Applying Guided Initialization for Particle 0")
        guided_wps_array = np.zeros((NUM_WAYPOINTS, 2))
        # Define some points to guide the path (very rough, adjust based on your specific obstacle layout)
        # This example tries to go above the first critical obstacle
        crit_obs_for_guidance = CRITICAL_INFLATED_OBSTACLES[0] # Assuming at least one critical
        pt1_x = START_POS[0] + (crit_obs_for_guidance.bounds[0] - START_POS[0]) * 0.5 # Mid X before critical
        pt1_y = crit_obs_for_guidance.bounds[3] + 5 + ROBOT_RADIUS # Y above critical
        pt2_x = crit_obs_for_guidance.bounds[2] + (GOAL_POS[0] - crit_obs_for_guidance.bounds[2]) * 0.2 # X after critical, towards goal
        pt2_y = pt1_y # Keep Y high over critical
        # Distribute waypoints
        num_wp_seg1, num_wp_seg2 = NUM_WAYPOINTS // 3, NUM_WAYPOINTS // 3
        num_wp_seg3 = NUM_WAYPOINTS - num_wp_seg1 - num_wp_seg2
        current_wp_idx = 0
        for k in range(num_wp_seg1): t = (k + 1) / num_wp_seg1; guided_wps_array[current_wp_idx, 0] = START_POS[0] + t * (pt1_x - START_POS[0]); guided_wps_array[current_wp_idx, 1] = START_POS[1] + t * (pt1_y - START_POS[1]); current_wp_idx += 1
        for k in range(num_wp_seg2): t = (k + 1) / num_wp_seg2; guided_wps_array[current_wp_idx, 0] = pt1_x + t * (pt2_x - pt1_x);     guided_wps_array[current_wp_idx, 1] = pt1_y + t * (pt2_y - pt1_y);     current_wp_idx += 1
        for k in range(num_wp_seg3): t = (k + 1) / num_wp_seg3; guided_wps_array[current_wp_idx, 0] = pt2_x + t * (GOAL_POS[0] - pt2_x); guided_wps_array[current_wp_idx, 1] = pt2_y + t * (GOAL_POS[1] - pt2_y); current_wp_idx += 1
        particles_pos[0, :] = guided_wps_array.flatten()
        
        # # --- Detailed check of the guided particle's path BEFORE PSO starts ---
        # print("\n--- DEBUGGING GUIDED PARTICLE 0 INITIAL PATH ---")
        # guided_particle_nodes = np.vstack([START_POS, particles_pos[0].reshape(NUM_WAYPOINTS, 2), GOAL_POS])
        # guided_path_hits_critical = False
        # for k_seg in range(len(guided_particle_nodes) - 1):
        #     p1_test, p2_test = guided_particle_nodes[k_seg], guided_particle_nodes[k_seg+1]
        #     # print(f"  GUIDED Seg {k_seg}: {p1_test.round(1)} -> {p2_test.round(1)}")
        #     if check_segment_collision_with_list(p1_test, p2_test, CRITICAL_INFLATED_OBSTACLES):
        #         # print(f"    WARN: GUIDED segment {k_seg} HITS CRITICAL!")
        #         guided_path_hits_critical = True
        # if not guided_path_hits_critical: print("  GUIDED particle 0 path appears CRITICALLY CLEAR initially.")
        # else: print("  ERROR: GUIDED particle 0 path HITS CRITICAL initially. Review guided logic/obstacles.")
        # print("--- END DEBUGGING GUIDED PARTICLE ---\n")
    # << End Guided Init >>

    particles_vel = np.random.randn(NUM_PARTICLES, DIMENSIONS) * 0.05 
    pbest_pos = np.copy(particles_pos)
    pbest_fitness = np.full(NUM_PARTICLES, np.inf)
    gbest_pos = np.full(DIMENSIONS, np.nan) 
    gbest_fitness = np.inf

    for i in range(NUM_PARTICLES): # Initial pbest/gbest evaluation
        fitness_val = fitness_function(particles_pos[i], START_POS, GOAL_POS, 
                                       NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, 
                                       MAP_BOUNDS, particle_idx_debug=i, iteration_debug=-1) # Iter -1 for initial
        pbest_fitness[i] = fitness_val
        if fitness_val < gbest_fitness:
            gbest_fitness = fitness_val
            gbest_pos = np.copy(particles_pos[i])

    fig_pso, ax_pso = plt.subplots(figsize=(10, 7))
    plt.ion()
    print(f"\nStarting PSO. Initial best fitness: {gbest_fitness:.2f}")
    plot_environment(0, gbest_pos, START_POS, GOAL_POS, 
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, gbest_fitness, title_suffix="Initial State")
    if not np.isfinite(gbest_fitness):
        print("CRITICAL INFO: No particle (even potentially guided) found a non-infinite path initially. PSO will likely fail or struggle significantly.")

    # PSO Main Loop
    for iteration in range(MAX_ITERATIONS):
        current_W = W_MAX - (W_MAX - W_MIN) * (iteration / MAX_ITERATIONS)
        num_finite_particles_this_iter = 0
        for i in range(NUM_PARTICLES):
            r1, r2 = np.random.rand(DIMENSIONS), np.random.rand(DIMENSIONS)
            gbest_term = np.zeros_like(particles_pos[i])
            if np.isfinite(gbest_fitness) and np.isfinite(gbest_pos).all():
                 gbest_term = C2 * r2 * (gbest_pos - particles_pos[i])
            
            # Ensure pbest_pos[i] is also finite for subtraction
            current_pbest_pos = pbest_pos[i]
            if not np.isfinite(current_pbest_pos).all(): # If pbest was from an 'inf' state
                current_pbest_pos = particles_pos[i] # Fallback to current position

            particles_vel[i] = (current_W * particles_vel[i] + 
                                C1 * r1 * (current_pbest_pos - particles_pos[i]) + 
                                gbest_term)
            
            vel_limit_x = (MAP_BOUNDS['x_max'] - MAP_BOUNDS['x_min']) * 0.04 # Slightly increased limit
            vel_limit_y = (MAP_BOUNDS['y_max'] - MAP_BOUNDS['y_min']) * 0.04
            particles_vel[i, ::2] = np.clip(particles_vel[i, ::2], -vel_limit_x, vel_limit_x)
            particles_vel[i, 1::2] = np.clip(particles_vel[i, 1::2], -vel_limit_y, vel_limit_y)
            particles_pos[i] += particles_vel[i]
            
            # Position Clamping
            current_waypoints = particles_pos[i].reshape(NUM_WAYPOINTS, 2)
            current_waypoints[:, 0] = np.clip(current_waypoints[:, 0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            current_waypoints[:, 1] = np.clip(current_waypoints[:, 1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
            particles_pos[i] = current_waypoints.flatten()

            current_fitness = fitness_function(particles_pos[i], START_POS, GOAL_POS, 
                                               NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                               MAP_BOUNDS, particle_idx_debug=i, iteration_debug=iteration)
            if np.isfinite(current_fitness): num_finite_particles_this_iter +=1
            if current_fitness < pbest_fitness[i]:
                pbest_fitness[i] = current_fitness; pbest_pos[i] = np.copy(particles_pos[i])
                if current_fitness < gbest_fitness:
                    gbest_fitness = current_fitness; gbest_pos = np.copy(particles_pos[i])
        
        if (iteration + 1) % 10 == 0 or iteration == MAX_ITERATIONS - 1 or \
           (iteration < 10 and np.isfinite(gbest_fitness)): # Plot more frequently at start if solution found
            plot_environment(iteration + 1, gbest_pos, START_POS, GOAL_POS, 
                             ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                             MAP_BOUNDS, ax_pso, gbest_fitness)
            print(f"Iter {iteration+1}/{MAX_ITERATIONS}: W={current_W:.2f}, BestFit={gbest_fitness:.2f}, FiniteInIter={num_finite_particles_this_iter}/{NUM_PARTICLES}")
            
            # Optional: Early stopping if a good enough, collision-free path is found
            if np.isfinite(gbest_fitness) and gbest_fitness < (NORMAL_OBSTACLE_PENALTY_CONSTANT / 1000) and num_finite_particles_this_iter > 0 :
                gbest_path_nodes_check = np.vstack([START_POS, gbest_pos.reshape(NUM_WAYPOINTS, 2), GOAL_POS])
                if not any(check_segment_collision_with_list(gbest_path_nodes_check[k], gbest_path_nodes_check[k+1], CRITICAL_INFLATED_OBSTACLES) for k in range(len(gbest_path_nodes_check)-1)):
                    # print(f"--- Potentially converged early at iter {iteration+1} with a good, critically-clear path. ---")
                    # break # Uncomment for early stopping
                    pass


    # --- Final Path Construction and Output ---
    print("\n--- PSO Finished ---")
    final_path_nodes_map = np.array([START_POS, GOAL_POS]) # Default if no valid path
    
    if np.isfinite(gbest_fitness) and gbest_pos is not None and np.isfinite(gbest_pos).all():
        print(f"Best Fitness Found: {gbest_fitness:.2f}")
        final_waypoints_map = gbest_pos.reshape(NUM_WAYPOINTS, 2)
        final_path_nodes_map = np.vstack([START_POS, final_waypoints_map, GOAL_POS])
    else:
        print("FAILED TO FIND A VALID NON-INFINITE PATH. Displaying direct Start-Goal line.")
    
    final_path_critical_collisions, final_path_normal_collisions = 0,0
    if len(final_path_nodes_map) > 1:
        for i in range(len(final_path_nodes_map) - 1):
            p1_node, p2_node = final_path_nodes_map[i], final_path_nodes_map[i+1]
            if check_segment_collision_with_list(p1_node, p2_node, CRITICAL_INFLATED_OBSTACLES): final_path_critical_collisions += 1
            elif check_segment_collision_with_list(p1_node, p2_node, NORMAL_INFLATED_OBSTACLES): final_path_normal_collisions += 1

    print(f"Displayed path is CRITICALLY collision-free: {final_path_critical_collisions == 0}.")
    if final_path_critical_collisions > 0: print(f"  WARNING: {final_path_critical_collisions} CRITICAL colliding segments in displayed path.")
    print(f"Displayed path is normally collision-free: {final_path_normal_collisions == 0}.")
    if final_path_normal_collisions > 0: print(f"  Displayed path has {final_path_normal_collisions} normal colliding segments.")

    print("Path to be Displayed (Map Coordinates):")
    for node_idx, node in enumerate(final_path_nodes_map): print(f"  Node {node_idx}: ({node[0]:.2f}, {node[1]:.2f})")

    plot_environment(MAX_ITERATIONS, gbest_pos, START_POS, GOAL_POS, 
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, gbest_fitness, title_suffix=" - Final Result")
    plt.ioff(); plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from shapely.geometry import LineString, Polygon, Point

# --- Configuration ---
NUM_PARTICLES = 150
MAX_ITERATIONS = 700 # Increased for more complex environment
NUM_WAYPOINTS = 20   # Increased for more complex paths with more obstacles
DIMENSIONS = NUM_WAYPOINTS * 2

# Inertia weight decay parameters
W_MAX = 0.9
W_MIN = 0.4
C1 = 1.8
C2 = 1.8

# Fitness function weights
W_LENGTH = 1.0
W_OBSTACLE_MULTIPLIER = 1.0 # For normal obstacles
NORMAL_OBSTACLE_PENALTY_CONSTANT = 20000.0
CRITICAL_OBSTACLE_PENALTY = float('inf') # Effectively infinite penalty
W_BOUNDS = 70.0

# Environment and Robot Setup
MAP_BOUNDS = {'x_min': 0, 'x_max': 100, 'y_min': 0, 'y_max': 55}
START_POS = np.array([5, 40])
GOAL_POS = np.array([95, 40])
ROBOT_RADIUS = 0.1 # <<-- ADJUST THIS! Set to 0 for point robot, or actual radius.

# --- Obstacle Simulation, Inflation, and Critical Marking ---
def get_and_process_obstacles(robot_radius):
    print("Processing simulated obstacles...")
    # Define obstacles: coords are [(x1,y1), (x2,y2), ...], is_critical, color for plotting
    simulated_obstacles_data = [
        # Original Obstacles (modified slightly or kept)
        {'coords': [(15, 29), (15, 51), (30, 51), (30, 29)], 'is_critical': False, 'color': 'rosybrown'}, # Obstacle 1 (original 1)
        {'coords': [(48, 9), (48, 37), (60, 37), (60, 9)], 'is_critical': True, 'color': 'red'},          # Obstacle 2 (original 2, critical)

        # Added Obstacles for a total of 10
        {'coords': [(5, 48), (5, 54), (20, 54), (20, 48)], 'is_critical': False, 'color': 'lightgreen'},     # Obstacle 3 (Top-leftish)
        {'coords': [(10, 5), (10, 20), (25, 20), (25, 5)], 'is_critical': False, 'color': 'lightblue'},      # Obstacle 4 (Bottom-left)
        {'coords': [(35, 45), (35, 52), (65, 52), (65, 45)], 'is_critical': True, 'color': 'darkred'},      # Obstacle 5 (Mid-top, wide, critical)
        {'coords': [(40, 2), (40, 8), (55, 8), (55, 2)], 'is_critical': False, 'color': 'gold'},           # Obstacle 6 (Mid-bottom, small)
        {'coords': [(70, 25), (70, 50), (80, 50), (80, 25)], 'is_critical': False, 'color': 'lightcoral'},   # Obstacle 7 (Right-mid, tall)
        {'coords': [(85, 48), (85, 54), (98, 54), (98, 48)], 'is_critical': True, 'color': 'orangered'},    # Obstacle 8 (Top-right near goal, critical)
        {'coords': [(75, 5), (75, 18), (90, 18), (90, 5)], 'is_critical': False, 'color': 'mediumpurple'}, # Obstacle 9 (Bottom-right)
        {'coords': [(35, 15), (35, 25), (70, 25), (70, 15)], 'is_critical': False, 'color': 'darkseagreen'},# Obstacle 10 (Center-ish, horizontal block)
    ]

    all_inflated_obstacles = []
    critical_inflated_obstacles = []
    normal_inflated_obstacles = []
    obstacle_colors = [] # To store colors corresponding to all_inflated_obstacles

    for obs_data in simulated_obstacles_data:
        coords = obs_data['coords']
        is_critical = obs_data['is_critical']
        color = obs_data['color']

        original_poly = Polygon(coords)
        if not original_poly.is_valid:
            # print(f"DEBUG: Original simulated obstacle with coords {coords} is invalid.")
            continue # Skip invalid original polygons

        inflated_poly = original_poly
        if robot_radius > 0:
            if original_poly.is_valid:
                 inflated_poly = original_poly.buffer(robot_radius, resolution=8)
                 if not inflated_poly.is_valid:
                    # print(f"DEBUG: Inflated obstacle (from {coords}) became invalid.")
                    continue
            else:
                # print(f"DEBUG: Cannot inflate invalid original polygon: {coords}")
                continue

        all_inflated_obstacles.append(inflated_poly)
        obstacle_colors.append(color)

        if is_critical:
            critical_inflated_obstacles.append(inflated_poly)
        else:
            normal_inflated_obstacles.append(inflated_poly)

    print(f"Generated {len(all_inflated_obstacles)} total inflated obstacles.")
    print(f"  {len(critical_inflated_obstacles)} critical inflated obstacles.")
    print(f"  {len(normal_inflated_obstacles)} normal inflated obstacles.")
    return all_inflated_obstacles, critical_inflated_obstacles, normal_inflated_obstacles, obstacle_colors

# --- PSO Helper Functions ---
def check_segment_collision_with_list(p1_coords, p2_coords, obstacles_to_check):
    """Checks if a line segment p1-p2 collides with any obstacle in the given list."""
    if np.array_equal(p1_coords, p2_coords): return False # Zero-length segment
    try:
        segment = LineString([p1_coords, p2_coords])
        if not segment.is_valid: return True # Cannot form valid segment
    except Exception:
        return True # Error forming segment, assume collision for safety
    for obs_poly in obstacles_to_check:
        if segment.intersects(obs_poly):
            return True
    return False

def fitness_function(particle_waypoints_flat, start_pos_map, goal_pos_map,
                     normal_obs_list, critical_obs_list, map_bnds_map,
                     particle_idx_debug=-1, iteration_debug=-1): # Debug params
    waypoints = particle_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    full_path_nodes = np.vstack([start_pos_map, waypoints, goal_pos_map])

    path_length = 0.0
    num_normal_collisions = 0
    num_critical_collisions = 0
    num_out_of_bounds = 0

    for i in range(len(full_path_nodes) - 1):
        p1 = full_path_nodes[i]
        p2 = full_path_nodes[i+1]
        path_length += np.linalg.norm(p2 - p1)

        segment_hit_critical_this_pass = False
        if check_segment_collision_with_list(p1, p2, critical_obs_list):
            num_critical_collisions += 1
            segment_hit_critical_this_pass = True

        if not segment_hit_critical_this_pass:
            if check_segment_collision_with_list(p1, p2, normal_obs_list):
                num_normal_collisions += 1

    for wp_idx in range(NUM_WAYPOINTS):
        wp_x, wp_y = particle_waypoints_flat[wp_idx*2], particle_waypoints_flat[wp_idx*2+1]
        if not (map_bnds_map['x_min'] <= wp_x <= map_bnds_map['x_max'] and \
                map_bnds_map['y_min'] <= wp_y <= map_bnds_map['y_max']):
            num_out_of_bounds += 1

    if num_critical_collisions > 0:
        return CRITICAL_OBSTACLE_PENALTY

    fitness_val = (W_LENGTH * path_length +
                   W_OBSTACLE_MULTIPLIER * num_normal_collisions * NORMAL_OBSTACLE_PENALTY_CONSTANT +
                   W_BOUNDS * num_out_of_bounds)
    return fitness_val

def plot_environment(iteration, gbest_pos_flat,
                     start_map, goal_map,
                     all_obstacles_infl, critical_obstacles_infl, obstacle_cols, # Pass colors
                     map_bnds_map, ax_plot, current_gbest_fitness, title_suffix=""):
    ax_plot.clear()
    for i, obs_poly in enumerate(all_obstacles_infl): # Iterate through all obstacles to use stored colors
        color_to_use = obstacle_cols[i]
        alpha_val = 0.6 # Default alpha
        is_this_critical = any(obs_poly.equals(crit_obs) for crit_obs in critical_obstacles_infl)
        if is_this_critical:
            alpha_val = 0.85 # Make critical obstacles more solid

        if obs_poly.exterior and len(obs_poly.exterior.coords) > 0 :
            ax_plot.add_patch(patches.Polygon(obs_poly.exterior.coords, closed=True, color=color_to_use, alpha=alpha_val, zorder=2))

    ax_plot.plot(start_map[0], start_map[1], 'go', markersize=10, label='Start', zorder=5)
    ax_plot.plot(goal_map[0], goal_map[1], 'mo', markersize=10, label='Goal', zorder=5)

    if gbest_pos_flat is not None and len(gbest_pos_flat) > 0 and \
       np.isfinite(gbest_pos_flat).all() and np.isfinite(current_gbest_fitness):
        gbest_waypoints = gbest_pos_flat.reshape(NUM_WAYPOINTS, 2)
        gbest_path_nodes_plot = np.vstack([start_map, gbest_waypoints, goal_map])
        ax_plot.plot(gbest_path_nodes_plot[:, 0], gbest_path_nodes_plot[:, 1], 'b-', linewidth=2.5, label='Best Path', zorder=4)
    elif iteration >= 0 and not np.isfinite(current_gbest_fitness) : # iteration >= 0 to avoid plotting this at init if gbest is inf
        ax_plot.text(np.mean(ax_plot.get_xlim()), np.mean(ax_plot.get_ylim()),
                      "No valid path found yet", ha='center', va='center', fontsize=12, color='gray', zorder=3)

    ax_plot.set_xlim(map_bnds_map['x_min'], map_bnds_map['x_max'])
    ax_plot.set_ylim(map_bnds_map['y_min'], map_bnds_map['y_max'])
    ax_plot.set_title(f"PSO Path Planning - Iter: {iteration} {title_suffix}")
    ax_plot.set_xlabel("X (Map)"); ax_plot.set_ylabel("Y (Map)")
    handles, labels = ax_plot.get_legend_handles_labels()
    if handles: ax_plot.legend(loc='best')
    ax_plot.set_aspect('equal', adjustable='box'); plt.pause(0.01)

# --- Main Execution ---
if __name__ == "__main__":
    ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, NORMAL_INFLATED_OBSTACLES, OBSTACLE_COLORS = \
        get_and_process_obstacles(ROBOT_RADIUS)

    if not ALL_INFLATED_OBSTACLES: print("Warning: No obstacles available for path planning.")
    if not CRITICAL_INFLATED_OBSTACLES: print("Info: No obstacles were marked as critical.")

    # Initialize particles' positions
    particles_pos = np.zeros((NUM_PARTICLES, DIMENSIONS))
    for i in range(NUM_PARTICLES):
        for d_wp in range(NUM_WAYPOINTS):
            particles_pos[i, d_wp*2] = random.uniform(MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            particles_pos[i, d_wp*2+1] = random.uniform(MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])

    # << Optional: Guided Initialization for particle 0 - Comment out if not needed >>
    # This guided initialization is very basic and might not be ideal for the new complex obstacle set.
    # It's mainly for ensuring at least one particle starts somewhat reasonably if all random ones fail.
    if NUM_PARTICLES > 0 and len(CRITICAL_INFLATED_OBSTACLES) > 0:
        # print("DEBUG: Applying Guided Initialization for Particle 0")
        guided_wps_array = np.zeros((NUM_WAYPOINTS, 2))
        # Define some points to guide the path (very rough, adjust based on your specific obstacle layout)
        # This example tries to go around the first critical obstacle encountered.
        crit_obs_for_guidance = CRITICAL_INFLATED_OBSTACLES[0] # First critical obstacle in the list
        
        # Attempt a path: slightly above start, then maneuver around critical, then towards goal
        # These points are highly dependent on the specific critical obstacle and map layout
        pt1_x = START_POS[0] + (crit_obs_for_guidance.bounds[0] - START_POS[0]) * 0.3 
        pt1_y = crit_obs_for_guidance.bounds[3] + 3 + ROBOT_RADIUS # Try to go above it
        if crit_obs_for_guidance.bounds[1] > (MAP_BOUNDS['y_min'] + MAP_BOUNDS['y_max'])/2 : # if critical is high
             pt1_y = crit_obs_for_guidance.bounds[1] - 3 - ROBOT_RADIUS # Try to go below it

        pt2_x = crit_obs_for_guidance.bounds[2] + (GOAL_POS[0] - crit_obs_for_guidance.bounds[2]) * 0.3
        pt2_y = pt1_y # Maintain y level

        # A simple three-segment guide
        num_wp_seg1, num_wp_seg2 = NUM_WAYPOINTS // 3, NUM_WAYPOINTS // 3
        num_wp_seg3 = NUM_WAYPOINTS - num_wp_seg1 - num_wp_seg2
        current_wp_idx = 0
        for k in range(num_wp_seg1): t = (k + 1) / (num_wp_seg1 +1e-9); guided_wps_array[current_wp_idx, 0] = START_POS[0] + t * (pt1_x - START_POS[0]); guided_wps_array[current_wp_idx, 1] = START_POS[1] + t * (pt1_y - START_POS[1]); current_wp_idx += 1
        for k in range(num_wp_seg2): t = (k + 1) / (num_wp_seg2 +1e-9); guided_wps_array[current_wp_idx, 0] = pt1_x + t * (pt2_x - pt1_x);     guided_wps_array[current_wp_idx, 1] = pt1_y + t * (pt2_y - pt1_y);     current_wp_idx += 1
        for k in range(num_wp_seg3): t = (k + 1) / (num_wp_seg3 +1e-9); guided_wps_array[current_wp_idx, 0] = pt2_x + t * (GOAL_POS[0] - pt2_x); guided_wps_array[current_wp_idx, 1] = pt2_y + t * (GOAL_POS[1] - pt2_y); current_wp_idx += 1
        
        # Clip guided waypoints to map bounds just in case
        guided_wps_array[:,0] = np.clip(guided_wps_array[:,0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
        guided_wps_array[:,1] = np.clip(guided_wps_array[:,1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
        particles_pos[0, :] = guided_wps_array.flatten()
    # << End Guided Init >>

    particles_vel = np.random.randn(NUM_PARTICLES, DIMENSIONS) * 0.05
    pbest_pos = np.copy(particles_pos)
    pbest_fitness = np.full(NUM_PARTICLES, np.inf)
    gbest_pos = np.full(DIMENSIONS, np.nan)
    gbest_fitness = np.inf

    for i in range(NUM_PARTICLES): # Initial pbest/gbest evaluation
        fitness_val = fitness_function(particles_pos[i], START_POS, GOAL_POS,
                                       NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                       MAP_BOUNDS, particle_idx_debug=i, iteration_debug=-1) # Iter -1 for initial
        pbest_fitness[i] = fitness_val
        if fitness_val < gbest_fitness:
            gbest_fitness = fitness_val
            gbest_pos = np.copy(particles_pos[i])

    fig_pso, ax_pso = plt.subplots(figsize=(12, 8)) # Slightly larger figure for more obstacles
    plt.ion()
    print(f"\nStarting PSO. Initial best fitness: {gbest_fitness:.2f}")
    plot_environment(0, gbest_pos, START_POS, GOAL_POS,
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, gbest_fitness, title_suffix="Initial State")
    if not np.isfinite(gbest_fitness):
        print("CRITICAL INFO: No particle (even potentially guided) found a non-infinite path initially. PSO will likely fail or struggle significantly.")

    # PSO Main Loop
    for iteration in range(MAX_ITERATIONS):
        current_W = W_MAX - (W_MAX - W_MIN) * (iteration / MAX_ITERATIONS)
        num_finite_particles_this_iter = 0
        for i in range(NUM_PARTICLES):
            r1, r2 = np.random.rand(DIMENSIONS), np.random.rand(DIMENSIONS)
            gbest_term = np.zeros_like(particles_pos[i])
            if np.isfinite(gbest_fitness) and np.isfinite(gbest_pos).all():
                 gbest_term = C2 * r2 * (gbest_pos - particles_pos[i])

            current_pbest_pos = pbest_pos[i]
            if not np.isfinite(current_pbest_pos).all():
                current_pbest_pos = particles_pos[i]

            particles_vel[i] = (current_W * particles_vel[i] +
                                C1 * r1 * (current_pbest_pos - particles_pos[i]) +
                                gbest_term)

            vel_limit_x = (MAP_BOUNDS['x_max'] - MAP_BOUNDS['x_min']) * 0.04
            vel_limit_y = (MAP_BOUNDS['y_max'] - MAP_BOUNDS['y_min']) * 0.04
            particles_vel[i, ::2] = np.clip(particles_vel[i, ::2], -vel_limit_x, vel_limit_x)
            particles_vel[i, 1::2] = np.clip(particles_vel[i, 1::2], -vel_limit_y, vel_limit_y)
            particles_pos[i] += particles_vel[i]

            current_waypoints = particles_pos[i].reshape(NUM_WAYPOINTS, 2)
            current_waypoints[:, 0] = np.clip(current_waypoints[:, 0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            current_waypoints[:, 1] = np.clip(current_waypoints[:, 1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
            particles_pos[i] = current_waypoints.flatten()

            current_fitness = fitness_function(particles_pos[i], START_POS, GOAL_POS,
                                               NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                               MAP_BOUNDS, particle_idx_debug=i, iteration_debug=iteration)
            if np.isfinite(current_fitness): num_finite_particles_this_iter +=1
            if current_fitness < pbest_fitness[i]:
                pbest_fitness[i] = current_fitness; pbest_pos[i] = np.copy(particles_pos[i])
                if current_fitness < gbest_fitness:
                    gbest_fitness = current_fitness; gbest_pos = np.copy(particles_pos[i])

        if (iteration + 1) % 20 == 0 or iteration == MAX_ITERATIONS - 1 or \
           (iteration < 20 and np.isfinite(gbest_fitness)): # Plot more frequently at start if solution found
            plot_environment(iteration + 1, gbest_pos, START_POS, GOAL_POS,
                             ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                             MAP_BOUNDS, ax_pso, gbest_fitness)
            print(f"Iter {iteration+1}/{MAX_ITERATIONS}: W={current_W:.2f}, BestFit={gbest_fitness:.2f}, FiniteInIter={num_finite_particles_this_iter}/{NUM_PARTICLES}")

            # Optional: Early stopping
            # if np.isfinite(gbest_fitness) and gbest_fitness < (NORMAL_OBSTACLE_PENALTY_CONSTANT / 1000) and num_finite_particles_this_iter > 0 :
            #     gbest_path_nodes_check = np.vstack([START_POS, gbest_pos.reshape(NUM_WAYPOINTS, 2), GOAL_POS])
            #     if not any(check_segment_collision_with_list(gbest_path_nodes_check[k], gbest_path_nodes_check[k+1], CRITICAL_INFLATED_OBSTACLES) for k in range(len(gbest_path_nodes_check)-1)):
            #         # print(f"--- Potentially converged early at iter {iteration+1} with a good, critically-clear path. ---")
            #         # break # Uncomment for early stopping
            #         pass


    # --- Final Path Construction and Output ---
    print("\n--- PSO Finished ---")
    final_path_nodes_map = np.array([START_POS, GOAL_POS]) # Default if no valid path

    if np.isfinite(gbest_fitness) and gbest_pos is not None and np.isfinite(gbest_pos).all():
        print(f"Best Fitness Found: {gbest_fitness:.2f}")
        final_waypoints_map = gbest_pos.reshape(NUM_WAYPOINTS, 2)
        final_path_nodes_map = np.vstack([START_POS, final_waypoints_map, GOAL_POS])
    else:
        print("FAILED TO FIND A VALID NON-INFINITE PATH. Displaying direct Start-Goal line.")

    final_path_critical_collisions, final_path_normal_collisions = 0,0
    if len(final_path_nodes_map) > 1:
        for i in range(len(final_path_nodes_map) - 1):
            p1_node, p2_node = final_path_nodes_map[i], final_path_nodes_map[i+1]
            if check_segment_collision_with_list(p1_node, p2_node, CRITICAL_INFLATED_OBSTACLES): final_path_critical_collisions += 1
            elif check_segment_collision_with_list(p1_node, p2_node, NORMAL_INFLATED_OBSTACLES): final_path_normal_collisions += 1

    print(f"Displayed path is CRITICALLY collision-free: {final_path_critical_collisions == 0}.")
    if final_path_critical_collisions > 0: print(f"  WARNING: {final_path_critical_collisions} CRITICAL colliding segments in displayed path.")
    print(f"Displayed path is normally collision-free: {final_path_normal_collisions == 0}.")
    if final_path_normal_collisions > 0: print(f"  Displayed path has {final_path_normal_collisions} normal colliding segments.")

    print("Path to be Displayed (Map Coordinates):")
    for node_idx, node in enumerate(final_path_nodes_map): print(f"  Node {node_idx}: ({node[0]:.2f}, {node[1]:.2f})")

    plot_environment(MAX_ITERATIONS, gbest_pos, START_POS, GOAL_POS,
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, gbest_fitness, title_suffix=" - Final Result")
    plt.ioff(); plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from shapely.geometry import LineString, Polygon, Point

# --- Configuration ---
NUM_PARTICLES = 150
MAX_ITERATIONS = 700 # Increased for more complex environment
NUM_WAYPOINTS = 20   # Increased for more complex paths with more obstacles
DIMENSIONS = NUM_WAYPOINTS * 2

# Inertia weight decay parameters
W_MAX = 0.9
W_MIN = 0.4
C1 = 1.8
C2 = 1.8

# Fitness function weights
W_LENGTH = 1.0
W_OBSTACLE_MULTIPLIER = 1.0 # For normal obstacles
NORMAL_OBSTACLE_PENALTY_CONSTANT = 20000.0
CRITICAL_OBSTACLE_PENALTY = float('inf') # Effectively infinite penalty
W_BOUNDS = 70.0

# Environment and Robot Setup
MAP_BOUNDS = {'x_min': 0, 'x_max': 100, 'y_min': 0, 'y_max': 55}
START_POS = np.array([5, 40])
GOAL_POS = np.array([95, 40])
ROBOT_RADIUS = 0.1 # <<-- ADJUST THIS! Set to 0 for point robot, or actual radius.

# --- Obstacle Simulation, Inflation, and Critical Marking ---
def get_and_process_obstacles(robot_radius):
    print("Processing simulated obstacles...")
    # Define obstacles: coords are [(x1,y1), (x2,y2), ...], is_critical, color for plotting
    simulated_obstacles_data = [
        # Original Obstacles (modified slightly or kept)
        {'coords': [(15, 29), (15, 51), (30, 51), (30, 29)], 'is_critical': False, 'color': 'rosybrown'}, # Obstacle 1 (original 1)
        {'coords': [(48, 9), (48, 37), (60, 37), (60, 9)], 'is_critical': True, 'color': 'red'},          # Obstacle 2 (original 2, critical)

        # Added Obstacles for a total of 10
        {'coords': [(5, 48), (5, 54), (20, 54), (20, 48)], 'is_critical': False, 'color': 'lightgreen'},     # Obstacle 3 (Top-leftish)
        {'coords': [(10, 5), (10, 20), (25, 20), (25, 5)], 'is_critical': False, 'color': 'lightblue'},      # Obstacle 4 (Bottom-left)
        {'coords': [(35, 45), (35, 52), (65, 52), (65, 45)], 'is_critical': True, 'color': 'darkred'},      # Obstacle 5 (Mid-top, wide, critical)
        {'coords': [(40, 2), (40, 8), (55, 8), (55, 2)], 'is_critical': False, 'color': 'gold'},           # Obstacle 6 (Mid-bottom, small)
        {'coords': [(70, 25), (70, 50), (80, 50), (80, 25)], 'is_critical': False, 'color': 'lightcoral'},   # Obstacle 7 (Right-mid, tall)
        {'coords': [(85, 48), (85, 54), (98, 54), (98, 48)], 'is_critical': True, 'color': 'orangered'},    # Obstacle 8 (Top-right near goal, critical)
        {'coords': [(75, 5), (75, 18), (90, 18), (90, 5)], 'is_critical': False, 'color': 'mediumpurple'}, # Obstacle 9 (Bottom-right)
        {'coords': [(35, 15), (35, 25), (70, 25), (70, 15)], 'is_critical': False, 'color': 'darkseagreen'},# Obstacle 10 (Center-ish, horizontal block)
    ]

    all_inflated_obstacles = []
    critical_inflated_obstacles = []
    normal_inflated_obstacles = []
    obstacle_colors = [] # To store colors corresponding to all_inflated_obstacles

    for obs_data in simulated_obstacles_data:
        coords = obs_data['coords']
        is_critical = obs_data['is_critical']
        color = obs_data['color']

        original_poly = Polygon(coords)
        if not original_poly.is_valid:
            # print(f"DEBUG: Original simulated obstacle with coords {coords} is invalid.")
            continue # Skip invalid original polygons

        inflated_poly = original_poly
        if robot_radius > 0:
            if original_poly.is_valid:
                 inflated_poly = original_poly.buffer(robot_radius, resolution=8)
                 if not inflated_poly.is_valid:
                    # print(f"DEBUG: Inflated obstacle (from {coords}) became invalid.")
                    continue
            else:
                # print(f"DEBUG: Cannot inflate invalid original polygon: {coords}")
                continue

        all_inflated_obstacles.append(inflated_poly)
        obstacle_colors.append(color)

        if is_critical:
            critical_inflated_obstacles.append(inflated_poly)
        else:
            normal_inflated_obstacles.append(inflated_poly)

    print(f"Generated {len(all_inflated_obstacles)} total inflated obstacles.")
    print(f"  {len(critical_inflated_obstacles)} critical inflated obstacles.")
    print(f"  {len(normal_inflated_obstacles)} normal inflated obstacles.")
    return all_inflated_obstacles, critical_inflated_obstacles, normal_inflated_obstacles, obstacle_colors

# --- PSO Helper Functions ---
def check_segment_collision_with_list(p1_coords, p2_coords, obstacles_to_check):
    """Checks if a line segment p1-p2 collides with any obstacle in the given list."""
    if np.array_equal(p1_coords, p2_coords): return False # Zero-length segment
    try:
        segment = LineString([p1_coords, p2_coords])
        if not segment.is_valid: return True # Cannot form valid segment
    except Exception:
        return True # Error forming segment, assume collision for safety
    for obs_poly in obstacles_to_check:
        if segment.intersects(obs_poly):
            return True
    return False

def fitness_function(particle_waypoints_flat, start_pos_map, goal_pos_map,
                     normal_obs_list, critical_obs_list, map_bnds_map,
                     particle_idx_debug=-1, iteration_debug=-1): # Debug params
    waypoints = particle_waypoints_flat.reshape(NUM_WAYPOINTS, 2)
    full_path_nodes = np.vstack([start_pos_map, waypoints, goal_pos_map])

    path_length = 0.0
    num_normal_collisions = 0
    num_critical_collisions = 0
    num_out_of_bounds = 0

    for i in range(len(full_path_nodes) - 1):
        p1 = full_path_nodes[i]
        p2 = full_path_nodes[i+1]
        path_length += np.linalg.norm(p2 - p1)

        segment_hit_critical_this_pass = False
        if check_segment_collision_with_list(p1, p2, critical_obs_list):
            num_critical_collisions += 1
            segment_hit_critical_this_pass = True

        if not segment_hit_critical_this_pass:
            if check_segment_collision_with_list(p1, p2, normal_obs_list):
                num_normal_collisions += 1

    for wp_idx in range(NUM_WAYPOINTS):
        wp_x, wp_y = particle_waypoints_flat[wp_idx*2], particle_waypoints_flat[wp_idx*2+1]
        if not (map_bnds_map['x_min'] <= wp_x <= map_bnds_map['x_max'] and \
                map_bnds_map['y_min'] <= wp_y <= map_bnds_map['y_max']):
            num_out_of_bounds += 1

    if num_critical_collisions > 0:
        return CRITICAL_OBSTACLE_PENALTY

    fitness_val = (W_LENGTH * path_length +
                   W_OBSTACLE_MULTIPLIER * num_normal_collisions * NORMAL_OBSTACLE_PENALTY_CONSTANT +
                   W_BOUNDS * num_out_of_bounds)
    return fitness_val

def plot_environment(iteration, gbest_pos_flat,
                     start_map, goal_map,
                     all_obstacles_infl, critical_obstacles_infl, obstacle_cols, # Pass colors
                     map_bnds_map, ax_plot, current_gbest_fitness, title_suffix=""):
    ax_plot.clear()
    for i, obs_poly in enumerate(all_obstacles_infl): # Iterate through all obstacles to use stored colors
        color_to_use = obstacle_cols[i]
        alpha_val = 0.6 # Default alpha
        is_this_critical = any(obs_poly.equals(crit_obs) for crit_obs in critical_obstacles_infl)
        if is_this_critical:
            alpha_val = 0.85 # Make critical obstacles more solid

        if obs_poly.exterior and len(obs_poly.exterior.coords) > 0 :
            ax_plot.add_patch(patches.Polygon(obs_poly.exterior.coords, closed=True, color=color_to_use, alpha=alpha_val, zorder=2))

    ax_plot.plot(start_map[0], start_map[1], 'go', markersize=10, label='Start', zorder=5)
    ax_plot.plot(goal_map[0], goal_map[1], 'mo', markersize=10, label='Goal', zorder=5)

    if gbest_pos_flat is not None and len(gbest_pos_flat) > 0 and \
       np.isfinite(gbest_pos_flat).all() and np.isfinite(current_gbest_fitness):
        gbest_waypoints = gbest_pos_flat.reshape(NUM_WAYPOINTS, 2)
        gbest_path_nodes_plot = np.vstack([start_map, gbest_waypoints, goal_map])
        ax_plot.plot(gbest_path_nodes_plot[:, 0], gbest_path_nodes_plot[:, 1], 'b-', linewidth=2.5, label='Best Path', zorder=4)
    elif iteration >= 0 and not np.isfinite(current_gbest_fitness) : # iteration >= 0 to avoid plotting this at init if gbest is inf
        ax_plot.text(np.mean(ax_plot.get_xlim()), np.mean(ax_plot.get_ylim()),
                      "No valid path found yet", ha='center', va='center', fontsize=12, color='gray', zorder=3)

    ax_plot.set_xlim(map_bnds_map['x_min'], map_bnds_map['x_max'])
    ax_plot.set_ylim(map_bnds_map['y_min'], map_bnds_map['y_max'])
    ax_plot.set_title(f"PSO Path Planning - Iter: {iteration} {title_suffix}")
    ax_plot.set_xlabel("X (Map)"); ax_plot.set_ylabel("Y (Map)")
    handles, labels = ax_plot.get_legend_handles_labels()
    if handles: ax_plot.legend(loc='best')

    # --- ADDED: Display current_gbest_fitness on the plot ---
    if np.isfinite(current_gbest_fitness):
        ax_plot.text(map_bnds_map['x_min'] + 1, map_bnds_map['y_min'] + 1,  # Position of the text
                     f"Best Fitness: {current_gbest_fitness:.2f}",
                     fontsize=9, color='black',
                     bbox=dict(facecolor='white', alpha=0.7, pad=0.3)) # Optional: background for better readability
    # --- END ADDED SECTION ---

    ax_plot.set_aspect('equal', adjustable='box'); plt.pause(0.01)

# --- Main Execution ---
if __name__ == "__main__":
    ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, NORMAL_INFLATED_OBSTACLES, OBSTACLE_COLORS = \
        get_and_process_obstacles(ROBOT_RADIUS)

    if not ALL_INFLATED_OBSTACLES: print("Warning: No obstacles available for path planning.")
    if not CRITICAL_INFLATED_OBSTACLES: print("Info: No obstacles were marked as critical.")

    # Initialize particles' positions
    particles_pos = np.zeros((NUM_PARTICLES, DIMENSIONS))
    for i in range(NUM_PARTICLES):
        for d_wp in range(NUM_WAYPOINTS):
            particles_pos[i, d_wp*2] = random.uniform(MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            particles_pos[i, d_wp*2+1] = random.uniform(MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])

    # << Optional: Guided Initialization for particle 0 - Comment out if not needed >>
    if NUM_PARTICLES > 0 and len(CRITICAL_INFLATED_OBSTACLES) > 0:
        # print("DEBUG: Applying Guided Initialization for Particle 0")
        guided_wps_array = np.zeros((NUM_WAYPOINTS, 2))
        crit_obs_for_guidance = CRITICAL_INFLATED_OBSTACLES[0]

        pt1_x = START_POS[0] + (crit_obs_for_guidance.bounds[0] - START_POS[0]) * 0.3
        pt1_y = crit_obs_for_guidance.bounds[3] + 3 + ROBOT_RADIUS
        if crit_obs_for_guidance.bounds[1] > (MAP_BOUNDS['y_min'] + MAP_BOUNDS['y_max'])/2 :
             pt1_y = crit_obs_for_guidance.bounds[1] - 3 - ROBOT_RADIUS

        pt2_x = crit_obs_for_guidance.bounds[2] + (GOAL_POS[0] - crit_obs_for_guidance.bounds[2]) * 0.3
        pt2_y = pt1_y

        num_wp_seg1, num_wp_seg2 = NUM_WAYPOINTS // 3, NUM_WAYPOINTS // 3
        num_wp_seg3 = NUM_WAYPOINTS - num_wp_seg1 - num_wp_seg2
        current_wp_idx = 0
        for k in range(num_wp_seg1): t = (k + 1) / (num_wp_seg1 +1e-9); guided_wps_array[current_wp_idx, 0] = START_POS[0] + t * (pt1_x - START_POS[0]); guided_wps_array[current_wp_idx, 1] = START_POS[1] + t * (pt1_y - START_POS[1]); current_wp_idx += 1
        for k in range(num_wp_seg2): t = (k + 1) / (num_wp_seg2 +1e-9); guided_wps_array[current_wp_idx, 0] = pt1_x + t * (pt2_x - pt1_x);     guided_wps_array[current_wp_idx, 1] = pt1_y + t * (pt2_y - pt1_y);     current_wp_idx += 1
        for k in range(num_wp_seg3): t = (k + 1) / (num_wp_seg3 +1e-9); guided_wps_array[current_wp_idx, 0] = pt2_x + t * (GOAL_POS[0] - pt2_x); guided_wps_array[current_wp_idx, 1] = pt2_y + t * (GOAL_POS[1] - pt2_y); current_wp_idx += 1

        guided_wps_array[:,0] = np.clip(guided_wps_array[:,0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
        guided_wps_array[:,1] = np.clip(guided_wps_array[:,1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
        particles_pos[0, :] = guided_wps_array.flatten()
    # << End Guided Init >>

    particles_vel = np.random.randn(NUM_PARTICLES, DIMENSIONS) * 0.05
    pbest_pos = np.copy(particles_pos)
    pbest_fitness = np.full(NUM_PARTICLES, np.inf)
    gbest_pos = np.full(DIMENSIONS, np.nan)
    gbest_fitness = np.inf

    for i in range(NUM_PARTICLES): # Initial pbest/gbest evaluation
        fitness_val = fitness_function(particles_pos[i], START_POS, GOAL_POS,
                                       NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                       MAP_BOUNDS, particle_idx_debug=i, iteration_debug=-1)
        pbest_fitness[i] = fitness_val
        if fitness_val < gbest_fitness:
            gbest_fitness = fitness_val
            gbest_pos = np.copy(particles_pos[i])

    fig_pso, ax_pso = plt.subplots(figsize=(12, 8))
    plt.ion()
    print(f"\nStarting PSO. Initial best fitness: {gbest_fitness:.2f}")
    plot_environment(0, gbest_pos, START_POS, GOAL_POS,
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, gbest_fitness, title_suffix="Initial State")
    if not np.isfinite(gbest_fitness):
        print("CRITICAL INFO: No particle (even potentially guided) found a non-infinite path initially. PSO will likely fail or struggle significantly.")

    # PSO Main Loop
    for iteration in range(MAX_ITERATIONS):
        current_W = W_MAX - (W_MAX - W_MIN) * (iteration / MAX_ITERATIONS)
        num_finite_particles_this_iter = 0
        for i in range(NUM_PARTICLES):
            r1, r2 = np.random.rand(DIMENSIONS), np.random.rand(DIMENSIONS)
            gbest_term = np.zeros_like(particles_pos[i])
            if np.isfinite(gbest_fitness) and np.isfinite(gbest_pos).all():
                 gbest_term = C2 * r2 * (gbest_pos - particles_pos[i])

            current_pbest_pos = pbest_pos[i]
            if not np.isfinite(current_pbest_pos).all():
                current_pbest_pos = particles_pos[i]

            particles_vel[i] = (current_W * particles_vel[i] +
                                C1 * r1 * (current_pbest_pos - particles_pos[i]) +
                                gbest_term)

            vel_limit_x = (MAP_BOUNDS['x_max'] - MAP_BOUNDS['x_min']) * 0.04
            vel_limit_y = (MAP_BOUNDS['y_max'] - MAP_BOUNDS['y_min']) * 0.04
            particles_vel[i, ::2] = np.clip(particles_vel[i, ::2], -vel_limit_x, vel_limit_x)
            particles_vel[i, 1::2] = np.clip(particles_vel[i, 1::2], -vel_limit_y, vel_limit_y)
            particles_pos[i] += particles_vel[i]

            current_waypoints = particles_pos[i].reshape(NUM_WAYPOINTS, 2)
            current_waypoints[:, 0] = np.clip(current_waypoints[:, 0], MAP_BOUNDS['x_min'], MAP_BOUNDS['x_max'])
            current_waypoints[:, 1] = np.clip(current_waypoints[:, 1], MAP_BOUNDS['y_min'], MAP_BOUNDS['y_max'])
            particles_pos[i] = current_waypoints.flatten()

            current_fitness = fitness_function(particles_pos[i], START_POS, GOAL_POS,
                                               NORMAL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES,
                                               MAP_BOUNDS, particle_idx_debug=i, iteration_debug=iteration)
            if np.isfinite(current_fitness): num_finite_particles_this_iter +=1
            if current_fitness < pbest_fitness[i]:
                pbest_fitness[i] = current_fitness; pbest_pos[i] = np.copy(particles_pos[i])
                if current_fitness < gbest_fitness:
                    gbest_fitness = current_fitness; gbest_pos = np.copy(particles_pos[i])

        if (iteration + 1) % 20 == 0 or iteration == MAX_ITERATIONS - 1 or \
           (iteration < 20 and np.isfinite(gbest_fitness)):
            plot_environment(iteration + 1, gbest_pos, START_POS, GOAL_POS,
                             ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                             MAP_BOUNDS, ax_pso, gbest_fitness)
            print(f"Iter {iteration+1}/{MAX_ITERATIONS}: W={current_W:.2f}, BestFit={gbest_fitness:.2f}, FiniteInIter={num_finite_particles_this_iter}/{NUM_PARTICLES}")

    # --- Final Path Construction and Output ---
    print("\n--- PSO Finished ---")
    final_path_nodes_map = np.array([START_POS, GOAL_POS])

    if np.isfinite(gbest_fitness) and gbest_pos is not None and np.isfinite(gbest_pos).all():
        print(f"Best Fitness Found: {gbest_fitness:.2f}")
        final_waypoints_map = gbest_pos.reshape(NUM_WAYPOINTS, 2)
        final_path_nodes_map = np.vstack([START_POS, final_waypoints_map, GOAL_POS])
    else:
        print("FAILED TO FIND A VALID NON-INFINITE PATH. Displaying direct Start-Goal line.")

    final_path_critical_collisions, final_path_normal_collisions = 0,0
    if len(final_path_nodes_map) > 1:
        for i in range(len(final_path_nodes_map) - 1):
            p1_node, p2_node = final_path_nodes_map[i], final_path_nodes_map[i+1]
            if check_segment_collision_with_list(p1_node, p2_node, CRITICAL_INFLATED_OBSTACLES): final_path_critical_collisions += 1
            elif check_segment_collision_with_list(p1_node, p2_node, NORMAL_INFLATED_OBSTACLES): final_path_normal_collisions += 1

    print(f"Displayed path is CRITICALLY collision-free: {final_path_critical_collisions == 0}.")
    if final_path_critical_collisions > 0: print(f"  WARNING: {final_path_critical_collisions} CRITICAL colliding segments in displayed path.")
    print(f"Displayed path is normally collision-free: {final_path_normal_collisions == 0}.")
    if final_path_normal_collisions > 0: print(f"  Displayed path has {final_path_normal_collisions} normal colliding segments.")

    print("Path to be Displayed (Map Coordinates):")
    for node_idx, node in enumerate(final_path_nodes_map): print(f"  Node {node_idx}: ({node[0]:.2f}, {node[1]:.2f})")

    plot_environment(MAX_ITERATIONS, gbest_pos, START_POS, GOAL_POS,
                     ALL_INFLATED_OBSTACLES, CRITICAL_INFLATED_OBSTACLES, OBSTACLE_COLORS,
                     MAP_BOUNDS, ax_pso, gbest_fitness, title_suffix=" - Final Result")
    plt.ioff(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Make plots appear inline in the notebook
%matplotlib inline
```

```python
# --- Environment Setup ---
START_POS = np.array([5, 5])
GOAL_POS = np.array([95, 95])
OBSTACLES = [
    # (center_x, center_y, radius)
    (30, 40, 10),
    (60, 60, 15),
    (75, 25, 8)
]
# The number of intermediate points in the path
N_WAYPOINTS = 10 

# Search space boundaries for the waypoints
SEARCH_BOUNDS_X = (0, 100)
SEARCH_BOUNDS_Y = (0, 100)

# --- PSO Hyperparameters ---
N_PARTICLES = 50
N_ITERATIONS = 150
W = 0.5   # Inertia
C1 = 2.0  # Cognitive coefficient
C2 = 2.0  # Social coefficient
```

```python
def plot_environment(path=None):
    """Visualizes the environment, obstacles, and optionally a path."""
    fig, ax = plt.subplots(figsize=(5, 5))
    
    # Plot Start and Goal
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')

    # Plot Obstacles
    for obs in OBSTACLES:
        circle = patches.Circle((obs[0], obs[1]), obs[2], color='b', alpha=0.5)
        ax.add_patch(circle)

    # Plot the path if provided
    if path is not None:
        path_points = np.vstack([START_POS, path, GOAL_POS])
        ax.plot(path_points[:, 0], path_points[:, 1], 'y-', label='Path')
        ax.plot(path_points[:, 0], path_points[:, 1], 'k.', markersize=5)

    ax.set_xlim(SEARCH_BOUNDS_X)
    ax.set_ylim(SEARCH_BOUNDS_Y)
    ax.set_title("PSO Path Planning")
    ax.legend()
    ax.grid(True)
    return fig, ax

def line_intersects_circle(p1, p2, circle_center, radius):
    """Checks if the line segment from p1 to p2 intersects a circle."""
    p1 = np.array(p1)
    p2 = np.array(p2)
    circle_center = np.array(circle_center)
    
    # Vector from p1 to p2
    d = p2 - p1
    # Vector from circle center to p1
    f = p1 - circle_center
    
    a = np.dot(d, d)
    b = 2 * np.dot(f, d)
    c = np.dot(f, f) - radius**2
    
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return False  # No real intersection points
    else:
        # Check if intersection lies on the segment
        t1 = (-b - np.sqrt(discriminant)) / (2*a)
        t2 = (-b + np.sqrt(discriminant)) / (2*a)
        if 0 <= t1 <= 1 or 0 <= t2 <= 1:
            return True # Intersection is on the line segment
    return False

# Let's visualize the initial setup

print("Initial Environment:")

plot_environment()

plt.show()
```

```python
def calculate_path_cost(waypoints_flat):
    """
    Calculates the cost of a path defined by waypoints.
    A particle's position is a flat array of waypoint coordinates.
    """
    # Reshape the flat array into a list of (x, y) waypoints
    waypoints = waypoints_flat.reshape(N_WAYPOINTS, 2)
    
    # Construct the full path from start to goal through the waypoints
    full_path = np.vstack([START_POS, waypoints, GOAL_POS])
    
    total_cost = 0.0
    
    # --- 1. Path Length Cost ---
    path_length = np.sum(np.sqrt(np.sum(np.diff(full_path, axis=0)**2, axis=1)))
    total_cost += path_length
    
    # --- 2. Obstacle Collision Penalty ---
    OBSTACLE_PENALTY = 1000  # A large penalty for hitting an obstacle
    for i in range(len(full_path) - 1):
        p1 = full_path[i]
        p2 = full_path[i+1]
        for obs_x, obs_y, obs_r in OBSTACLES:
            if line_intersects_circle(p1, p2, (obs_x, obs_y), obs_r):
                total_cost += OBSTACLE_PENALTY
                
    return total_cost
```

```python
def pso_path_planning():
    # --- INITIALIZATION ---
    n_dimensions = N_WAYPOINTS * 2
    
    # Define search bounds for the flattened waypoint array
    min_bound = np.array([SEARCH_BOUNDS_X[0], SEARCH_BOUNDS_Y[0]] * N_WAYPOINTS)
    max_bound = np.array([SEARCH_BOUNDS_X[1], SEARCH_BOUNDS_Y[1]] * N_WAYPOINTS)
    
    # Initialize particle positions (paths) and velocities
    positions = np.random.rand(N_PARTICLES, n_dimensions) * (max_bound - min_bound) + min_bound
    velocities = np.random.rand(N_PARTICLES, n_dimensions) * 0.1

    # Initialize personal and global bests
    pbest_positions = np.copy(positions)
    pbest_scores = np.array([calculate_path_cost(p) for p in positions])
    
    gbest_idx = np.argmin(pbest_scores)
    gbest_position = pbest_positions[gbest_idx]
    gbest_score = pbest_scores[gbest_idx]
    
    # Store history of best scores and positions for plotting
    history = []

    # --- MAIN LOOP ---
    for i in range(N_ITERATIONS):
        for j in range(N_PARTICLES):
            # Update pbest
            if calculate_path_cost(positions[j]) < pbest_scores[j]:
                pbest_scores[j] = calculate_path_cost(positions[j])
                pbest_positions[j] = positions[j]

            # Update gbest
            if pbest_scores[j] < gbest_score:
                gbest_score = pbest_scores[j]
                gbest_position = pbest_positions[j]
        
        # Store the best position of this iteration
        history.append(gbest_position.copy())
        
        # Update velocities and positions
        r1 = np.random.rand(N_PARTICLES, n_dimensions)
        r2 = np.random.rand(N_PARTICLES, n_dimensions)
        
        velocities = (W * velocities +
                      C1 * r1 * (pbest_positions - positions) +
                      C2 * r2 * (gbest_position - positions))
        
        positions = positions + velocities
        
        # Clamp positions to stay within bounds
        positions = np.clip(positions, min_bound, max_bound)

        if i % 20 == 0:
            print(f"Iteration {i}: Best Cost = {gbest_score:.2f}")

    print(f"\nFinished! Final Best Cost = {gbest_score:.2f}")
    return gbest_position, history

# --- Run the algorithm ---
best_path_flat, history = pso_path_planning()
```

```python
# Reshape the final best path for plotting
final_path = best_path_flat.reshape(N_WAYPOINTS, 2)



print("\nVisualizing the best path found:")
plot_environment(path=final_path)

plt.show()
```
