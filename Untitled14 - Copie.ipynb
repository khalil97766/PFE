```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Make plots appear inline in the notebook
%matplotlib inline
# --- Environment Setup ---
START_POS = np.array([5, 5])
GOAL_POS = np.array([95, 95])
OBSTACLES = [
    # (center_x, center_y, radius)
    (30, 40, 10),
    (60, 60, 15),
    (75, 25, 8)
]
# The number of intermediate points in the path
N_WAYPOINTS = 10 

# Search space boundaries for the waypoints
SEARCH_BOUNDS_X = (0, 100)
SEARCH_BOUNDS_Y = (0, 100)

# --- PSO Hyperparameters ---
N_PARTICLES = 50
N_ITERATIONS = 150
W = 0.5   # Inertia
C1 = 2.0  # Cognitive coefficient
C2 = 2.0  # Social coefficient
def plot_environment(path=None):
    """Visualizes the environment, obstacles, and optionally a path."""
    fig, ax = plt.subplots(figsize=(5, 5))
    
    # Plot Start and Goal
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')

    # Plot Obstacles
    for obs in OBSTACLES:
        circle = patches.Circle((obs[0], obs[1]), obs[2], color='b', alpha=0.5)
        ax.add_patch(circle)

    # Plot the path if provided
    if path is not None:
        path_points = np.vstack([START_POS, path, GOAL_POS])
        ax.plot(path_points[:, 0], path_points[:, 1], 'y-', label='Path')
        ax.plot(path_points[:, 0], path_points[:, 1], 'k.', markersize=5)

    ax.set_xlim(SEARCH_BOUNDS_X)
    ax.set_ylim(SEARCH_BOUNDS_Y)
    ax.set_title("PSO Path Planning")
    ax.legend()
    ax.grid(True)
    return fig, ax

def line_intersects_circle(p1, p2, circle_center, radius):
    """Checks if the line segment from p1 to p2 intersects a circle."""
    p1 = np.array(p1)
    p2 = np.array(p2)
    circle_center = np.array(circle_center)
    
    # Vector from p1 to p2
    d = p2 - p1
    # Vector from circle center to p1
    f = p1 - circle_center
    
    a = np.dot(d, d)
    b = 2 * np.dot(f, d)
    c = np.dot(f, f) - radius**2
    
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return False  # No real intersection points
    else:
        # Check if intersection lies on the segment
        t1 = (-b - np.sqrt(discriminant)) / (2*a)
        t2 = (-b + np.sqrt(discriminant)) / (2*a)
        if 0 <= t1 <= 1 or 0 <= t2 <= 1:
            return True # Intersection is on the line segment
    return False

# Let's visualize the initial setup

print("Initial Environment:")

plot_environment()

plt.show()
def calculate_path_cost(waypoints_flat):
    """
    Calculates the cost of a path defined by waypoints.
    A particle's position is a flat array of waypoint coordinates.
    """
    # Reshape the flat array into a list of (x, y) waypoints
    waypoints = waypoints_flat.reshape(N_WAYPOINTS, 2)
    
    # Construct the full path from start to goal through the waypoints
    full_path = np.vstack([START_POS, waypoints, GOAL_POS])
    
    total_cost = 0.0
    
    # --- 1. Path Length Cost ---
    path_length = np.sum(np.sqrt(np.sum(np.diff(full_path, axis=0)**2, axis=1)))
    total_cost += path_length
    
    # --- 2. Obstacle Collision Penalty ---
    OBSTACLE_PENALTY = 1000  # A large penalty for hitting an obstacle
    for i in range(len(full_path) - 1):
        p1 = full_path[i]
        p2 = full_path[i+1]
        for obs_x, obs_y, obs_r in OBSTACLES:
            if line_intersects_circle(p1, p2, (obs_x, obs_y), obs_r):
                total_cost += OBSTACLE_PENALTY
                
    return total_cost
def pso_path_planning():
    # --- INITIALIZATION ---
    n_dimensions = N_WAYPOINTS * 2
    
    # Define search bounds for the flattened waypoint array
    min_bound = np.array([SEARCH_BOUNDS_X[0], SEARCH_BOUNDS_Y[0]] * N_WAYPOINTS)
    max_bound = np.array([SEARCH_BOUNDS_X[1], SEARCH_BOUNDS_Y[1]] * N_WAYPOINTS)
    
    # Initialize particle positions (paths) and velocities
    positions = np.random.rand(N_PARTICLES, n_dimensions) * (max_bound - min_bound) + min_bound
    velocities = np.random.rand(N_PARTICLES, n_dimensions) * 0.1

    # Initialize personal and global bests
    pbest_positions = np.copy(positions)
    pbest_scores = np.array([calculate_path_cost(p) for p in positions])
    
    gbest_idx = np.argmin(pbest_scores)
    gbest_position = pbest_positions[gbest_idx]
    gbest_score = pbest_scores[gbest_idx]
    
    # Store history of best scores and positions for plotting
    history = []

    # --- MAIN LOOP ---
    for i in range(N_ITERATIONS):
        for j in range(N_PARTICLES):
            # Update pbest
            if calculate_path_cost(positions[j]) < pbest_scores[j]:
                pbest_scores[j] = calculate_path_cost(positions[j])
                pbest_positions[j] = positions[j]

            # Update gbest
            if pbest_scores[j] < gbest_score:
                gbest_score = pbest_scores[j]
                gbest_position = pbest_positions[j]
        
        # Store the best position of this iteration
        history.append(gbest_position.copy())
        
        # Update velocities and positions
        r1 = np.random.rand(N_PARTICLES, n_dimensions)
        r2 = np.random.rand(N_PARTICLES, n_dimensions)
        
        velocities = (W * velocities +
                      C1 * r1 * (pbest_positions - positions) +
                      C2 * r2 * (gbest_position - positions))
        
        positions = positions + velocities
        
        # Clamp positions to stay within bounds
        positions = np.clip(positions, min_bound, max_bound)

        if i % 20 == 0:
            print(f"Iteration {i}: Best Cost = {gbest_score:.2f}")

    print(f"\nFinished! Final Best Cost = {gbest_score:.2f}")
    return gbest_position, history

# --- Run the algorithm ---
best_path_flat, history = pso_path_planning()
# Reshape the final best path for plotting
final_path = best_path_flat.reshape(N_WAYPOINTS, 2)



print("\nVisualizing the best path found:")
plot_environment(path=final_path)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
import time

# ===================================================================
# 1. CONFIGURATION AND HYPERPARAMETERS
# ===================================================================

# --- Environment Setup ---
START_POS = np.array([5, 5])
GOAL_POS = np.array([95, 95])
OBSTACLES = [
    # (center_x, center_y, radius)
    (30, 40, 10),
    (60, 60, 15),
    (75, 25, 8)
]

# --- Graph Discretization ---
# We convert the continuous space into a grid of nodes
GRID_SIZE = 5  # The distance between nodes in the grid
X_MIN, X_MAX = 0, 100
Y_MIN, Y_MAX = 0, 100

# --- ACO Hyperparameters ---
N_ANTS = 40          # Number of ants per iteration
N_ITERATIONS = 500   # Number of iterations for the algorithm
ALPHA = 1.0          # Pheromone influence factor
BETA = 5.0           # Heuristic (visibility) influence factor
RHO = 0.5            # Pheromone evaporation rate (1 - rho is the persistence)
Q = 100              # Pheromone deposit constant


# ===================================================================
# 2. HELPER FUNCTIONS AND ENVIRONMENT SETUP
# ===================================================================

def create_graph_environment():
    """Creates the grid of nodes and identifies valid nodes."""
    print("Creating the graph environment...")
    x_coords = np.arange(X_MIN, X_MAX + GRID_SIZE, GRID_SIZE)
    y_coords = np.arange(Y_MIN, Y_MAX + GRID_SIZE, GRID_SIZE)
    nodes = np.array(np.meshgrid(x_coords, y_coords)).T.reshape(-1, 2)
    
    # Identify valid nodes (not inside any obstacle)
    valid_nodes_mask = np.ones(len(nodes), dtype=bool)
    for i, node in enumerate(nodes):
        for ox, oy, r in OBSTACLES:
            if np.linalg.norm(node - np.array([ox, oy])) < r:
                valid_nodes_mask[i] = False
                break
    
    valid_nodes = nodes[valid_nodes_mask]
    n_valid_nodes = len(valid_nodes)
    
    # Create an adjacency matrix for valid nodes only
    adj_matrix = np.zeros((n_valid_nodes, n_valid_nodes))
    for i in range(n_valid_nodes):
        for j in range(i + 1, n_valid_nodes):
            dist = np.linalg.norm(valid_nodes[i] - valid_nodes[j])
            if dist <= GRID_SIZE * 1.5:  # Connect direct neighbors
                adj_matrix[i, j] = adj_matrix[j, i] = dist

    start_node_idx = np.argmin(np.linalg.norm(valid_nodes - START_POS, axis=1))
    goal_node_idx = np.argmin(np.linalg.norm(valid_nodes - GOAL_POS, axis=1))
    
    print(f"Graph created with {n_valid_nodes} valid nodes.")
    return valid_nodes, adj_matrix, start_node_idx, goal_node_idx

def plot_aco_environment(nodes, path_indices=None, title="ACO Path Planning"):
    """Visualizes the ACO environment, including the grid."""
    fig, ax = plt.subplots(figsize=(5, 5))
    
    ax.plot(nodes[:, 0], nodes[:, 1], 'k.', markersize=2, alpha=0.3, label='Grid Nodes')
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')

    for obs in OBSTACLES:
        circle = patches.Circle((obs[0], obs[1]), obs[2], color='b', alpha=0.5)
        ax.add_patch(circle)

    if path_indices is not None and len(path_indices) > 0:
        path_coords = nodes[path_indices]
        ax.plot(path_coords[:, 0], path_coords[:, 1], 'y-', linewidth=3, label='Best Path')

    ax.set_xlim(X_MIN, X_MAX)
    ax.set_ylim(Y_MIN, Y_MAX)
    ax.set_title(title)
    ax.legend()
    ax.grid(True)
    return fig, ax


# ===================================================================
# 3. ANT COLONY OPTIMIZATION ALGORITHM
# ===================================================================

def aco_path_planning(nodes, adjacency, start_node, goal_node):
    """The main ACO algorithm logic."""
    n_nodes = len(nodes)
    
    # Heuristic matrix (eta)
    eta = 1.0 / (adjacency + 1e-10)
    
    # Pheromone matrix (tau)
    pheromones = np.ones((n_nodes, n_nodes)) * 0.1
    
    best_path = []
    best_path_length = np.inf
    history = []

    print("\nStarting ACO algorithm...")
    start_time = time.time()
    
    for i in range(N_ITERATIONS):
        ant_paths = []
        ant_path_lengths = []

        # a. Construct paths for all ants
        for ant in range(N_ANTS):
            visited = [start_node]
            current_node = start_node
            
            while current_node != goal_node:
                neighbors = np.where(adjacency[current_node] > 0)[0]
                unvisited_neighbors = [n for n in neighbors if n not in visited]

                if not unvisited_neighbors:
                    break  # Ant is stuck
                
                # b. Probabilistic path selection
                ph = pheromones[current_node, unvisited_neighbors] ** ALPHA
                he = eta[current_node, unvisited_neighbors] ** BETA
                probabilities = ph * he
                probabilities /= np.sum(probabilities)
                
                next_node = np.random.choice(unvisited_neighbors, p=probabilities)
                visited.append(next_node)
                current_node = next_node
            
            if visited[-1] == goal_node:
                path_length = sum(adjacency[visited[k], visited[k+1]] for k in range(len(visited)-1))
                ant_paths.append(visited)
                ant_path_lengths.append(path_length)

                if path_length < best_path_length:
                    best_path_length = path_length
                    best_path = visited
        
        # c. Pheromone update
        pheromones *= (1 - RHO)  # Evaporation
        
        for path, length in zip(ant_paths, ant_path_lengths):
            for k in range(len(path) - 1):
                pheromones[path[k], path[k+1]] += Q / length
                pheromones[path[k+1], path[k]] += Q / length
        
        history.append(best_path)
        if (i + 1) % 50 == 0:
            print(f"Iteration {i+1}/{N_ITERATIONS}: Best Path Length = {best_path_length:.2f}")

    end_time = time.time()
    print(f"\nFinished! Total time: {end_time - start_time:.2f} seconds.")
    print(f"Final Best Path Length = {best_path_length:.2f}")
    return best_path, history

# ===================================================================
# 4. MAIN EXECUTION BLOCK
# ===================================================================

if __name__ == "__main__":
    # --- Step 1: Set up the environment ---
    nodes, adjacency, start_node, goal_node = create_graph_environment()
    
    # --- Step 2: Show the initial, empty environment ---
    fig, ax = plot_aco_environment(nodes, title="Initial Environment Setup")
    plt.show()

    # --- Step 3: Run the ACO algorithm ---
    best_path_indices, history = aco_path_planning(nodes, adjacency, start_node, goal_node)
    
    # --- Step 4: Show the final best path ---
    if best_path_indices:
        fig, ax = plot_aco_environment(nodes, path_indices=best_path_indices, title="Final Best Path Found by ACO")
        plt.show()
    else:
        print("Could not find a valid path.")

    # --- Step 5: Create and save the convergence animation ---
    print("\nCreating and saving animation... (this may take a moment)")
    fig, ax = plot_aco_environment(nodes, title="ACO Path Evolution")
    
    # Ensure there is a path to draw for the first frame
    initial_path_coords = nodes[history[0]] if history and history[0] else np.array([[],[]]).T
    line, = ax.plot(initial_path_coords[:, 0], initial_path_coords[:, 1], 'y-', linewidth=3)
    title_text = ax.text(0.5, 1.01, '', ha='center', va='bottom', transform=ax.transAxes, fontsize=12)

    def update_animation(frame):
        path_indices = history[frame]
        if path_indices:
            path_coords = nodes[path_indices]
            line.set_data(path_coords[:, 0], path_coords[:, 1])
        else:
            line.set_data([], [])
        title_text.set_text(f"ACO Path Evolution (Iteration {frame})")
        return line, title_text

    # Note: blit=False can be more robust on some systems for complex plots
    ani = FuncAnimation(fig, update_animation, frames=N_ITERATIONS, blit=False, interval=50)

    # Save the animation to a file. Requires ffmpeg.
    try:
        ani.save('aco_path_animation.mp4', writer='ffmpeg', dpi=100)
        print("Animation successfully saved to 'aco_path_animation.mp4'")
    except Exception as e:
        print(f"Error saving animation: {e}")
        print("Please ensure you have ffmpeg installed and accessible in your system's PATH.")
    
    plt.close(fig)
```
