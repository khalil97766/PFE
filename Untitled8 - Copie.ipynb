```python
import geopandas as gpd
import networkx as nx
from shapely.geometry import LineString, Point
from shapely.ops import nearest_points
import matplotlib.pyplot as plt

# Load data
gdf_route = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_building = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# Create graph from route lines
G = nx.Graph()

for idx, row in gdf_route.iterrows():
    geom = row.geometry
    if isinstance(geom, LineString):
        coords = list(geom.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i+1]
            dist = Point(p1).distance(Point(p2))
            G.add_edge(p1, p2, weight=dist)

# Remove edges that intersect buildings
for edge in list(G.edges):
    line = LineString([edge[0], edge[1]])
    if gdf_building.intersects(line).any():
        G.remove_edge(*edge)

# Define A* heuristic (Euclidean distance)
def heuristic(a, b):
    return Point(a).distance(Point(b))

# Example start and goal points (adjust to your coordinates)
start = list(G.nodes)[0]
goal = list(G.nodes)[-1]

# Run A* pathfinding
path = nx.astar_path(G, start, goal, heuristic=heuristic)

# Plotting
fig, ax = plt.subplots()
gdf_route.plot(ax=ax, color='gray')
gdf_building.plot(ax=ax, color='red')
# Draw path
path_line = LineString(path)
gpd.GeoSeries([path_line]).plot(ax=ax, color='blue')
plt.show()

```
