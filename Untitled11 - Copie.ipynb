```python
import geopandas as gpd
import networkx as nx
from shapely.geometry import LineString, Point
import matplotlib.pyplot as plt
import random

# === Load GeoJSON Files ===
gdf_route = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_building = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# === Create Graph from LineString ===
G = nx.Graph()
for _, row in gdf_route.iterrows():
    if isinstance(row.geometry, LineString):
        coords = list(row.geometry.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i + 1]
            dist = Point(p1).distance(Point(p2))
            G.add_edge(p1, p2, weight=dist)

# === Remove Edges that Intersect Buildings ===
for u, v in list(G.edges()):
    line = LineString([u, v])
    if gdf_building.intersects(line).any():
        G.remove_edge(u, v)

# === Extract Valid Coordinates from Route for Start/Goal ===
all_coords = []
for geom in gdf_route.geometry:
    if isinstance(geom, LineString):
        all_coords.extend(geom.coords)

# Pick valid start and goal from route
start = all_coords[10]  # Adjust index if needed
goal = all_coords[-10]

# Ensure start ≠ goal
if start == goal:
    goal = all_coords[-20]

# Check if path exists
if not nx.has_path(G, start, goal):
    print("⚠️ Chosen start and goal are not connected in the graph.")
    exit()

print("✅ Start and goal set properly:")
print("Start:", start)
print("Goal:", goal)

# === BFO Parameters ===
num_bacteria = 50
num_generations = 50
max_steps = 200

# === Heuristic Path Cost ===
def path_cost(path):
    return sum(G[u][v]['weight'] for u, v in zip(path[:-1], path[1:]))

# === Goal-biased Random Walk ===
def random_walk(graph, start, goal, max_steps=100):
    path = [start]
    current = start
    for _ in range(max_steps):
        neighbors = list(graph.neighbors(current))
        neighbors = [n for n in neighbors if n not in path]  # avoid loops
        if not neighbors:
            break
        neighbors.sort(key=lambda n: Point(n).distance(Point(goal)))  # goal bias
        current = neighbors[0]
        path.append(current)
        if current == goal:
            return path
    return None

# === Initial Bacteria Population ===
population = []
for _ in range(num_bacteria):
    p = random_walk(G, start, goal, max_steps)
    if p:
        population.append(p)

if not population:
    print("❌ No valid initial paths found. Try adjusting the parameters.")
    exit()

# === BFO-like Optimization Loop ===
best_path = None
best_cost = float('inf')

for _ in range(num_generations):
    new_population = []
    for path in population:
        new_path = path
        for _ in range(3):  # chemotaxis
            mutated = random_walk(G, start, goal, max_steps)
            if mutated and path_cost(mutated) < path_cost(new_path):
                new_path = mutated
        if new_path and path_cost(new_path) < best_cost:
            best_path = new_path
            best_cost = path_cost(new_path)
        new_population.append(new_path)
    population = new_population

# === Plot Result ===
fig, ax = plt.subplots()
gdf_route.plot(ax=ax, color='gray', linewidth=0.5)
gdf_building.plot(ax=ax, color='red', alpha=0.5)

# Plot all graph nodes (for debugging)
x, y = zip(*G.nodes)
ax.scatter(x, y, color='black', s=5, alpha=0.3)

if best_path:
    # To reverse path direction (goal to start), uncomment below:
    # best_path = best_path[::-1]

    gpd.GeoSeries([LineString(best_path)]).plot(ax=ax, color='blue', linewidth=2)
    plt.scatter(*start, color='green', label='Start', zorder=5)
    plt.scatter(*goal, color='purple', label='Goal', zorder=5)
    plt.legend()
    print(f"✅ Best path found with cost: {best_cost}")
else:
    print("❌ No path found after BFO.")

plt.title("BFO Path Planning")
plt.axis("equal")
plt.show()

```
