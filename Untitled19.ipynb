```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.animation import FuncAnimation
import time
import os

# Suppress TensorFlow informational messages
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
from tensorflow.keras.optimizers import Adam


# ===================================================================
# 1. ENVIRONMENT AND MODEL CONFIGURATION
# ===================================================================

# --- Environment Setup ---
START_POS = np.array([5.0, 5.0])
GOAL_POS = np.array([95.0, 95.0])
OBSTACLES = [ (30, 40, 10), (60, 60, 15), (75, 25, 8) ]
X_MIN, X_MAX = 0, 100
Y_MIN, Y_MAX = 0, 100
GOAL_THRESHOLD = 3.0
MAX_STEPS_PER_EPISODE = 300

# --- LSTM Model & Training Hyperparameters ---
# As per paper's findings (Table 6, No. 4: L128F64)
LSTM_UNITS = 128
FC_UNITS = 64
N_INPUTS = 7  # 5 sensors + Angle_theta + r_GR
N_OUTPUTS = 1 # Angular velocity omega
LEARNING_RATE = 0.005 # From paper's findings (Table 5)

# --- Reinforcement Learning Hyperparameters ---
# As per paper (Table 8)
RL_TRAINING_ROUNDS = 45 # Paper uses ~44 rounds for convergence
RL_EPISODES_PER_ROUND = 20 # Number of paths to generate to find the best one
INITIAL_EXPLORATION_RATE = 0.5
TERMINATE_EXPLORATION_RATE = 0.1

# ===================================================================
# 2. HELPER FUNCTIONS (STATE, VISUALIZATION)
# ===================================================================

def get_state_representation(current_pos, current_angle, goal_pos, last_goal_dist):
    """Calculates the 7-dimensional state vector as described in the paper."""
    # 1. 5-dimensional Lidar data
    sensor_angles = np.array([-90, -45, 0, 45, 90]) + np.rad2deg(current_angle)
    sensor_readings = np.full(5, 50.0) # Max range
    for i, angle in enumerate(sensor_angles):
        rad = np.deg2rad(angle)
        for step in np.arange(0, 50, 0.5):
            px = current_pos[0] + step * np.cos(rad)
            py = current_pos[1] + step * np.sin(rad)
            if not (X_MIN < px < X_MAX and Y_MIN < py < Y_MAX):
                sensor_readings[i] = step
                break
            for ox, oy, r in OBSTACLES:
                if np.linalg.norm([px - ox, py - oy]) < r:
                    sensor_readings[i] = step
                    break
            if sensor_readings[i] != 50.0:
                break
    
    # 2. Angle_theta: Angle between robot's heading and goal
    goal_vec = goal_pos - current_pos
    goal_angle = np.arctan2(goal_vec[1], goal_vec[0])
    angle_theta = goal_angle - current_angle
    # Normalize to [-pi, pi]
    angle_theta = (angle_theta + np.pi) % (2 * np.pi) - np.pi

    # 3. r_GR: Distance relationship to goal (Equation 8)
    current_goal_dist = np.linalg.norm(goal_pos - current_pos)
    dGNR = current_goal_dist
    dGLR = last_goal_dist
    if dGNR < 5.0 and abs(angle_theta) < np.pi / 2:
        r_GR = 1.0
    else:
        r_GR = dGLR - dGNR

    state = np.hstack([sensor_readings / 50.0, [angle_theta / np.pi], [r_GR]])
    return state, current_goal_dist

def calculate_linear_velocity(angular_velocity, state):
    """Calculates linear velocity based on angular vel and distance (Eq. 9)"""
    d_front = state[2] * 50.0 # Front sensor reading
    v = min(max(d_front - 1.5, 0.001), 0.6) * (1 - abs(angular_velocity) / (np.pi/2))
    return max(v, 0.05) # Ensure minimum movement

def plot_environment(path=None, title="Path Planning"):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.plot(START_POS[0], START_POS[1], 'go', markersize=15, label='Start')
    ax.plot(GOAL_POS[0], GOAL_POS[1], 'ro', markersize=15, label='Goal')
    for ox, oy, r in OBSTACLES:
        ax.add_patch(patches.Circle((ox, oy), r, color='b', alpha=0.5))
    if path is not None:
        path_arr = np.array(path)
        ax.plot(path_arr[:, 0], path_arr[:, 1], 'y-', label='Path')
    ax.set_xlim(X_MIN, X_MAX); ax.set_ylim(Y_MIN, Y_MAX)
    ax.set_title(title); ax.legend(); ax.grid(True)
    return fig, ax

# ===================================================================
# 3. PHASE 1: RULE-BASED CONTROLLER FOR DATA GENERATION
# ===================================================================

class RuleBasedController:
    """A simplified controller to generate initial training data,
    mimicking the role of the Fuzzy Logic controller in the paper."""
    def get_action(self, state):
        sensor_readings = state[:5] * 50.0
        angle_theta = state[5] * np.pi
        
        # Rule 1: If path is clear, head to goal
        if min(sensor_readings) > 15:
            return angle_theta * 0.8
            
        # Rule 2: Obstacle avoidance
        front_dist = sensor_readings[2]
        left_dist = sensor_readings[1]
        right_dist = sensor_readings[3]

        if front_dist < 8.0:
            # Turn away from closer side obstacle
            return -np.sign(left_dist - right_dist) * np.pi / 2.5
        
        # Rule 3: Follow direction to goal if sides are clear enough
        return angle_theta * 0.5

# ===================================================================
# 4. PHASE 2: LSTM MODEL DEFINITION AND PRE-TRAINING
# ===================================================================

def create_lstm_model():
    """Creates the LSTM model based on the paper's architecture."""
    model = Sequential([
        Input(shape=(None, N_INPUTS)), # None for variable sequence length
        LSTM(LSTM_UNITS, return_sequences=True),
        Dense(FC_UNITS, activation='relu'),
        Dense(N_OUTPUTS)
    ])
    model.compile(optimizer=Adam(learning_rate=LEARNING_RATE), loss='mse')
    return model

# ===================================================================
# 5. PHASE 3: REINFORCEMENT LEARNING TRAINER
# ===================================================================

class RL_Trainer:
    def __init__(self, model):
        self.model = model
        self.strengthened_data = []

    def run_episode(self, start_pos, goal_pos, exploration_rate):
        """Runs a single pathfinding episode."""
        current_pos = np.copy(start_pos)
        current_angle = np.arctan2(goal_pos[1] - start_pos[1], goal_pos[0] - start_pos[0])
        last_goal_dist = np.linalg.norm(goal_pos - start_pos)
        
        episode_memory = [] # Store (state, action) for this path

        for step in range(MAX_STEPS_PER_EPISODE):
            state, current_goal_dist = get_state_representation(current_pos, current_angle, goal_pos, last_goal_dist)
            last_goal_dist = current_goal_dist
            
            # Predict base action from model
            # Model expects a sequence: (batch, timesteps, features)
            state_input = np.reshape(state, (1, 1, N_INPUTS))
            base_omega = self.model.predict(state_input, verbose=0)[0][0][0]
            
            # Exploration vs. Exploitation (Îµ-greedy like)
            rand_val = np.random.rand()
            if rand_val < exploration_rate / 2:
                omega = base_omega - 0.5 # Explore left
            elif rand_val < exploration_rate:
                omega = base_omega + 0.5 # Explore right
            else:
                omega = base_omega # Exploit

            omega = np.clip(omega, -np.pi/2, np.pi/2)
            linear_v = calculate_linear_velocity(omega, state)
            
            # Store state and the *final* chosen action
            episode_memory.append((state, omega))

            # Update robot position
            current_pos[0] += linear_v * np.cos(current_angle)
            current_pos[1] += linear_v * np.sin(current_angle)
            current_angle += omega * 0.1 # dt=0.1
            current_angle = (current_angle + np.pi) % (2 * np.pi) - np.pi

            if current_goal_dist < GOAL_THRESHOLD:
                # Success! Calculate reward (lower is better)
                path_len = sum(np.linalg.norm(np.array(episode_memory[i+1][0][:2]) - np.array(episode_memory[i][0][:2])) for i in range(len(episode_memory)-1))
                reward = 0.5 * path_len + 0.5 * len(episode_memory)
                return episode_memory, reward # Return successful path and its cost

        return None, float('inf') # Failure

    def train_round(self, round_num):
        """Performs one round of RL training as described in the paper."""
        exploration_rate = max(INITIAL_EXPLORATION_RATE - round_num * ((INITIAL_EXPLORATION_RATE - TERMINATE_EXPLORATION_RATE) / RL_TRAINING_ROUNDS), TERMINATE_EXPLORATION_RATE)
        
        successful_paths = []
        path_rewards = []
        
        print(f"\n--- RL Round {round_num+1}/{RL_TRAINING_ROUNDS} (Exploration: {exploration_rate:.2f}) ---")
        # Generate N episodes and find the best one
        i = 0
        while i < RL_EPISODES_PER_ROUND:
            # Use random start/goal for robustness
            start = np.random.rand(2) * 80 + 10
            goal = np.random.rand(2) * 80 + 10
            path, reward = self.run_episode(start, goal, exploration_rate)
            if path:
                successful_paths.append(path)
                path_rewards.append(reward)
                print(f"  Episode {i+1}/{RL_EPISODES_PER_ROUND} succeeded. Cost: {reward:.2f}")
                i += 1
            else:
                print(f"  Episode failed, retrying...")
        
        if not successful_paths:
            print("No successful paths found in this round. Skipping model update.")
            return

        # Select the best path from this round and add to strengthened data
        best_path_idx = np.argmin(path_rewards)
        best_path = successful_paths[best_path_idx]
        print(f"Best path cost in this round: {path_rewards[best_path_idx]:.2f}. Adding to dataset.")
        
        # Convert path to training format (X, y)
        path_X, path_y = [], []
        for state, action in best_path:
            path_X.append(state)
            path_y.append(action)
        
        # Add to master dataset
        self.strengthened_data.append((np.array(path_X), np.array(path_y)))

        # Fine-tune the model on the strengthened data
        # Prepare all data collected so far
        all_X, all_y = [], []
        for pX, pY in self.strengthened_data:
            all_X.append(pX)
            all_y.append(pY)
            
        # We need to reshape for LSTM: (num_samples, timesteps, features)
        # Here, each path is a sample. We pad to max length.
        max_len = max(len(p) for p in all_X)
        X_train = tf.keras.preprocessing.sequence.pad_sequences(all_X, maxlen=max_len, dtype='float32', padding='post')
        y_train = tf.keras.preprocessing.sequence.pad_sequences(all_y, maxlen=max_len, dtype='float32', padding='post')
        
        # Reshape y for the model
        y_train = y_train.reshape(*y_train.shape, 1)

        print(f"Fine-tuning LSTM_FTR model on {len(X_train)} best paths...")
        self.model.fit(X_train, y_train, epochs=5, batch_size=4, verbose=0)


# ===================================================================
# 6. MAIN EXECUTION SCRIPT
# ===================================================================

if __name__ == "__main__":
    # --- PHASE 1: Generate Data with Rule-Based Controller ---
    print("--- PHASE 1: Generating data using Rule-Based Controller ---")
    controller = RuleBasedController()
    initial_dataset = []
    for _ in range(50): # Generate 50 paths
        path, _ = RL_Trainer(None).run_episode(START_POS, GOAL_POS, 0) # No exploration
        if path:
            initial_dataset.extend(path)
    
    print(f"Generated an initial dataset of {len(initial_dataset)} (state, action) pairs.")
    
    # --- PHASE 2: Pre-train the LSTM_FT model ---
    print("\n--- PHASE 2: Pre-training the LSTM_FT model ---")
    X_pre, y_pre = [], []
    for state, action in initial_dataset:
        X_pre.append(state)
        y_pre.append(action)
    
    X_pre = np.array(X_pre).reshape(1, len(X_pre), N_INPUTS)
    y_pre = np.array(y_pre).reshape(1, len(y_pre), N_OUTPUTS)

    lstm_ft_model = create_lstm_model()
    lstm_ft_model.fit(X_pre, y_pre, epochs=100, verbose=1)
    lstm_ft_model.save('lstm_ft_model.h5')
    print("LSTM_FT model pre-trained and saved.")

    # --- PHASE 3: Fine-tune with Reinforcement Learning ---
    print("\n--- PHASE 3: Fine-tuning to create LSTM_FTR model ---")
    lstm_ftr_model = tf.keras.models.load_model('lstm_ft_model.h5')
    rl_trainer = RL_Trainer(lstm_ftr_model)
    for i in range(RL_TRAINING_ROUNDS):
        rl_trainer.train_round(i)
    
    lstm_ftr_model.save('lstm_ftr_model.h5')
    print("\nLSTM_FTR model fine-tuned and saved.")

    # --- PHASE 4: Final Evaluation ---
    print("\n--- PHASE 4: Evaluating the final LSTM_FTR model ---")
    final_model = tf.keras.models.load_model('lstm_ftr_model.h5')
    eval_path, eval_cost = RL_Trainer(final_model).run_episode(START_POS, GOAL_POS, 0) # No exploration
    
    if eval_path:
        print(f"Final model successfully found a path with cost: {eval_cost:.2f}")
        path_coords = []
        pos, angle = np.copy(START_POS), np.arctan2(GOAL_POS[1]-START_POS[1], GOAL_POS[0]-START_POS[0])
        for state, omega in eval_path:
            path_coords.append(pos.copy())
            linear_v = calculate_linear_velocity(omega, state)
            pos[0] += linear_v * np.cos(angle)
            pos[1] += linear_v * np.sin(angle)
            angle += omega * 0.1
        
        fig, ax = plot_environment(path_coords, title="Final Path by LSTM_FTR Model")
        plt.show()
    else:
        print("Final model failed to find a path.")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import random

# --- (UNCHANGED) 1. Environment Constants and Robot Simulation ---

# Constants
START_POS = (0.0, 0.0)
GOAL_POS = (10.0, 10.0)
ROBOT_SPEED = 1.0
MAX_TURN_RATE = np.deg2rad(45)
DT = 0.1
MAX_EPISODE_STEPS = 200
GOAL_RADIUS = 0.5

N_INPUTS = 2
N_OUTPUTS = 1

class Robot:
    def __init__(self, x=0.0, y=0.0, theta=np.pi / 4):
        self.x = x
        self.y = y
        self.theta = theta

    def move(self, omega, dt=DT, speed=ROBOT_SPEED):
        self.theta += omega * dt
        self.theta = math.atan2(math.sin(self.theta), math.cos(self.theta))
        self.x += speed * math.cos(self.theta) * dt
        self.y += speed * math.sin(self.theta) * dt

    def get_pos(self):
        return (self.x, self.y)

# --- (UNCHANGED) 2. RL Trainer and Model Creation ---

def get_state_from_robot(robot, goal_pos):
    """Calculates the state vector. Moved outside class to be a general utility."""
    goal_x, goal_y = goal_pos
    dx = goal_x - robot.x
    dy = goal_y - robot.y
    distance = math.sqrt(dx**2 + dy**2)
    angle_to_goal = math.atan2(dy, dx)
    angle_diff = angle_to_goal - robot.theta
    angle_diff = math.atan2(math.sin(angle_diff), math.cos(angle_diff))
    return np.array([distance, angle_diff])

class RL_Trainer:
    def __init__(self, model):
        self.model = model

    def run_episode(self, start_pos, start_theta, goal_pos):
        robot = Robot(x=start_pos[0], y=start_pos[1], theta=start_theta)
        path_taken = []
        for step in range(MAX_EPISODE_STEPS):
            current_pos = robot.get_pos()
            path_taken.append(current_pos)
            distance_to_goal = math.sqrt((current_pos[0] - goal_pos[0])**2 + (current_pos[1] - goal_pos[1])**2)
            if distance_to_goal < GOAL_RADIUS:
                return path_taken, True
            
            state = get_state_from_robot(robot, goal_pos)
            state_input = np.reshape(state, (1, N_INPUTS))
            base_omega_normalized = self.model.predict(state_input, verbose=0)[0][0]
            action_omega = base_omega_normalized * MAX_TURN_RATE
            robot.move(omega=action_omega)
        return path_taken, False

def create_rl_model(n_inputs, n_outputs):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(n_inputs,)),
        Dense(64, activation='relu'),
        Dense(n_outputs, activation='tanh')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mean_squared_error')
    print("Keras model created successfully.")
    model.summary()
    return model

# --- 3. Main Execution Block with NEW Training Logic ---

if __name__ == "__main__":
    
    # =========================================================================
    # PART 1: Generate Training Data with an "Expert" Policy
    # =========================================================================
    print("\n--- Part 1: Generating Training Data ---")

    def expert_policy(state):
        """
        A simple proportional controller. It turns the robot towards the goal.
        The turning rate is proportional to the angle difference.
        """
        distance, angle_diff = state
        # A simple proportional gain. Increase it for sharper turns.
        gain = 2.0 
        omega = gain * angle_diff
        # Ensure the turning rate doesn't exceed the robot's physical limits
        return np.clip(omega, -MAX_TURN_RATE, MAX_TURN_RATE)

    # We will generate data from multiple random starting points to make the model robust.
    num_expert_episodes = 200
    states_data = []
    actions_data = []

    for i in range(num_expert_episodes):
        # Start from random positions and angles to get diverse data
        start_x = random.uniform(-2, 2)
        start_y = random.uniform(-2, 2)
        start_theta = random.uniform(-np.pi, np.pi)
        
        robot = Robot(x=start_x, y=start_y, theta=start_theta)
        
        for _ in range(MAX_EPISODE_STEPS):
            state = get_state_from_robot(robot, GOAL_POS)
            
            # Check for success condition
            if state[0] < GOAL_RADIUS:
                break
            
            action = expert_policy(state)
            
            # Store the state and the expert action
            states_data.append(state)
            # Normalize the action to be between -1 and 1 for the tanh output layer
            normalized_action = action / MAX_TURN_RATE
            actions_data.append(normalized_action)
            
            # Move the robot according to the expert
            robot.move(omega=action)

    X_train = np.array(states_data)
    y_train = np.array(actions_data)

    print(f"Generated {X_train.shape[0]} training samples from {num_expert_episodes} expert episodes.")

    # =========================================================================
    # PART 2: Create and Train the Neural Network
    # =========================================================================
    print("\n--- Part 2: Training the Model ---")
    
    my_model = create_rl_model(N_INPUTS, N_OUTPUTS)
    
    history = my_model.fit(
        X_train,
        y_train,
        epochs=30,
        batch_size=32,
        validation_split=0.2, # Use 20% of data for validation
        verbose=1
    )

    # Plot training loss
    plt.figure()
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)
    plt.show()


    # =========================================================================
    # PART 3: Evaluate the Trained Model
    # =========================================================================
    print("\n--- Part 3: Evaluating the Trained Model ---")
    
    rl_trainer = RL_Trainer(my_model)
    
    num_eval_paths = 10
    plt.figure(figsize=(10, 10))
    
    for i in range(num_eval_paths):
        # Start from different positions to test generalization
        start_x = random.uniform(-1, 1)
        start_y = random.uniform(-1, 1)
        start_theta = random.uniform(np.pi/4 - 0.5, np.pi/4 + 0.5) # Start generally facing the goal
        
        path, success = rl_trainer.run_episode((start_x, start_y), start_theta, GOAL_POS)
        
        path_color = 'g-' if success else 'm-' # Green for success, Magenta for fail
        path_label = 'Successful Path' if success else 'Failed Path'
        
        if path:
            path_np = np.array(path)
            # Avoid duplicate labels in the legend
            if i > 0: path_label = None
            plt.plot(path_np[:, 0], path_np[:, 1], path_color, alpha=0.7, label=path_label)

    # Plot Start and Goal markers
    plt.plot(GOAL_POS[0], GOAL_POS[1], 'r*', markersize=20, label='Goal')
    goal_circle = plt.Circle(GOAL_POS, GOAL_RADIUS, color='r', fill=False, linestyle='--', label='Goal Radius')
    plt.gca().add_patch(goal_circle)

    plt.title(f"{num_eval_paths} Paths Generated by TRAINED Model")
    plt.xlabel("X Position (m)")
    plt.ylabel("Y Position (m)")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```
