```python
import geopandas as gpd
import networkx as nx
from shapely.geometry import LineString, Point
import matplotlib.pyplot as plt

# Load data
gdf_route = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_building = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# Create graph from route lines
G = nx.Graph()

for idx, row in gdf_route.iterrows():
    geom = row.geometry
    if isinstance(geom, LineString):
        coords = list(geom.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i + 1]
            dist = Point(p1).distance(Point(p2))
            G.add_edge(p1, p2, weight=dist)

# Remove edges that intersect buildings
for u, v in list(G.edges()):
    line = LineString([u, v])
    if gdf_building.intersects(line).any():
        G.remove_edge(u, v)

# Define start and goal (first and last nodes in the graph)
start = list(G.nodes)[0]
goal = list(G.nodes)[-1]

# Run Dijkstra's algorithm
path = nx.dijkstra_path(G, start, goal, weight='weight')

# Plotting
fig, ax = plt.subplots()
gdf_route.plot(ax=ax, color='gray', linewidth=0.5)
gdf_building.plot(ax=ax, color='red', alpha=0.5)

# Draw shortest path
gpd.GeoSeries([LineString(path)]).plot(ax=ax, color='blue', linewidth=2)
plt.scatter(*start, color='green', label='Start')
plt.scatter(*goal, color='purple', label='Goal')
plt.legend()
plt.show()

```
