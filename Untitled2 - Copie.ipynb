```python
import geopandas as gpd
import matplotlib.pyplot as plt
import networkx as nx
from shapely.geometry import LineString, Point
import numpy as np
import random

# Load GeoJSON files
gdf_rute = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_bulding = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# Build graph
G = nx.Graph()
for _, row in gdf_rute.iterrows():
    if isinstance(row.geometry, LineString):
        coords = list(row.geometry.coords)
        for i in range(len(coords) - 1):
            p1, p2 = coords[i], coords[i + 1]
            dist = LineString([p1, p2]).length
            G.add_edge(p1, p2, weight=dist)

# Ensure same CRS
if gdf_bulding.crs != gdf_rute.crs:
    gdf_bulding = gdf_bulding.to_crs(gdf_rute.crs)

# Get building centroid
valid_buildings = gdf_bulding[gdf_bulding.is_valid & gdf_bulding.geometry.type.isin(['Polygon', 'MultiPolygon'])]
building_geom = valid_buildings.geometry.iloc[0]
building_point = building_geom.centroid

# Find nearest node in graph
graph_nodes = [Point(n) for n in G.nodes]
start_node_idx = min(range(len(graph_nodes)), key=lambda i: building_point.distance(graph_nodes[i]))
start_node = list(G.nodes)[start_node_idx]
end_node = random.choice(list(G.nodes))

print("Start:", start_node)
print("End:", end_node)

# --- ACO Implementation ---
class AntColony:
    def __init__(self, graph, ants, evaporation, alpha=1.0, beta=2.0, iterations=100):
        self.graph = graph
        self.ants = ants
        self.evaporation = evaporation
        self.alpha = alpha
        self.beta = beta
        self.iterations = iterations
        self.pheromone = {edge: 1.0 for edge in graph.edges}

    def run(self, start, end):
        best_path = None
        best_cost = float("inf")

        for _ in range(self.iterations):
            all_paths = []
            for _ in range(self.ants):
                path = self.construct_path(start, end)
                if path:
                    cost = self.calculate_cost(path)
                    if cost < best_cost:
                        best_path = path
                        best_cost = cost
                    all_paths.append((path, cost))

            # Evaporate pheromones
            for edge in self.pheromone:
                self.pheromone[edge] *= (1 - self.evaporation)

            # Deposit new pheromones
            for path, cost in all_paths:
                for i in range(len(path) - 1):
                    edge = (path[i], path[i + 1])
                    if edge in self.pheromone:
                        self.pheromone[edge] += 1.0 / cost
                    elif (edge[1], edge[0]) in self.pheromone:
                        self.pheromone[(edge[1], edge[0])] += 1.0 / cost

        return best_path, best_cost

    def construct_path(self, start, end):
        path = [start]
        visited = set(path)
        current = start
        while current != end:
            neighbors = [n for n in self.graph.neighbors(current) if n not in visited]
            if not neighbors:
                return None
            probs = []
            for neighbor in neighbors:
                edge = (current, neighbor)
                weight = self.graph[current][neighbor]['weight']
                pher = self.pheromone.get(edge, 1.0)
                prob = (pher ** self.alpha) * ((1.0 / weight) ** self.beta)
                probs.append(prob)
            total = sum(probs)
            if total == 0:
                return None
            probs = [p / total for p in probs]
            current = random.choices(neighbors, weights=probs, k=1)[0]
            path.append(current)
            visited.add(current)
        return path

    def calculate_cost(self, path):
        return sum(self.graph[path[i]][path[i + 1]]['weight'] for i in range(len(path) - 1))


# Run ACO
aco = AntColony(G, ants=20, evaporation=0.1, alpha=1, beta=3, iterations=100)
shortest_path, path_cost = aco.run(start_node, end_node)
print("Path cost:", path_cost)

# --- Plotting the result ---
fig, ax = plt.subplots(figsize=(10, 8))
gdf_rute.plot(ax=ax, color="gray", linewidth=0.5)
gdf_bulding.plot(ax=ax, color="red", alpha=0.5)

# Plot path
if shortest_path:
    path_line = LineString(shortest_path)
    gpd.GeoSeries([path_line], crs=gdf_rute.crs).plot(ax=ax, color="blue", linewidth=2)

plt.title("Shortest Path using ACO")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
