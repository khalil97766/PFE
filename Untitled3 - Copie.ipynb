```python
import geopandas as gpd
import matplotlib.pyplot as plt
import networkx as nx
from shapely.geometry import LineString, Point
import random
import numpy as np
from mealpy.swarm_based.GWO import OriginalGWO
from mealpy.utils.problem import Problem
from mealpy.utils.space import IntegerVar

# 1.   GeoJSON
gdf_rute = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_bulding = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# 2.   
gdf_rute.plot(color="blue", markersize=5)
plt.title("The Route of Bejaia")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()

gdf_bulding.plot(color="red", markersize=5)
plt.title("The Buildings of Bejaia")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()

# 3.     
G = nx.Graph()
for _, row in gdf_rute.iterrows():
    if isinstance(row.geometry, LineString):
        coords = list(row.geometry.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i + 1]
            dist = LineString([p1, p2]).length
            G.add_edge(p1, p2, weight=dist)

print(" :", len(G.nodes))
print(" :", len(G.edges))

# 4.     
if gdf_bulding.crs != gdf_rute.crs:
    gdf_bulding = gdf_bulding.to_crs(gdf_rute.crs)

valid_buildings = gdf_bulding[gdf_bulding.is_valid & gdf_bulding.geometry.type.isin(['Polygon', 'MultiPolygon'])]
if len(valid_buildings) == 0:
    raise ValueError("       !")

building_point = valid_buildings.geometry.iloc[0].centroid
graph_nodes = [Point(n) for n in G.nodes]
start_node_idx = min(range(len(graph_nodes)), key=lambda i: building_point.distance(graph_nodes[i]))
start_node = list(G.nodes)[start_node_idx]
end_node = random.choice(list(G.nodes))

print("  :", start_node)
print("  :", end_node)

# 5.  
def fitness(path):
    total_distance = 0
    for i in range(len(path) - 1):
        if G.has_edge(path[i], path[i + 1]):
            total_distance += G[path[i]][path[i + 1]]['weight']
        else:
            total_distance += 1e6  # 
    return total_distance

# 6.       
def decode_solution_to_path(solution):
    path = []
    for idx in solution:
        path.append(list(G.nodes)[int(idx)])
    return path

# 7.      
class PathPlanningProblem(Problem):
    def __init__(self, graph, dim):
        self.G = graph
        self.dim = dim
        vars = [IntegerVar(0, len(graph.nodes) - 1) for _ in range(dim)]
        super().__init__(bounds=vars, minmax="min")

    def obj_func(self, x):
        path = decode_solution_to_path(x)
        path[0] = start_node
        path[-1] = end_node
        return fitness(path)

# 8.  GWO
problem = PathPlanningProblem(G, dim=20)
model = OriginalGWO(epoch=100, pop_size=30)
g_best = model.solve(problem)

# 9.  
best_position = g_best.solution
best_path = decode_solution_to_path(best_position)
best_path[0] = start_node
best_path[-1] = end_node

print("\n  :")
for i, node in enumerate(best_path):
    print(f"{i+1}: {node}")
print(f"  : {fitness(best_path):.2f}")

# 10.     
base = gdf_rute.plot(color="lightblue", linewidth=1, label="Roads")

#  
gdf_bulding.plot(ax=base, color="salmon", label="Buildings")

#      
from shapely.geometry import LineString
path_line = LineString(best_path)
gpd.GeoSeries([path_line]).plot(ax=base, color="green", linewidth=2, label="Optimal Path")

#    
plt.scatter(*start_node, color='blue', label='Start', zorder=5)
plt.scatter(*end_node, color='black', label='End', zorder=5)

#  
plt.legend()
plt.title("Shortest Path using GWO")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import networkx as nx
from shapely.geometry import LineString, Point, Polygon
import osmnx as ox

# 1. Load Data
gdf_rute = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_bulding = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# 2. Build the Graph from Roads
G = nx.Graph()
for _, row in gdf_rute.iterrows():
    if isinstance(row.geometry, LineString):
        coords = list(row.geometry.coords)
        for i in range(len(coords) - 1):
            p1 = coords[i]
            p2 = coords[i + 1]
            # Use osmnx to get a realistic distance weight
            dist = ox.distance.great_circle_vec(p1[1], p1[0], p2[1], p2[0])
            G.add_edge(p1, p2, weight=dist)

print(f"Graph created with {len(G.nodes)} nodes and {len(G.edges)} edges.")

# 3. Incorporate Buildings as Obstacles
# Remove any road nodes that fall inside a building polygon
nodes_to_remove = []
for node in G.nodes():
    point = Point(node)
    # Check if the node is within any building polygon
    if any(building.contains(point) for building in gdf_bulding.geometry):
        nodes_to_remove.append(node)

G.remove_nodes_from(nodes_to_remove)
print(f"Removed {len(nodes_to_remove)} nodes that were inside buildings.")
print(f"Graph now has {len(G.nodes)} nodes and {len(G.edges)} edges.")


# 4. Define Start and End Points
# Use the same logic for the start node (closest to a building centroid)
building_point = gdf_bulding.geometry.iloc[0].centroid
start_node = ox.distance.nearest_nodes(G, building_point.x, building_point.y)

# Select a more distant end point for a better demonstration
end_node_point = gdf_rute.geometry.iloc[-1].centroid
end_node = ox.distance.nearest_nodes(G, end_node_point.x, end_node_point.y)

print(f" Start Point: {start_node}")
print(f" End Point: {end_node}")

# 5. Find the Shortest Path using A*
try:
    # Heuristic function for A* (straight-line distance)
    def heuristic(u, v):
        return ox.distance.great_circle_vec(u[1], u[0], v[1], v[0])

    shortest_path_nodes = nx.astar_path(G, start_node, end_node, heuristic=heuristic, weight='weight')
    shortest_path_length = nx.astar_path_length(G, start_node, end_node, heuristic=heuristic, weight='weight')

    print(f"\n A* Algorithm Successful!")
    print(f" Optimal Path Length: {shortest_path_length:.2f} meters")

    # 6. Visualization
    base = gdf_rute.plot(color="lightblue", linewidth=0.5, label="All Roads", figsize=(12, 10))
    gdf_bulding.plot(ax=base, color="salmon", label="Buildings")

    # Draw the optimal path
    path_line = LineString(shortest_path_nodes)
    gpd.GeoSeries([path_line]).plot(ax=base, color="green", linewidth=2.5, label="A* Optimal Path")

    # Mark Start and End
    plt.scatter(*start_node, color='blue', s=100, label='Start', zorder=5)
    plt.scatter(*end_node, color='black', s=100, label='End', zorder=5)

    plt.legend()
    plt.title("Optimal Path in Bejaia using A* Algorithm")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.grid(True)
    plt.show()

except nx.NetworkXNoPath:
    print(f" No path could be found between {start_node} and {end_node}.")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import networkx as nx
from shapely.geometry import LineString, Point
import random
import numpy as np
import osmnx as ox
from mealpy.swarm_based.GWO import OriginalGWO
from mealpy.utils.problem import Problem
from mealpy.utils.space import FloatVar

# 1. Load Data
gdf_rute = gpd.read_file("C:/Users/acer/Downloads/rute.geojson")
gdf_bulding = gpd.read_file("C:/Users/acer/Downloads/bulding.geojson")

# 2. Build the Graph
G = nx.Graph()
for _, row in gdf_rute.iterrows():
    if isinstance(row.geometry, LineString):
        coords = list(row.geometry.coords)
        for i in range(len(coords) - 1):
            p1, p2 = coords[i], coords[i+1]
            dist = ox.distance.great_circle_vec(p1[1], p1[0], p2[1], p2[0])
            G.add_edge(p1, p2, weight=dist)

print(f"Graph created with {len(G.nodes)} nodes and {len(G.edges)} edges.")

# 3. Remove Nodes Inside Buildings (Obstacles)
nodes_to_remove = [node for node in G.nodes() if any(building.contains(Point(node)) for building in gdf_bulding.geometry)]
G.remove_nodes_from(nodes_to_remove)
print(f"Removed {len(nodes_to_remove)} nodes inside buildings.")
print(f"Graph now has {len(G.nodes)} nodes and {len(G.edges)} edges.")

# 4. Define Start and End Points
building_point = gdf_bulding.geometry.iloc[0].centroid
start_node = ox.distance.nearest_nodes(G, building_point.x, building_point.y)
# Use a consistent end point for comparison
end_node_point = gdf_rute.geometry.iloc[-1].centroid
end_node = ox.distance.nearest_nodes(G, end_node_point.x, end_node_point.y)

print(f" Start Point: {start_node}")
print(f" End Point: {end_node}")

# Heuristic function for distance penalty
def heuristic(u, v):
    return ox.distance.great_circle_vec(u[1], u[0], v[1], v[0])

# 5. Improved Fitness and Problem Definition for GWO
class PathPlanningProblemGWO(Problem):
    def __init__(self, graph, start, end, max_path_len):
        self.G = graph
        self.start_node = start
        self.end_node = end
        self.max_path_len = max_path_len
        # The solution is a list of choices (floats between 0 and 1)
        bounds = [FloatVar(0, 1) for _ in range(max_path_len)]
        super().__init__(bounds=bounds, minmax="min")

    def obj_func(self, solution):
        current_node = self.start_node
        path = [current_node]
        total_distance = 0.0
        
        for i in range(self.max_path_len):
            if current_node == self.end_node:
                return total_distance  # Path found

            neighbors = list(self.G.neighbors(current_node))
            if not neighbors:
                # Dead end, huge penalty
                return total_distance + 1e7

            # Use the solution vector to decide which neighbor to visit
            choice_index = int(solution[i] * len(neighbors))
            next_node = neighbors[choice_index]

            # Penalize revisiting nodes to avoid loops
            if next_node in path:
                total_distance += self.G[current_node][next_node]['weight'] * 1.5 # Penalty for loop
            else:
                total_distance += self.G[current_node][next_node]['weight']
            
            path.append(next_node)
            current_node = next_node

        # If loop finishes without reaching the end, add a penalty based on distance to target
        penalty = heuristic(current_node, self.end_node) * 10 # Heuristic penalty
        return total_distance + penalty

# Helper to reconstruct the path from the best GWO solution
def decode_gwo_solution_to_path(g, solution, start, end, max_len):
    current_node = start
    path = [current_node]
    for i in range(max_len):
        if current_node == end:
            break
        neighbors = list(g.neighbors(current_node))
        if not neighbors:
            break
        choice_index = int(solution[i] * len(neighbors))
        next_node = neighbors[choice_index]
        path.append(next_node)
        current_node = next_node
    # Ensure the path ends exactly at the end_node if it was reached
    if end in path:
        return path[:path.index(end)+1]
    return path

# 6. Run the Improved GWO
# max_path_len is now an upper limit, not a fixed size.
problem = PathPlanningProblemGWO(G, start_node, end_node, max_path_len=200) 
model = OriginalGWO(epoch=100, pop_size=50)
g_best = model.solve(problem)

# 7. Analyze and Visualize the Result
best_solution_vector = g_best.solution
best_path_gwo = decode_gwo_solution_to_path(G, best_solution_vector, start_node, end_node, 200)
final_path_length = sum(G[best_path_gwo[i]][best_path_gwo[i+1]]['weight'] for i in range(len(best_path_gwo)-1))

print("\n--- GWO Optimization Results ---")
print(f" Best Path Found by GWO:")
print(f" Path Length: {final_path_length:.2f} meters")

# 8. Visualization
base = gdf_rute.plot(color="lightblue", linewidth=0.5, label="All Roads", figsize=(12, 10))
gdf_bulding.plot(ax=base, color="salmon", label="Buildings")
path_line_gwo = LineString(best_path_gwo)
gpd.GeoSeries([path_line_gwo]).plot(ax=base, color="purple", linewidth=2.5, label="GWO Path")
plt.scatter(*start_node, color='blue', s=100, label='Start', zorder=5)
plt.scatter(*end_node, color='black', s=100, label='End', zorder=5)
plt.legend()
plt.title("Shortest Path in Bejaia using Improved GWO")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
