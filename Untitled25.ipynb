```python
import osmnx as ox
import networkx as nx
import math
import matplotlib.pyplot as plt
import numpy as np

# ==============================================================================
# Part 1: Artificial Potential Field (APF) Algorithm
# (No changes here)
# ==============================================================================

class APF:
    def __init__(self, attractive_gain, repulsive_gain, obstacle_influence_radius):
        self.K_att = attractive_gain
        self.K_rep = repulsive_gain
        self.rho0 = obstacle_influence_radius

    def compute_total_force(self, robot_xy, waypoint_xy, dynamic_obstacles_xy):
        att_force_x = self.K_att * (waypoint_xy[0] - robot_xy[0])
        att_force_y = self.K_att * (waypoint_xy[1] - robot_xy[1])
        total_rep_force_x, total_rep_force_y = 0, 0
        for obs_x, obs_y in dynamic_obstacles_xy:
            dist_to_obs = math.hypot(robot_xy[0] - obs_x, robot_xy[1] - obs_y)
            if dist_to_obs < self.rho0 and dist_to_obs > 0.01:
                magnitude = self.K_rep * (1.0/dist_to_obs - 1.0/self.rho0) * (1.0/(dist_to_obs**2))
                angle = math.atan2(robot_xy[1] - obs_y, robot_xy[0] - obs_x)
                total_rep_force_x += magnitude * math.cos(angle)
                total_rep_force_y += magnitude * math.sin(angle)
        return att_force_x + total_rep_force_x, att_force_y + total_rep_force_y

# ==============================================================================
# Part 2: Robot Simulation Physics
# (No changes here)
# ==============================================================================
class RobotSimulation:
    def __init__(self, start_x, start_y):
        self.x = start_x
        self.y = start_y

    def update_state(self, force_x, force_y, dt, max_speed=15.0):
        magnitude = math.hypot(force_x, force_y)
        if magnitude < 0.01: return
        speed = min(max_speed, magnitude)
        self.x += (force_x / magnitude) * speed * dt
        self.y += (force_y / magnitude) * speed * dt

# ==============================================================================
# Main Simulation Program
# ==============================================================================
if __name__ == "__main__":
    
    # --- 1. GLOBAL PLANNING: Get a real-world route from OpenStreetMap ---
    print("--- Downloading real-world map data from OpenStreetMap... ---")
    
    # --- FIX: Use a central point and distance instead of a place name ---
    # This is a much more robust way to define the area of interest.
    # We will define a point roughly in the middle of our desired start and end.
    map_center_point = (37.7845, -122.4036) 
    
    # Download the street network for driving within 1500 meters of our center point.
    graph = ox.graph_from_point(map_center_point, dist=1500, network_type='drive')
    
    # --- The rest of the code works as before ---
    
    # Define real start and end points using latitude and longitude
    start_latlon = (37.7819, -122.4072) # Near Moscone Center
    end_latlon = (37.7870, -122.4000)   # Near Union Square
    
    # Find the nearest street intersections (nodes) to our start/end points
    start_node = ox.distance.nearest_nodes(graph, X=start_latlon[1], Y=start_latlon[0])
    end_node = ox.distance.nearest_nodes(graph, X=end_latlon[1], Y=end_latlon[0])
    
    # Find the shortest path using Dijkstra/A*
    print(f"--- Calculating shortest real-world path from {start_latlon} to {end_latlon}... ---")
    route_nodes = nx.shortest_path(graph, start_node, end_node, weight='length')
    
    # Convert the route from GPS nodes to a list of (x, y) waypoints in meters.
    route_waypoints_xy = [(graph.nodes[node]['x'], graph.nodes[node]['y']) for node in route_nodes]
    print(f"Global path found with {len(route_waypoints_xy)} waypoints.")

    # --- 2. SETUP THE SIMULATION ---
    # Place a dynamic obstacle on the path
    obstacle_position = (route_waypoints_xy[2][0], route_waypoints_xy[2][1] + 30)
    DYNAMIC_OBSTACLES = [obstacle_position]
    print(f"Placing a dynamic obstacle at {obstacle_position}")

    robot_sim = RobotSimulation(route_waypoints_xy[0][0], route_waypoints_xy[0][1])
    apf_controller = APF(attractive_gain=1.0, repulsive_gain=1000.0, obstacle_influence_radius=20.0)
    
    robot_path_history = [(robot_sim.x, robot_sim.y)]
    path_index = 1
    dt = 0.1

    print("\n--- Starting High-Fidelity Simulation ---")
    # --- 3. RUN THE MAIN SIMULATION LOOP ---
    for step in range(5000):
        if path_index >= len(route_waypoints_xy):
            print(f"--- Mission accomplished in {step} steps! ---")
            break

        current_pos_xy = (robot_sim.x, robot_sim.y)
        current_waypoint_xy = route_waypoints_xy[path_index]
        
        force_x, force_y = apf_controller.compute_total_force(
            current_pos_xy, 
            current_waypoint_xy, 
            DYNAMIC_OBSTACLES
        )
        
        robot_sim.update_state(force_x, force_y, dt)
        robot_path_history.append((robot_sim.x, robot_sim.y))
        
        dist_to_waypoint = math.hypot(current_pos_xy[0] - current_waypoint_xy[0], current_pos_xy[1] - current_waypoint_xy[1])
        if dist_to_waypoint < 10.0:
            print(f"Waypoint {path_index} reached. Moving to the next.")
            path_index += 1
            
    if path_index < len(route_waypoints_xy):
        print("--- Simulation ended, but goal was not reached. ---")
        
    # --- 4. VISUALIZE THE RESULTS ---
    print("--- Generating plot... ---")
    
    fig, ax = ox.plot_graph_route(graph, route_nodes, route_color='g', route_linewidth=4,
                                  node_size=0, bgcolor='#FFFFFF', edge_color='#AAAAAA',
                                  show=False, close=False)
    
    robot_x_coords, robot_y_coords = zip(*robot_path_history)
    ax.plot(robot_x_coords, robot_y_coords, color='b', linewidth=3, solid_capstyle='round', label='Actual Robot Path')
    
    ax.scatter(obstacle_position[0], obstacle_position[1], c='r', s=200, marker='X', label='Dynamic Obstacle')
    
    ax.set_title("Real-World Navigation Simulation (San Francisco)")
    ax.legend()
    plt.show()
```
