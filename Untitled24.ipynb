```python
import heapq
import math
import matplotlib.pyplot as plt
import numpy as np

# ==============================================================================
# Part 1: Global Planner - Dijkstra's Algorithm
# ==============================================================================

def dijkstra_path(grid_map, start_xy, goal_xy):
    """
    Calculates the shortest path on a grid map using Dijkstra's algorithm.
    It operates with (x, y) coordinates, but accesses the map via [y][x].
    """
    print("\n--- Starting Global Planning using Dijkstra ---")
    rows, cols = len(grid_map), len(grid_map[0])
    # Possible moves (right, left, down, up) -> (dx, dy)
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)] 
    
    # Priority queue stores (cost, position)
    priority_queue = [(0, start_xy)]
    # Dictionary to store the cost to reach each node
    cost_so_far = {start_xy: 0}
    # Dictionary to backtrack the path
    came_from = {start_xy: None}

    while priority_queue:
        current_cost, current_pos_xy = heapq.heappop(priority_queue)

        if current_pos_xy == goal_xy:
            break # Goal reached

        if current_cost > cost_so_far.get(current_pos_xy, float('inf')):
            continue

        for dx, dy in directions:
            neighbor_pos_xy = (current_pos_xy[0] + dx, current_pos_xy[1] + dy)
            nx, ny = neighbor_pos_xy

            # Check if the neighbor is within bounds and is not a wall
            # IMPORTANT: Access the map via grid[y][x]
            if 0 <= nx < cols and 0 <= ny < rows and grid_map[ny][nx] == 0:
                new_cost = current_cost + 1 # Assuming cost of 1 for each step
                
                if new_cost < cost_so_far.get(neighbor_pos_xy, float('inf')):
                    cost_so_far[neighbor_pos_xy] = new_cost
                    came_from[neighbor_pos_xy] = current_pos_xy
                    heapq.heappush(priority_queue, (new_cost, neighbor_pos_xy))
    
    # Reconstruct the path from goal to start
    path = []
    current = goal_xy
    while current is not None:
        path.append(current)
        current = came_from.get(current)
    
    if not path or path[-1] != start_xy:
        print("!!! Global path not found !!!")
        return []

    path.reverse() # Reverse the path to go from start to goal
    print(f"Found a global path with {len(path)} waypoints.")
    return path

# ==============================================================================
# Part 2: Local Planner - Artificial Potential Field (APF) Algorithm
# ==============================================================================

class APF:
    """
    Calculates the total force (attractive + repulsive) to guide the robot.
    """
    def __init__(self, attractive_gain, repulsive_gain, obstacle_influence_radius):
        self.K_att = attractive_gain  # Gain for the attractive force
        self.K_rep = repulsive_gain   # Gain for the repulsive force
        self.rho0 = obstacle_influence_radius   # Radius of influence for obstacles (in meters/units)

    def compute_total_force(self, robot_xy, waypoint_xy, dynamic_obstacles_xy):
        # --- Calculate Attractive Force towards the waypoint ---
        att_force_x = self.K_att * (waypoint_xy[0] - robot_xy[0])
        att_force_y = self.K_att * (waypoint_xy[1] - robot_xy[1])

        total_rep_force_x = 0
        total_rep_force_y = 0

        # --- Calculate Repulsive Force from each dynamic obstacle ---
        for obs_x, obs_y in dynamic_obstacles_xy:
            dist_to_obs = math.hypot(robot_xy[0] - obs_x, robot_xy[1] - obs_y)
            
            # If the obstacle is within the influence radius, calculate a repulsive force
            if dist_to_obs < self.rho0 and dist_to_obs > 0.01: # Avoid division by zero
                # The magnitude of the force increases quadratically as the robot gets closer
                magnitude = self.K_rep * (1.0/dist_to_obs - 1.0/self.rho0) * (1.0/(dist_to_obs**2))
                
                # The direction of the force is directly away from the obstacle
                angle = math.atan2(robot_xy[1] - obs_y, robot_xy[0] - obs_x)
                total_rep_force_x += magnitude * math.cos(angle)
                total_rep_force_y += magnitude * math.sin(angle)
        
        # --- Sum the forces ---
        final_force_x = att_force_x + total_rep_force_x
        final_force_y = att_force_y + total_rep_force_y
        
        return final_force_x, final_force_y

# ==============================================================================
# Part 3: Visualization
# ==============================================================================
def plot_results(static_map, start_xy, goal_xy, global_path, robot_history_path, dynamic_obstacles_xy):
    """
    Plots the complete simulation results using matplotlib.
    """
    rows, cols = len(static_map), len(static_map[0])
    
    plt.figure(figsize=(8, 10))
    
    # 1. Plot the static map (walls and open space)
    # origin='lower' places the (0,0) coordinate at the bottom-left corner
    plt.imshow(static_map, cmap='Greys', origin='lower', interpolation='none', extent=[-0.5, cols-0.5, -0.5, rows-0.5])
    
    # 2. Plot the global path (Dijkstra)
    if global_path:
        path_x, path_y = zip(*global_path)
        plt.plot(path_x, path_y, 'g--', linewidth=2, label='Global Path (Dijkstra)')

    # 3. Plot the actual robot path (APF)
    if robot_history_path:
        robot_x, robot_y = zip(*robot_history_path)
        plt.plot(robot_x, robot_y, 'b-', linewidth=3, label='Actual Robot Path (APF)')

    # 4. Plot start, goal, and dynamic obstacles
    plt.plot(start_xy[0], start_xy[1], 'go', markersize=15, label='Start Point')
    plt.plot(goal_xy[0], goal_xy[1], 'r*', markersize=20, label='Goal Point')
    if dynamic_obstacles_xy:
        obs_x, obs_y = zip(*dynamic_obstacles_xy)
        plt.plot(obs_x, obs_y, 'rX', markersize=15, mew=3, label='Dynamic Obstacles')

    plt.title("Robot Navigation in a Blocked Corridor", fontsize=16)
    plt.xlabel("X Coordinate", fontsize=12)
    plt.ylabel("Y Coordinate", fontsize=12)
    plt.xticks(np.arange(0, cols, 1))
    plt.yticks(np.arange(0, rows, 1))
    plt.grid(True, which='both', linestyle=':', linewidth=0.5)
    plt.legend(fontsize=12)
    plt.axis('equal') # Ensure aspect ratio is equal for a true-to-scale view
    plt.show()

# ==============================================================================
# Main Simulation Program
# ==============================================================================
if __name__ == "__main__":
    
    # --- Challenging Map: A narrow corridor with a U-shaped detour ---
    STATIC_MAP = [
      # x=0  1  2  3  4
        [1, 1, 1, 1, 1], # y=0
        [1, 0, 1, 1, 1], # y=1
        [1, 0, 1, 1, 1], # y=2
        [1, 0, 0, 0, 1], # y=3
        [1, 1, 1, 0, 1], # y=4
        [1, 0, 0, 0, 1], # y=5
        [1, 0, 1, 1, 1], # y=6
        [1, 0, 1, 1, 1], # y=7
        [1, 0, 1, 1, 1], # y=8
        [1, 1, 1, 1, 1], # y=9
    ]
    START_POS_XY = (1, 1)
    GOAL_POS_XY = (1, 8)
    
    # --- Critical Obstacle: Placed to completely block the shortest path ---
    dynamic_obstacles_in_route_xy = [(1.0, 4.0)]
    
    # --- 1. Global Planning ---
    global_path = dijkstra_path(STATIC_MAP, START_POS_XY, GOAL_POS_XY)
    
    if global_path:
        print("Calculated Global Path (x, y):", global_path)
        print(f"\n--- Placing a critical dynamic obstacle at: {dynamic_obstacles_in_route_xy} ---")
        print("\n--- Starting Actual Motion Simulation ---")
        
        # --- 2. Initialize Simulation State ---
        robot_pos_xy = list(START_POS_XY)
        robot_path_history = [tuple(robot_pos_xy)]
        path_index = 1
        
        # --- Tuned Parameters: More aggressive repulsion for tight spaces ---
        apf_controller = APF(attractive_gain=1.0, repulsive_gain=3.0, obstacle_influence_radius=1.5)
        
        # --- 3. Run Simulation Loop ---
        for step in range(600): # Increased max steps for the longer detour
            if path_index >= len(global_path):
                print(f"Reached the final waypoint in {step} steps.")
                break

            current_waypoint_xy = global_path[path_index]
            
            # Calculate the net force on the robot
            force_x, force_y = apf_controller.compute_total_force(
                tuple(robot_pos_xy), 
                current_waypoint_xy,
                dynamic_obstacles_in_route_xy
            )
            
            # Move the robot one step in the direction of the calculated force
            magnitude = math.hypot(force_x, force_y)
            step_size = 0.1 # Very small step size for precision in tight corridors
            if magnitude > 0.01:
                robot_pos_xy[0] += (force_x / magnitude) * step_size
                robot_pos_xy[1] += (force_y / magnitude) * step_size
                robot_path_history.append(tuple(robot_pos_xy))

            # Check if the robot has reached the current waypoint
            dist_to_waypoint = math.hypot(robot_pos_xy[0] - current_waypoint_xy[0], robot_pos_xy[1] - current_waypoint_xy[1])
            if dist_to_waypoint < 0.2:
                print(f"--- Reached waypoint {current_waypoint_xy}. Moving to the next. ---")
                path_index += 1
        
        # --- 4. Plot the final results ---
        plot_results(STATIC_MAP, START_POS_XY, GOAL_POS_XY, global_path, robot_path_history, dynamic_obstacles_in_route_xy)

        # --- 5. Print final status message ---
        if path_index < len(global_path):
            print("\nSimulation finished, but the robot did not reach the final goal.")
        else:
            print("\nSuccess! The robot successfully navigated the blockage and reached the final goal.")
```
